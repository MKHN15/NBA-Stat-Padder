```python
def gameSim1():
    T1Score = ((rnd.gauss(t1_score1,t1a_drtg)) + rnd.gauss(t1_score2,t1a_drtg))/2
    T2Score = ((rnd.gauss(t2_score1,t2a_drtg)) + rnd.gauss(t2_score2,t2a_drtg))/2
    if int(round(T1Score)) > int(round(T2Score)):
        return 1
    elif int(round(T1Score)) < int(round(T2Score)):
        return -1
    else: return 0
```

```python
def gamesSim1(ns):
    gamesout = []
    team1win = 0
    team2win = 0
    for i in range(ns):
        gm = gameSim1()
        gamesout.append(gm)
        if gm == 1:
            team1win +=1 
        elif gm == -1:
            team2win +=1
    print('Team 1 Win ', round((team1win/(team1win+team2win))*100,2),'%')
    print('Team 2 Win ', round((team2win/(team1win+team2win))*100,2),'%')
    return gamesout
```
