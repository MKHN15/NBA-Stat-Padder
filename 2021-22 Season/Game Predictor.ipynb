```python
def gameSim():
    PHIScore = (rnd.gauss(PHI_score,PHI_sdpts)+ rnd.gauss(PHI_oppmeanpoints,PHI_oppstdpts))/2
    MILScore = (rnd.gauss(MIL_score,MIL_sdpts)+ rnd.gauss(MIL_oppmeanpoints,MIL_oppstdpts))/2
    if int(round(PHIScore)) > int(round(MILScore)):
        return 1
    elif int(round(PHIScore)) < int(round(MILScore)):
        return -1
    else: return 0
```

```python
def gamesSim(ns):
    gamesout = []
    team1win = 0
    team2win = 0
    for i in range(ns):
        gm = gameSim()
        gamesout.append(gm)
        if gm == 1:
            team1win +=1 
        elif gm == -1:
            team2win +=1
    print('PHI Win ', round((team1win/(team1win+team2win))*100,2),'%')
    print('MIL Win ', round((team2win/(team1win+team2win))*100,2),'%')
    return gamesout
```
