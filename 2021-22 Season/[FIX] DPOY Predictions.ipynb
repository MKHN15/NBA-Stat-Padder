#### Cleaning DataFrame ####

###### Cleaning Normal Stats ######

```python
#renames columns to strings only#
nba.rename(
    columns={
        'FG%': 'FGP',
        '3P': 'TP',
        '3PA': 'TPA',
        '3P%': 'TPP',
        '2P': 'TWOP',
        '2PA': 'TWOPA',
        '2P%': 'TWOPP',
        'eFG%': 'eFGP',
        'FT%': 'FTP'
    }, inplace=True)

#removing duplicate players (stats of players that played on different teams)#
nba.drop_duplicates(subset=['Player'], keep=False, inplace=True)
#Dropping "Rk" Column#
nba.drop('Rk', axis=1, inplace=True)

#replacing NaN with 0#
nba.fillna(0, inplace=True)

nba[["G", "GS","MP","FG","FGA","FGP","TP","TPA","TPP","TWOP","TWOPA","TWOPP","eFGP","FT","FTA","FTP","ORB","DRB","TRB","AST","STL","BLK","TOV","PF","PTS"]] = nba[["G", "GS","MP","FG","FGA","FGP","TP","TPA","TPP","TWOP","TWOPA","TWOPP","eFGP","FT","FTA","FTP","ORB","DRB","TRB","AST","STL","BLK","TOV","PF","PTS"]].apply(pd.to_numeric)
nba
```

###### Cleaning Advanced Stats ######

```python
#removing duplicate players (stats of players that played on different teams)#
advanced.drop_duplicates(subset=['Player'], keep=False, inplace=True)
#Dropping "Rk" Column#
advanced.drop('Rk', axis=1, inplace=True)

#replacing NaN with 0#
advanced.fillna(0, inplace=True)

advanced[['Age', 'G', 'MP', 'PER', 'TS%', '3PAr', 'FTr','ORB%', 'DRB%', 'TRB%', 'AST%', 'STL%', 'BLK%', 'TOV%', 'USG%','Unnamed: 19', 'OWS', 'DWS', 'WS', 'WS/48', 'Unnamed: 24', 'OBPM','DBPM', 'BPM', 'VORP']] = advanced[['Age', 'G', 'MP', 'PER', 'TS%', '3PAr', 'FTr','ORB%', 'DRB%', 'TRB%', 'AST%', 'STL%', 'BLK%', 'TOV%', 'USG%','Unnamed: 19', 'OWS', 'DWS', 'WS', 'WS/48', 'Unnamed: 24', 'OBPM','DBPM', 'BPM', 'VORP']].apply(pd.to_numeric)
advanced.drop(['Player','Pos','Age','Tm','G','MP','TS%','3PAr','FTr','ORB%','AST%','TOV%','USG%','OWS','WS/48','OBPM','VORP','Unnamed: 19','Unnamed: 24','PER','WS','BPM'], axis=1, inplace=True)

shorter_advanced = advanced
shorter_advanced
```

###### Cleaning Per 100 Possession Stats ######

```python
#removing duplicate players (stats of players that played on different teams)#
poss.drop_duplicates(subset=['Player'], keep=False, inplace=True)
#Dropping "Rk" Column#
poss.drop('Rk', axis=1, inplace=True)

#replacing NaN with 0#
poss.fillna(0, inplace=True)

poss.drop(['Player','Pos','Age','Tm','G','GS','MP','FG','FGA','FG%','3P','3PA','3P%','2P','2PA','2P%','FT','FTA','FT%','ORB','DRB','TRB','AST','STL','BLK','TOV','PF','PTS','ORtg','Unnamed: 29'], axis=1, inplace=True)
poss[['DRtg']] = poss[['DRtg']].apply(pd.to_numeric)
shorter_poss = poss
shorter_poss
```

```python
nba = nba.loc[~nba.index.duplicated(keep='first')]
shorter_advanced = shorter_advanced.loc[~shorter_advanced.index.duplicated(keep='first')]
shorter_poss = shorter_poss.loc[~shorter_poss.index.duplicated(keep='first')]
full = pd.concat([nba, shorter_advanced, shorter_poss], axis=1)
full
```

#### Filtering DataFrame ####

###### Filtering Normal Stats ######

```python
full.columns
```

```python
full.drop(['GS','MP','FG','FGA','FGP','TP','TPA','TPP','TWOP','TWOPA','TWOPP','eFGP','FT','FTA','FTP','ORB','AST','TOV','PTS'], axis=1, inplace=True)
```

```python
full.columns
```

```python
full[['Age','G','DRB','TRB','STL','BLK','PF','DRB%','TRB%','STL%','BLK%','DWS','DBPM','DRtg']] = full[['Age','G','DRB','TRB','STL','BLK','PF','DRB%','TRB%','STL%','BLK%','DWS','DBPM','DRtg']].apply(pd.to_numeric)
```

```python
full.drop(full[full['G'] <= 4].index, inplace = True)
full.drop(full[full['STL'] <= 0.1].index, inplace = True)
full.drop(full[full['BLK'] <= 0.1].index, inplace = True)
full.drop(full[full['DRB'] <= 0.5].index, inplace = True)
full.drop(full[full['DBPM'] <= 0.01].index, inplace = True)
full.drop(full[full['DWS'] <= 0.01].index, inplace = True)
```

```python
full
```

#### Processing the Data ####

```python
for index, row in full.iterrows():
    full.loc[index, "DPOY Score"] = row['DBPM']*((1/(1-row['DWS']))*(row['STL']*row['DRB']*row['BLK']*row['STL%']*row['BLK%']))/(row['DRtg'])
    
#((row['DRB%']*(row['DRB']/row['TRB']))+(row['STL']*(1+(row['STL%'])))+(row['BLK']*(1+(row['BLK%'])))-(row['PF']**4))*(((row['DWS']**3)*(row['DBPM']*2))*row['DRtg']))/((row['DBPM']-100)/row['DWS'])*-1#
#((((((((2*row['STL'])+row['BLK'])**2)+((row['DRB']/row['DRB%'])*row['TRB']))*(row['STL%']+row['BLK%']*(row['DRB%']/row['TRB%'])))*row['DWS'])/(row['DRtg']-100)))*(row['DWS']**2)/100#
```

```python
full.sort_values(by=['DPOY Score'], ascending=False, inplace=True)
#full.drop(['G','GS','MP','FG','FGA','FGP','TP','TPA','TPP','TWOP','TWOPA','TWOPP','eFGP','FT','FTA','FTP','ORB','DRB','TRB','AST','STL','BLK','TOV','PF','PTS','Avg. Box +/-'], axis=1, inplace=True)#
dpoy = full
dpoy['DPOY Score'] = round(dpoy['DPOY Score'],2)
dpoy = dpoy.head(15)
dpoy = dpoy.copy()
```
