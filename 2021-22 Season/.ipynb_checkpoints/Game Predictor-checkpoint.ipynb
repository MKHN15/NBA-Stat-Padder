```python
PHIdf = gdf[gdf.TEAM == 'PHI']
MILdf = gdf[gdf.TEAM == 'MIL']
```

```python
PHI_winpct = PHIdf.WINP.mean()
PHI_meanpts = PHIdf.PTS.mean()
PHI_fgp = PHIdf.FGP.mean()
PHI_tpp = PHIdf.TPP.mean()
PHI_ftp = PHIdf.FTP.mean()
PHI_dreb = PHIdf.DREB.mean()
PHI_ast = PHIdf.AST.mean()
PHI_tov = PHIdf.TOV.mean()
PHI_stl = PHIdf.STL.mean()
PHI_blk = PHIdf.BLK.mean()
PHI_pfd = PHIdf.PFD.mean()
PHI_offrtg = PHIdf.ORTG.mean()
PHI_defrtg = PHIdf.DRTG.mean()
PHI_clutchwinpct = PHIdf.CWINP.mean()

print("PHI Win% ", PHI_winpct)
print("PHI Avg. Points ", PHI_meanpts)
print("PHI FG% ", PHI_fgp)
print("PHI 3P% ", PHI_tpp)
print("PHI FT% ", PHI_ftp)
print("PHI Def. Reb. ", PHI_dreb)
print("PHI Avg. Assists ", PHI_ast)
print("PHI Avg. Turnovers ", PHI_tov)
print("PHI Avg. Steals ", PHI_stl)
print("PHI Avg. Blocks ", PHI_blk)
print("PHI Avg. Personal Fouls Drawn ", PHI_pfd)
print("PHI Off. Rtg ", PHI_offrtg)
print("PHI Def. Rtg ", PHI_defrtg)
print("PHI Clutch Win% ", PHI_clutchwinpct)
```

```python
MIL_winpct = MILdf.WINP.mean()
MIL_meanpts = MILdf.PTS.mean()
MIL_fgp = MILdf.FGP.mean()
MIL_tpp = MILdf.TPP.mean()
MIL_ftp = MILdf.FTP.mean()
MIL_dreb = MILdf.DREB.mean()
MIL_ast = MILdf.AST.mean()
MIL_tov = MILdf.TOV.mean()
MIL_stl = MILdf.STL.mean()
MIL_blk = MILdf.BLK.mean()
MIL_pfd = MILdf.PFD.mean()
MIL_offrtg = PHIdf.ORTG.mean()
MIL_defrtg = PHIdf.DRTG.mean()
MIL_clutchwinpct = PHIdf.CWINP.mean()

print("MIL Win% ", MIL_winpct)
print("MIL Avg. Points ", MIL_meanpts)
print("MIL FG% ", MIL_fgp)
print("MIL 3P% ", MIL_tpp)
print("MIL FT% ", MIL_ftp)
print("MIL Def. Reb. ", MIL_dreb)
print("MIL Avg. Assists ", MIL_ast)
print("MIL Avg. Turnovers ", MIL_tov)
print("MIL Avg. Steals ", MIL_stl)
print("MIL Avg. Blocks ", MIL_blk)
print("MIL Avg. Personal Fouls Drawn ", MIL_pfd)
print("MIL Off. Rtg ", MIL_offrtg)
print("MIL Def. Rtg ", MIL_defrtg)
print("MIL Clutch Win% ", MIL_clutchwinpct)
```

```python
PHI_scoring = PHI_meanpts*(PHI_fgp*PHI_tpp*PHI_ftp+PHI_offrtg)
PHI_misc = PHI_scoring+PHI_dreb+PHI_ast+PHI_stl+PHI_blk+PHI_pfd+PHI_defrtg
PHI_top = PHI_misc*PHI_winpct*PHI_clutchwinpct
PHI_score = PHI_top/(PHI_tov*10000)
PHI_score

print(PHI_score)
```

```python
MIL_scoring = MIL_meanpts*(MIL_fgp*MIL_tpp*MIL_ftp+MIL_offrtg)
MIL_misc = MIL_scoring+MIL_dreb+MIL_ast+MIL_stl+MIL_blk+MIL_pfd+MIL_defrtg
MIL_top = MIL_misc*MIL_winpct*MIL_clutchwinpct
MIL_score = MIL_top/(MIL_tov*10000)
MIL_score

print(MIL_score)
```

```python
PHI_meanpoints = PHI1df.TeamPTS.mean()
PHI_meanfgp = PHI1df.TeamFGP.mean()
PHI_sdpts = PHI1df.TeamPTS.std()
PHI_sdfgp = PHI1df.TeamFGP.std()
PHI_oppmeanpoints = PHI1df.OpponentPTS.mean()
PHI_oppfgp = PHI1df.OpponentFGP.mean()
PHI_oppstdpts = PHI1df.OpponentPTS.std()
PHI_oppstdfgp = PHI1df.OpponentFGP.std()

print("PHI Avg. Pts ", PHI_meanpoints)
print("PHI Avg. FG% ", PHI_meanfgp)
print("PHI Pts Deviation ", PHI_sdpts)
print("PHI FG% Deviation ", PHI_sdfgp)
print("PHI Opp. Avg. Pts ", PHI_oppmeanpoints)
print("PHI Opp. Avg. FG% ", PHI_oppfgp)
print("PHI Opp. Pts Deviation ", PHI_oppstdpts)
print("PHI Opp. FG% Deviation ", PHI_oppstdfgp)
```

```python
MIL_meanpoints = MIL1df.TeamPTS.mean()
MIL_meanfgp = MIL1df.TeamFGP.mean()
MIL_sdpts = MIL1df.TeamPTS.std()
MIL_sdfgp = MIL1df.TeamFGP.std()
MIL_oppmeanpoints = MIL1df.OpponentPTS.mean()
MIL_oppfgp = MIL1df.OpponentFGP.mean()
MIL_oppstdpts = MIL1df.OpponentPTS.std()
MIL_oppstdfgp = MIL1df.OpponentFGP.std()

print("MIL Avg. Pts ", MIL_meanpoints)
print("MIL Avg. FG% ", MIL_meanfgp)
print("MIL Pts Deviation ", MIL_sdpts)
print("MIL FG% Deviation ", MIL_sdfgp)
print("MIL Opp. Avg. Pts ", MIL_oppmeanpoints)
print("MIL Opp. Avg. FG% ", MIL_oppfgp)
print("MIL Opp. Pts Deviation ", MIL_oppstdpts)
print("MIL Opp. FG% Deviation ", MIL_oppstdfgp)
```

```python
def gameSim():
    PHIScore = (rnd.gauss(PHI_score,PHI_sdpts)+ rnd.gauss(PHI_oppmeanpoints,PHI_oppstdpts))/2
    MILScore = (rnd.gauss(MIL_score,MIL_sdpts)+ rnd.gauss(MIL_oppmeanpoints,MIL_oppstdpts))/2
    if int(round(PHIScore)) > int(round(MILScore)):
        return 1
    elif int(round(PHIScore)) < int(round(MILScore)):
        return -1
    else: return 0
```

```python
def gamesSim(ns):
    gamesout = []
    team1win = 0
    team2win = 0
    for i in range(ns):
        gm = gameSim()
        gamesout.append(gm)
        if gm == 1:
            team1win +=1 
        elif gm == -1:
            team2win +=1
    print('PHI Win ', round((team1win/(team1win+team2win))*100,2),'%')
    print('MIL Win ', round((team2win/(team1win+team2win))*100,2),'%')
    return gamesout
```

```python
gamesSim(10000)
```
