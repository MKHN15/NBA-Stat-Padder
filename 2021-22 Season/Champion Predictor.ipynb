#### Importing Data

```python
import pandas as pd 
import random as rnd
import numpy as np 
import matplotlib.pyplot as plt
from bs4 import BeautifulSoup
import requests
```

Team Stats

```python
url = 'https://www.basketball-reference.com/leagues/NBA_2021.html'
page = requests.get(url)
soup = BeautifulSoup(page.content, 'html.parser')
table = soup.find('table', id='per_game-team')
teamdf = pd.read_html(str(table))[0]
```

Team Opponent Stats

```python
url = 'https://www.basketball-reference.com/leagues/NBA_2021.html'
page = requests.get(url)
soup = BeautifulSoup(page.content, 'html.parser')
table = soup.find('table', id='per_game-opponent')
oppdf = pd.read_html(str(table))[0]
oppdf.columns
```

```python
oppdf.rename(
    columns={
        'Rk':'O_Rk',
        'Team':'O_Team', 
        'G':'O_G', 
        'MP':'O_MP', 
        'FG':'O_FG', 
        'FGA':'O_FGA', 
        'FG%':'O_FG%', 
        '3P':'O_3P', 
        '3PA':'O_3PA', 
        '3P%':'O_3P%', 
        '2P':'O_2P',
        '2PA':'O_2PA', 
        '2P%':'O_2P%', 
        'FT':'O_FT', 
        'FTA':'O_FTA', 
        'FT%':'O_FT%', 
        'ORB':'O_ORB', 
        'DRB':'O_DRB', 
        'TRB':'O_TRB', 
        'AST':'O_AST', 
        'STL':'O_STL',
        'BLK':'O_BLK', 
        'TOV':'O_TOV', 
        'PF':'O_PF', 
        'PTS':'O_PTS'
    }, inplace=True)
oppdf.drop(columns=['O_Rk', 'O_G','O_MP'], inplace=True)
```

Team Advanced Stats

```python
url = 'https://www.basketball-reference.com/leagues/NBA_2021.html'
page = requests.get(url)
soup = BeautifulSoup(page.content, 'html.parser')
table = soup.find('table', id='advanced-team')
advdf = pd.read_html(str(table))[0]
advdf.columns = advdf.columns.map('_'.join)
advdf.rename_axis('Rk').reset_index()
advdf.drop(columns=['Unnamed: 17_level_0_Unnamed: 17_level_1', 'Unnamed: 22_level_0_Unnamed: 22_level_1', 'Unnamed: 27_level_0_Unnamed: 27_level_1', 'Unnamed: 28_level_0_Arena', 'Unnamed: 29_level_0_Attend.', 'Unnamed: 30_level_0_Attend./G'], inplace=True)
advdf.rename(
    columns={
        'Unnamed: 0_level_0_Rk': 'A_Rk',
        'Unnamed: 1_level_0_Team': 'A_Team',
        'Unnamed: 2_level_0_Age': 'A_Age',
        'Unnamed: 3_level_0_W': 'A_W',
        'Unnamed: 4_level_0_L': 'A_L',
        'Unnamed: 5_level_0_PW': 'A_PW',
        'Unnamed: 6_level_0_PL': 'A_PL',
        'Unnamed: 7_level_0_MOV': 'A_MOV',
        'Unnamed: 8_level_0_SOS': 'A_SOS',
        'Unnamed: 9_level_0_SRS': 'A_SRS',
        'Unnamed: 10_level_0_ORtg': 'A_ORtg',
        'Unnamed: 11_level_0_DRtg': 'A_DRtg',
        'Unnamed: 12_level_0_NRtg': 'A_NRtg',
        'Unnamed: 13_level_0_Pace': 'A_Pace',
        'Unnamed: 14_level_0_FTr': 'A_FTr',
        'Unnamed: 15_level_0_3PAr': 'A_3PAr',
        'Unnamed: 16_level_0_TS%': 'A_TS%',
        'Offense Four Factors_eFG%': 'A_O_eFG%',
        'Offense Four Factors_TOV%': 'A_O_TOV%',
        'Offense Four Factors_ORB%': 'A_O_ORB%',
        'Offense Four Factors_FT/FGA': 'A_O_FT/FGA',
        'Defense Four Factors_eFG%': 'A_D_eFG%',
        'Defense Four Factors_TOV%': 'A_D_TOV%',
        'Defense Four Factors_DRB%': 'A_D_DRB%',
        'Defense Four Factors_FT/FGA': 'A_D_FT/FGA'
    }, inplace=True)
advdf.drop(columns=['A_Rk', 'A_Age','A_PW','A_PL'], inplace=True)
```

```python
teamdf.dropna(inplace=True)
oppdf.dropna(inplace=True)
advdf.dropna(inplace=True)

teamdf.sort_values(by=['Team'], ascending=True, inplace=True)
oppdf.sort_values(by=['O_Team'], ascending=True, inplace=True)
advdf.sort_values(by=['A_Team'], ascending=True, inplace=True)

td = teamdf
od = oppdf
ad = advdf
```

```python
full = pd.concat([td, od, ad], axis=1)
full
```
