```python
DAL_meanpoints = DAL1df.TeamPTS.mean()
DAL_meanfgp = DAL1df.TeamFGP.mean()
DAL_sdpts = DAL1df.TeamPTS.std()
DAL_sdfgp = DAL1df.TeamFGP.std()
DAL_oppmeanpoints = DAL1df.OpponentPTS.mean()
DAL_oppfgp = DAL1df.OpponentFGP.mean()
DAL_oppstdpts = DAL1df.OpponentPTS.std()
DAL_oppstdfgp = DAL1df.OpponentFGP.std()

print("DAL Avg. Pts ", DAL_meanpoints)
print("DAL Avg. FG% ", DAL_meanfgp)
print("DAL Pts Deviation ", DAL_sdpts)
print("DAL FG% Deviation ", DAL_sdfgp)
print("DAL Opp. Avg. Pts ", DAL_oppmeanpoints)
print("DAL Opp. Avg. FG% ", DAL_oppfgp)
print("DAL Opp. Pts Deviation ", DAL_oppstdpts)
print("DAL Opp. FG% Deviation ", DAL_oppstdfgp)
```

```python
OKC_meanpoints = OKC1df.TeamPTS.mean()
OKC_meanfgp = OKC1df.TeamFGP.mean()
OKC_sdpts = OKC1df.TeamPTS.std()
OKC_sdfgp = OKC1df.TeamFGP.std()
OKC_oppmeanpoints = OKC1df.OpponentPTS.mean()
OKC_oppfgp = OKC1df.OpponentFGP.mean()
OKC_oppstdpts = OKC1df.OpponentPTS.std()
OKC_oppstdfgp = OKC1df.OpponentFGP.std()

print("OKC Avg. Pts ", OKC_meanpoints)
print("OKC Avg. FG% ", OKC_meanfgp)
print("OKC Pts Deviation ", OKC_sdpts)
print("OKC FG% Deviation ", OKC_sdfgp)
print("OKC Opp. Avg. Pts ", OKC_oppmeanpoints)
print("OKC Opp. Avg. FG% ", OKC_oppfgp)
print("OKC Opp. Pts Deviation ", OKC_oppstdpts)
print("OKC Opp. FG% Deviation ", OKC_oppstdfgp)
```

```python
def gameSim():
    DALScore = (rnd.gauss(DAL_score,DAL_sdpts)+ rnd.gauss(DAL_oppmeanpoints,DAL_oppstdpts))/2
    OKCScore = (rnd.gauss(OKC_score,OKC_sdpts)+ rnd.gauss(OKC_oppmeanpoints,OKC_oppstdpts))/2
    if int(round(DALScore)) > int(round(OKCScore)):
        return 1
    elif int(round(DALScore)) < int(round(OKCScore)):
        return -1
    else: return 0
```

```python
def gamesSim(ns):
    gamesout = []
    team1win = 0
    team2win = 0
    for i in range(ns):
        gm = gameSim()
        gamesout.append(gm)
        if gm == 1:
            team1win +=1 
        elif gm == -1:
            team2win +=1
    print('DAL Win ', round((team1win/(team1win+team2win))*100,2),'%')
    print('OKC Win ', round((team2win/(team1win+team2win))*100,2),'%')
    return gamesout
```

```python
gamesSim(10000)
```
