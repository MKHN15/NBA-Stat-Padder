```python
import pandas as pd 
import random as rnd
import numpy as np 
import matplotlib.pyplot as plt
```

Importing Data

```python
html_url = "https://www.basketball-reference.com/leagues/NBA_2021_per_game.html"
```

```python
nba_tables = pd.read_html(html_url)
```

```python
len(nba_tables)
```

```python
nba = nba_tables[0]
```

```python
nba.head()
```

```python
nba.columns
```

Cleaning DataFrame

```python
#renames columns to strings only#
nba.rename(
    columns={
        'FG%': 'FGP',
        '3P': 'TP',
        '3PA': 'TPA',
        '3P%': 'TPP',
        '2P': 'TWOP',
        '2PA': 'TWOPA',
        '2P%': 'TWOPP',
        'eFG%': 'eFGP',
        'FT%': 'FTP'
    }, inplace=True)

#replacing NaN with 0#
nba.fillna(0)
```

```python
nba.drop_duplicates(subset=['Player'], keep='first', inplace=True)
nba.head()
```

```python
nba[["G", "GS","MP","FG","FGA","FGP","TP","TPA","TPP","TWOP","TWOPA","TWOPP","eFGP","FT","FTA","FTP","ORB","DRB","TRB","AST","STL","BLK","TOV","PF","PTS"]] = nba[["G", "GS","MP","FG","FGA","FGP","TP","TPA","TPP","TWOP","TWOPA","TWOPP","eFGP","FT","FTA","FTP","ORB","DRB","TRB","AST","STL","BLK","TOV","PF","PTS"]].apply(pd.to_numeric)
nba.loc[nba['G'] > 20]
nba.loc[nba['GS'] > 20]
nba.loc[nba['MP'] > 15]
nba.loc[nba['FG'] > 0.01]
nba.loc[nba['FGA'] > 0.01]
nba.loc[nba['FGP'] > 0.01]
nba.loc[nba['TP'] > 0.01]
nba.loc[nba['TPA'] > 0.01]
nba.loc[nba['TPP'] > 0.01]
nba.loc[nba['TWOP'] > 0.01]
nba.loc[nba['TWOPA'] > 0.01]
nba.loc[nba['TWOPP'] > 0.3]
nba.loc[nba['eFGP'] > 0.4]
nba.loc[nba['FT'] > 0.01]
nba.loc[nba['FTA'] > 0.01]
nba.loc[nba['FTP'] > 0.4]
nba.loc[nba['ORB'] > 0.01]
nba.loc[nba['DRB'] > 0.01]
nba.loc[nba['TRB'] > 0.01]
nba.loc[nba['AST'] > 0.01]
nba.loc[nba['STL'] > 0.01]
nba.loc[nba['BLK'] > 0.01]
nba.loc[nba['TOV'] > 0.01]
nba.loc[nba['PF'] > 0.01]
nba.loc[nba['PTS'] > 13]
```

Processing the Data

```python
'''
scoring = (FGP/TPP/FTP)*(PTS)/(1-eFGP)
key = scoring/(TRB*AST*STL*BLK)
score = ((key*(TOV/PF))
score
'''
```
