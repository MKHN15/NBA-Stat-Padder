```python
import pandas as pd 
import random as rnd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
gdf = pd.read_csv('24jan.csv')
```

```python
gdf.columns
```

```python
bam = gdf[gdf.Player == 'Bam Adebayo']

bam_winp = bam.WINP.mean()
bam_fgp = bam.FGP.mean()
bam_3pp = bam.TPP.mean()
bam_efgp = bam.eFGP.mean()
bam_ftp = bam.FTP.mean()
bam_reb = bam.TRB.mean()
bam_ast = bam.AST.mean()
bam_stl = bam.STL.mean()
bam_blk = bam.BLK.mean()
bam_tov = bam.TOV.mean()
bam_pf = bam.PF.mean()
bam_pts = bam.PTS.mean()

print("bam FG% ", bam_fgp)
print("bam 3P% ", bam_3pp)
print("bam eFG% ", bam_efgp)
print("bam FT% ", bam_ftp)
print("bam REB ", bam_reb)
print("bam AST ", bam_ast)
print("bam STL ", bam_stl)
print("bam BLK ", bam_blk)
print("bam TOV ", bam_tov)
print("bam PF ", bam_pf)
print("bam PTS ", bam_pts)

print("------------------------------------------------------")

bam_scoring = bam_pts*((bam_fgp*bam_3pp*bam_ftp)*(bam_efgp))
bam_key = bam_scoring*(bam_reb*bam_ast*bam_stl*bam_blk)
bam_score = (bam_key/(bam_tov*bam_pf))*bam_winp
bam_score

print("bam score ", bam_score)
```
