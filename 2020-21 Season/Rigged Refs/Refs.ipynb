```python
achiuwa = gdf[gdf.Player == 'Precious Achiuwa']

achiuwa_fgp = achiuwa.FGP.mean()
achiuwa_3pp = achiuwa.TPP.mean()
achiuwa_ftp = achiuwa.FTP.mean()
achiuwa_reb = achiuwa.RPG.mean()
achiuwa_ast = achiuwa.APG.mean()
achiuwa_stl = achiuwa.STL.mean()
achiuwa_blk = achiuwa.BLK.mean()
achiuwa_tov = achiuwa.TOV.mean()
achiuwa_pf = achiuwa.PF.mean()
achiuwa_pts = achiuwa.PPG.mean()

print("achiuwa FG% ", achiuwa_fgp)
print("achiuwa 3P% ", achiuwa_3pp)
print("achiuwa FT% ", achiuwa_ftp)
print("achiuwa REB ", achiuwa_reb)
print("achiuwa AST ", achiuwa_ast)
print("achiuwa STL ", achiuwa_stl)
print("achiuwa BLK ", achiuwa_blk)
print("achiuwa TOV ", achiuwa_tov)
print("achiuwa PF ", achiuwa_pf)
print("achiuwa PTS ", achiuwa_pts)

print("------------------------------------------------------")

achiuwa_scoring = ((achiuwa_fgp/achiuwa_ftp)*achiuwa_pts)+((achiuwa_3pp)*10)
achiuwa_key = achiuwa_scoring*((achiuwa_reb+achiuwa_ast)/(achiuwa_stl+achiuwa_blk))
achiuwa_score = (achiuwa_key*(achiuwa_tov/achiuwa_pf))*10

print("achiuwa score ", achiuwa_score)
```
