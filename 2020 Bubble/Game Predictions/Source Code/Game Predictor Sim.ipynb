```python
LAL_meanpoints = LAL1df.TeamPTS.mean()
LAL_meanfgp = LAL1df.TeamFGP.mean()
LAL_sdpts = LAL1df.TeamPTS.std()
LAL_sdfgp = LAL1df.TeamFGP.std()
LAL_oppmeanpoints = LAL1df.OpponentPTS.mean()
LAL_oppfgp = LAL1df.OpponentFGP.mean()
LAL_oppstdpts = LAL1df.OpponentPTS.std()
LAL_oppstdfgp = LAL1df.OpponentFGP.std()

print("LAL Avg. Pts ", LAL_meanpoints)
print("LAL Avg. FG% ", LAL_meanfgp)
print("LAL Pts Deviation ", LAL_sdpts)
print("LAL FG% Deviation ", LAL_sdfgp)
print("LAL Opp. Avg. Pts ", LAL_oppmeanpoints)
print("LAL Opp. Avg. FG% ", LAL_oppfgp)
print("LAL Opp. Pts Deviation ", LAL_oppstdpts)
print("LAL Opp. FG% Deviation ", LAL_oppstdfgp)
```

```python
MIL_meanpoints = MIL1df.TeamPTS.mean()
MIL_meanfgp = MIL1df.TeamFGP.mean()
MIL_sdpts = MIL1df.TeamPTS.std()
MIL_sdfgp = MIL1df.TeamFGP.std()
MIL_oppmeanpoints = MIL1df.OpponentPTS.mean()
MIL_oppfgp = MIL1df.OpponentFGP.mean()
MIL_oppstdpts = MIL1df.OpponentPTS.std()
MIL_oppstdfgp = MIL1df.OpponentFGP.std()

print("MIL Avg. Pts ", MIL_meanpoints)
print("MIL Avg. FG% ", MIL_meanfgp)
print("MIL Pts Deviation ", MIL_sdpts)
print("MIL FG% Deviation ", MIL_sdfgp)
print("MIL Opp. Avg. Pts ", MIL_oppmeanpoints)
print("MIL Opp. Avg. FG% ", MIL_oppfgp)
print("MIL Opp. Pts Deviation ", MIL_oppstdpts)
print("MIL Opp. FG% Deviation ", MIL_oppstdfgp)
```

```python
def gameSim():
    LALScore = (rnd.gauss(LAL_score,LAL_sdpts)+ rnd.gauss(LAL_oppmeanpoints,LAL_oppstdpts))/2
    MILScore = (rnd.gauss(MIL_score,MIL_sdpts)+ rnd.gauss(MIL_oppmeanpoints,MIL_oppstdpts))/2
    if int(round(LALScore)) > int(round(MILScore)):
        return 1
    elif int(round(LALScore)) < int(round(MILScore)):
        return -1
    else: return 0
```

```python
def gamesSim(ns):
    gamesout = []
    team1win = 0
    team2win = 0
    for i in range(ns):
        gm = gameSim()
        gamesout.append(gm)
        if gm == 1:
            team1win +=1 
        elif gm == -1:
            team2win +=1
    print('LAL Win ', round((team1win/(team1win+team2win))*100,2),'%')
    print('MIL Win ', round((team2win/(team1win+team2win))*100,2),'%')
    return gamesout
```
