```python
IND_meanpoints = IND1df.TeamPTS.mean()
IND_meanfgp = IND1df.TeamFGP.mean()
IND_sdpts = IND1df.TeamPTS.std()
IND_sdfgp = IND1df.TeamFGP.std()
IND_oppmeanpoints = IND1df.OpponentPTS.mean()
IND_oppfgp = IND1df.OpponentFGP.mean()
IND_oppstdpts = IND1df.OpponentPTS.std()
IND_oppstdfgp = IND1df.OpponentFGP.std()

print("IND Avg. Pts ", IND_meanpoints)
print("IND Avg. FG% ", IND_meanfgp)
print("IND Pts Deviation ", IND_sdpts)
print("IND FG% Deviation ", IND_sdfgp)
print("IND Opp. Avg. Pts ", IND_oppmeanpoints)
print("IND Opp. Avg. FG% ", IND_oppfgp)
print("IND Opp. Pts Deviation ", IND_oppstdpts)
print("IND Opp. FG% Deviation ", IND_oppstdfgp)
```

```python
PHX_meanpoints = PHX1df.TeamPTS.mean()
PHX_meanfgp = PHX1df.TeamFGP.mean()
PHX_sdpts = PHX1df.TeamPTS.std()
PHX_sdfgp = PHX1df.TeamFGP.std()
PHX_oppmeanpoints = PHX1df.OpponentPTS.mean()
PHX_oppfgp = PHX1df.OpponentFGP.mean()
PHX_oppstdpts = PHX1df.OpponentPTS.std()
PHX_oppstdfgp = PHX1df.OpponentFGP.std()

print("PHX Avg. Pts ", PHX_meanpoints)
print("PHX Avg. FG% ", PHX_meanfgp)
print("PHX Pts Deviation ", PHX_sdpts)
print("PHX FG% Deviation ", PHX_sdfgp)
print("PHX Opp. Avg. Pts ", PHX_oppmeanpoints)
print("PHX Opp. Avg. FG% ", PHX_oppfgp)
print("PHX Opp. Pts Deviation ", PHX_oppstdpts)
print("PHX Opp. FG% Deviation ", PHX_oppstdfgp)
```

```python
def gameSim():
    INDScore = (rnd.gauss(IND_score,IND_sdpts)+ rnd.gauss(IND_oppmeanpoints,IND_oppstdpts))/2
    PHXScore = (rnd.gauss(PHX_score,PHX_sdpts)+ rnd.gauss(PHX_oppmeanpoints,PHX_oppstdpts))/2
    if int(round(INDScore)) > int(round(PHXScore)):
        return 1
    elif int(round(INDScore)) < int(round(PHXScore)):
        return -1
    else: return 0
```

```python
def gamesSim(ns):
    gamesout = []
    team1win = 0
    team2win = 0
    for i in range(ns):
        gm = gameSim()
        gamesout.append(gm)
        if gm == 1:
            team1win +=1 
        elif gm == -1:
            team2win +=1
    print('IND Win ', round((team1win/(team1win+team2win))*100,2),'%')
    print('PHX Win ', round((team2win/(team1win+team2win))*100,2),'%')
    return gamesout
```

```python
gamesSim(10000)
```
