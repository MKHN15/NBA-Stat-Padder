```python
MIA_meanpoints = MIA1df.TeamPTS.mean()
MIA_meanfgp = MIA1df.TeamFGP.mean()
MIA_sdpts = MIA1df.TeamPTS.std()
MIA_sdfgp = MIA1df.TeamFGP.std()
MIA_oppmeanpoints = MIA1df.OpponentPTS.mean()
MIA_oppfgp = MIA1df.OpponentFGP.mean()
MIA_oppstdpts = MIA1df.OpponentPTS.std()
MIA_oppstdfgp = MIA1df.OpponentFGP.std()

print("MIA Avg. Pts ", MIA_meanpoints)
print("MIA Avg. FG% ", MIA_meanfgp)
print("MIA Pts Deviation ", MIA_sdpts)
print("MIA FG% Deviation ", MIA_sdfgp)
print("MIA Opp. Avg. Pts ", MIA_oppmeanpoints)
print("MIA Opp. Avg. FG% ", MIA_oppfgp)
print("MIA Opp. Pts Deviation ", MIA_oppstdpts)
print("MIA Opp. FG% Deviation ", MIA_oppstdfgp)
```

```python
MIL_meanpoints = MIL1df.TeamPTS.mean()
MIL_meanfgp = MIL1df.TeamFGP.mean()
MIL_sdpts = MIL1df.TeamPTS.std()
MIL_sdfgp = MIL1df.TeamFGP.std()
MIL_oppmeanpoints = MIL1df.OpponentPTS.mean()
MIL_oppfgp = MIL1df.OpponentFGP.mean()
MIL_oppstdpts = MIL1df.OpponentPTS.std()
MIL_oppstdfgp = MIL1df.OpponentFGP.std()

print("MIL Avg. Pts ", MIL_meanpoints)
print("MIL Avg. FG% ", MIL_meanfgp)
print("MIL Pts Deviation ", MIL_sdpts)
print("MIL FG% Deviation ", MIL_sdfgp)
print("MIL Opp. Avg. Pts ", MIL_oppmeanpoints)
print("MIL Opp. Avg. FG% ", MIL_oppfgp)
print("MIL Opp. Pts Deviation ", MIL_oppstdpts)
print("MIL Opp. FG% Deviation ", MIL_oppstdfgp)
```

```python
def gameSim():
    MIAScore = (rnd.gauss(MIA_score,MIA_sdpts)+ rnd.gauss(MIA_oppmeanpoints,MIA_oppstdpts))/2
    MILScore = (rnd.gauss(MIL_score,MIL_sdpts)+ rnd.gauss(MIL_oppmeanpoints,MIL_oppstdpts))/2
    if int(round(MIAScore)) > int(round(MILScore)):
        return 1
    elif int(round(MIAScore)) < int(round(MILScore)):
        return -1
    else: return 0
```

```python
def gamesSim(ns):
    gamesout = []
    team1win = 0
    team2win = 0
    for i in range(ns):
        gm = gameSim()
        gamesout.append(gm)
        if gm == 1:
            team1win +=1 
        elif gm == -1:
            team2win +=1
    print('MIA Win ', round((team1win/(team1win+team2win))*100,2),'%')
    print('MIL Win ', round((team2win/(team1win+team2win))*100,2),'%')
    return gamesout
```

```python
gamesSim(10000)
```
