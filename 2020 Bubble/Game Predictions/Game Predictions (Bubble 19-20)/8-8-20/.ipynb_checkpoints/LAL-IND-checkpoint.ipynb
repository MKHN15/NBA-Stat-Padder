```python
IND_meanpoints = IND1df.TeamPTS.mean()
IND_meanfgp = IND1df.TeamFGP.mean()
IND_sdpts = IND1df.TeamPTS.std()
IND_sdfgp = IND1df.TeamFGP.std()
IND_oppmeanpoints = IND1df.OpponentPTS.mean()
IND_oppfgp = IND1df.OpponentFGP.mean()
IND_oppstdpts = IND1df.OpponentPTS.std()
IND_oppstdfgp = IND1df.OpponentFGP.std()

print("IND Avg. Pts ", IND_meanpoints)
print("IND Avg. FG% ", IND_meanfgp)
print("IND Pts Deviation ", IND_sdpts)
print("IND FG% Deviation ", IND_sdfgp)
print("IND Opp. Avg. Pts ", IND_oppmeanpoints)
print("IND Opp. Avg. FG% ", IND_oppfgp)
print("IND Opp. Pts Deviation ", IND_oppstdpts)
print("IND Opp. FG% Deviation ", IND_oppstdfgp)
```

```python
LAL_meanpoints = LAL1df.TeamPTS.mean()
LAL_meanfgp = LAL1df.TeamFGP.mean()
LAL_sdpts = LAL1df.TeamPTS.std()
LAL_sdfgp = LAL1df.TeamFGP.std()
LAL_oppmeanpoints = LAL1df.OpponentPTS.mean()
LAL_oppfgp = LAL1df.OpponentFGP.mean()
LAL_oppstdpts = LAL1df.OpponentPTS.std()
LAL_oppstdfgp = LAL1df.OpponentFGP.std()

print("LAL Avg. Pts ", LAL_meanpoints)
print("LAL Avg. FG% ", LAL_meanfgp)
print("LAL Pts Deviation ", LAL_sdpts)
print("LAL FG% Deviation ", LAL_sdfgp)
print("LAL Opp. Avg. Pts ", LAL_oppmeanpoints)
print("LAL Opp. Avg. FG% ", LAL_oppfgp)
print("LAL Opp. Pts Deviation ", LAL_oppstdpts)
print("LAL Opp. FG% Deviation ", LAL_oppstdfgp)
```

```python
def gameSim():
    INDScore = (rnd.gauss(IND_score,IND_sdpts)+ rnd.gauss(IND_oppmeanpoints,IND_oppstdpts))/2
    LALScore = (rnd.gauss(LAL_score,LAL_sdpts)+ rnd.gauss(LAL_oppmeanpoints,LAL_oppstdpts))/2
    if int(round(INDScore)) > int(round(LALScore)):
        return 1
    elif int(round(INDScore)) < int(round(LALScore)):
        return -1
    else: return 0
```

```python
def gamesSim(ns):
    gamesout = []
    team1win = 0
    team2win = 0
    for i in range(ns):
        gm = gameSim()
        gamesout.append(gm)
        if gm == 1:
            team1win +=1 
        elif gm == -1:
            team2win +=1
    print('IND Win ', round((team1win/(team1win+team2win))*100,2),'%')
    print('LAL Win ', round((team2win/(team1win+team2win))*100,2),'%')
    return gamesout
```

```python
gamesSim(10000)
```
