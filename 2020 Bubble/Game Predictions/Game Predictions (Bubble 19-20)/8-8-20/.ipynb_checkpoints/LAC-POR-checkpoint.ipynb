```python
LAC_meanpoints = LAC1df.TeamPTS.mean()
LAC_meanfgp = LAC1df.TeamFGP.mean()
LAC_sdpts = LAC1df.TeamPTS.std()
LAC_sdfgp = LAC1df.TeamFGP.std()
LAC_oppmeanpoints = LAC1df.OpponentPTS.mean()
LAC_oppfgp = LAC1df.OpponentFGP.mean()
LAC_oppstdpts = LAC1df.OpponentPTS.std()
LAC_oppstdfgp = LAC1df.OpponentFGP.std()

print("LAC Avg. Pts ", LAC_meanpoints)
print("LAC Avg. FG% ", LAC_meanfgp)
print("LAC Pts Deviation ", LAC_sdpts)
print("LAC FG% Deviation ", LAC_sdfgp)
print("LAC Opp. Avg. Pts ", LAC_oppmeanpoints)
print("LAC Opp. Avg. FG% ", LAC_oppfgp)
print("LAC Opp. Pts Deviation ", LAC_oppstdpts)
print("LAC Opp. FG% Deviation ", LAC_oppstdfgp)
```

```python
POR_meanpoints = POR1df.TeamPTS.mean()
POR_meanfgp = POR1df.TeamFGP.mean()
POR_sdpts = POR1df.TeamPTS.std()
POR_sdfgp = POR1df.TeamFGP.std()
POR_oppmeanpoints = POR1df.OpponentPTS.mean()
POR_oppfgp = POR1df.OpponentFGP.mean()
POR_oppstdpts = POR1df.OpponentPTS.std()
POR_oppstdfgp = POR1df.OpponentFGP.std()

print("POR Avg. Pts ", POR_meanpoints)
print("POR Avg. FG% ", POR_meanfgp)
print("POR Pts Deviation ", POR_sdpts)
print("POR FG% Deviation ", POR_sdfgp)
print("POR Opp. Avg. Pts ", POR_oppmeanpoints)
print("POR Opp. Avg. FG% ", POR_oppfgp)
print("POR Opp. Pts Deviation ", POR_oppstdpts)
print("POR Opp. FG% Deviation ", POR_oppstdfgp)
```

```python
def gameSim():
    LACScore = (rnd.gauss(LAC_score,LAC_sdpts)+ rnd.gauss(LAC_oppmeanpoints,LAC_oppstdpts))/2
    PORScore = (rnd.gauss(POR_score,POR_sdpts)+ rnd.gauss(POR_oppmeanpoints,POR_oppstdpts))/2
    if int(round(LACScore)) > int(round(PORScore)):
        return 1
    elif int(round(LACScore)) < int(round(PORScore)):
        return -1
    else: return 0
```

```python
def gamesSim(ns):
    gamesout = []
    team1win = 0
    team2win = 0
    for i in range(ns):
        gm = gameSim()
        gamesout.append(gm)
        if gm == 1:
            team1win +=1 
        elif gm == -1:
            team2win +=1
    print('LAC Win ', round((team1win/(team1win+team2win))*100,2),'%')
    print('POR Win ', round((team2win/(team1win+team2win))*100,2),'%')
    return gamesout
```

```python
gamesSim(10000)
```
