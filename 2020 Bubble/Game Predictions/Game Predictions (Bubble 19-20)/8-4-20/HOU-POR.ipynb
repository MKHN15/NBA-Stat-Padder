```python
HOU_meanpoints = HOU1df.TeamPTS.mean()
HOU_meanfgp = HOU1df.TeamFGP.mean()
HOU_sdpts = HOU1df.TeamPTS.std()
HOU_sdfgp = HOU1df.TeamFGP.std()
HOU_oppmeanpoints = HOU1df.OpponentPTS.mean()
HOU_oppfgp = HOU1df.OpponentFGP.mean()
HOU_oppstdpts = HOU1df.OpponentPTS.std()
HOU_oppstdfgp = HOU1df.OpponentFGP.std()

print("HOU Avg. Pts ", HOU_meanpoints)
print("HOU Avg. FG% ", HOU_meanfgp)
print("HOU Pts Deviation ", HOU_sdpts)
print("HOU FG% Deviation ", HOU_sdfgp)
print("HOU Opp. Avg. Pts ", HOU_oppmeanpoints)
print("HOU Opp. Avg. FG% ", HOU_oppfgp)
print("HOU Opp. Pts Deviation ", HOU_oppstdpts)
print("HOU Opp. FG% Deviation ", HOU_oppstdfgp)
```

```python
POR_meanpoints = POR1df.TeamPTS.mean()
POR_meanfgp = POR1df.TeamFGP.mean()
POR_sdpts = POR1df.TeamPTS.std()
POR_sdfgp = POR1df.TeamFGP.std()
POR_oppmeanpoints = POR1df.OpponentPTS.mean()
POR_oppfgp = POR1df.OpponentFGP.mean()
POR_oppstdpts = POR1df.OpponentPTS.std()
POR_oppstdfgp = POR1df.OpponentFGP.std()

print("POR Avg. Pts ", POR_meanpoints)
print("POR Avg. FG% ", POR_meanfgp)
print("POR Pts Deviation ", POR_sdpts)
print("POR FG% Deviation ", POR_sdfgp)
print("POR Opp. Avg. Pts ", POR_oppmeanpoints)
print("POR Opp. Avg. FG% ", POR_oppfgp)
print("POR Opp. Pts Deviation ", POR_oppstdpts)
print("POR Opp. FG% Deviation ", POR_oppstdfgp)
```

```python
def gameSim():
    HOUScore = (rnd.gauss(HOU_score,HOU_sdpts)+ rnd.gauss(HOU_oppmeanpoints,HOU_oppstdpts))/2
    PORScore = (rnd.gauss(POR_score,POR_sdpts)+ rnd.gauss(POR_oppmeanpoints,POR_oppstdpts))/2
    if int(round(HOUScore)) > int(round(PORScore)):
        return 1
    elif int(round(HOUScore)) < int(round(PORScore)):
        return -1
    else: return 0
```

```python
def gamesSim(ns):
    gamesout = []
    team1win = 0
    team2win = 0
    for i in range(ns):
        gm = gameSim()
        gamesout.append(gm)
        if gm == 1:
            team1win +=1 
        elif gm == -1:
            team2win +=1
    print('HOU Win ', round((team1win/(team1win+team2win))*100,2),'%')
    print('POR Win ', round((team2win/(team1win+team2win))*100,2),'%')
    return gamesout
```

```python
gamesSim(10000)
```
