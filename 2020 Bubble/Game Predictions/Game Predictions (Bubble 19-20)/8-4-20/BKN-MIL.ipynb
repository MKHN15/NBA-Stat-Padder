```python
BKN_meanpoints = BKN1df.TeamPTS.mean()
BKN_meanfgp = BKN1df.TeamFGP.mean()
BKN_sdpts = BKN1df.TeamPTS.std()
BKN_sdfgp = BKN1df.TeamFGP.std()
BKN_oppmeanpoints = BKN1df.OpponentPTS.mean()
BKN_oppfgp = BKN1df.OpponentFGP.mean()
BKN_oppstdpts = BKN1df.OpponentPTS.std()
BKN_oppstdfgp = BKN1df.OpponentFGP.std()

print("BKN Avg. Pts ", BKN_meanpoints)
print("BKN Avg. FG% ", BKN_meanfgp)
print("BKN Pts Deviation ", BKN_sdpts)
print("BKN FG% Deviation ", BKN_sdfgp)
print("BKN Opp. Avg. Pts ", BKN_oppmeanpoints)
print("BKN Opp. Avg. FG% ", BKN_oppfgp)
print("BKN Opp. Pts Deviation ", BKN_oppstdpts)
print("BKN Opp. FG% Deviation ", BKN_oppstdfgp)
```

```python
MIL_meanpoints = MIL1df.TeamPTS.mean()
MIL_meanfgp = MIL1df.TeamFGP.mean()
MIL_sdpts = MIL1df.TeamPTS.std()
MIL_sdfgp = MIL1df.TeamFGP.std()
MIL_oppmeanpoints = MIL1df.OpponentPTS.mean()
MIL_oppfgp = MIL1df.OpponentFGP.mean()
MIL_oppstdpts = MIL1df.OpponentPTS.std()
MIL_oppstdfgp = MIL1df.OpponentFGP.std()

print("MIL Avg. Pts ", MIL_meanpoints)
print("MIL Avg. FG% ", MIL_meanfgp)
print("MIL Pts Deviation ", MIL_sdpts)
print("MIL FG% Deviation ", MIL_sdfgp)
print("MIL Opp. Avg. Pts ", MIL_oppmeanpoints)
print("MIL Opp. Avg. FG% ", MIL_oppfgp)
print("MIL Opp. Pts Deviation ", MIL_oppstdpts)
print("MIL Opp. FG% Deviation ", MIL_oppstdfgp)
```

```python
def gameSim():
    BKNScore = (rnd.gauss(BKN_score,BKN_sdpts)+ rnd.gauss(BKN_oppmeanpoints,BKN_oppstdpts))/2
    MILScore = (rnd.gauss(MIL_score,MIL_sdpts)+ rnd.gauss(MIL_oppmeanpoints,MIL_oppstdpts))/2
    if int(round(BKNScore)) > int(round(MILScore)):
        return 1
    elif int(round(BKNScore)) < int(round(MILScore)):
        return -1
    else: return 0
```

```python
def gamesSim(ns):
    gamesout = []
    team1win = 0
    team2win = 0
    for i in range(ns):
        gm = gameSim()
        gamesout.append(gm)
        if gm == 1:
            team1win +=1 
        elif gm == -1:
            team2win +=1
    print('BKN Win ', round((team1win/(team1win+team2win))*100,2),'%')
    print('MIL Win ', round((team2win/(team1win+team2win))*100,2),'%')
    return gamesout
```
