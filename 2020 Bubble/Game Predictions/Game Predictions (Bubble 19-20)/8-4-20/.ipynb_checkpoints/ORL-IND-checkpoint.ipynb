```python
IND_winpct = INDdf.WINP.mean()
IND_meanpts = INDdf.PTS.mean()
IND_fgp = INDdf.FGP.mean()
IND_tpp = INDdf.TPP.mean()
IND_ftp = INDdf.FTP.mean()
IND_dreb = INDdf.DREB.mean()
IND_ast = INDdf.AST.mean()
IND_tov = INDdf.TOV.mean()
IND_stl = INDdf.STL.mean()
IND_blk = INDdf.BLK.mean()
IND_pfd = INDdf.PFD.mean()
IND_offrtg = ORLdf.ORTG.mean()
IND_defrtg = ORLdf.DRTG.mean()
IND_clutchwinpct = ORLdf.CWINP.mean()

print("IND Win% ", IND_winpct)
print("IND Avg. Points ", IND_meanpts)
print("IND FG% ", IND_fgp)
print("IND 3P% ", IND_tpp)
print("IND FT% ", IND_ftp)
print("IND Def. Reb. ", IND_dreb)
print("IND Avg. Assists ", IND_ast)
print("IND Avg. Turnovers ", IND_tov)
print("IND Avg. Steals ", IND_stl)
print("IND Avg. Blocks ", IND_blk)
print("IND Avg. Personal Fouls Drawn ", IND_pfd)
print("IND Off. Rtg ", IND_offrtg)
print("IND Def. Rtg ", IND_defrtg)
print("IND Clutch Win% ", IND_clutchwinpct)
```

```python
ORL_scoring = ORL_meanpts*(ORL_fgp*ORL_tpp*ORL_ftp+ORL_offrtg)
ORL_misc = ORL_scoring+ORL_dreb+ORL_ast+ORL_stl+ORL_blk+ORL_pfd+ORL_defrtg
ORL_top = ORL_misc*ORL_winpct*ORL_clutchwinpct
ORL_score = ORL_top/(ORL_tov*10000)
ORL_score

print(ORL_score)
```

```python
ORL_meanpoints = ORL1df.TeamPTS.mean()
ORL_meanfgp = ORL1df.TeamFGP.mean()
ORL_sdpts = ORL1df.TeamPTS.std()
ORL_sdfgp = ORL1df.TeamFGP.std()
ORL_oppmeanpoints = ORL1df.OpponentPTS.mean()
ORL_oppfgp = ORL1df.OpponentFGP.mean()
ORL_oppstdpts = ORL1df.OpponentPTS.std()
ORL_oppstdfgp = ORL1df.OpponentFGP.std()

print("ORL Avg. Pts ", ORL_meanpoints)
print("ORL Avg. FG% ", ORL_meanfgp)
print("ORL Pts Deviation ", ORL_sdpts)
print("ORL FG% Deviation ", ORL_sdfgp)
print("ORL Opp. Avg. Pts ", ORL_oppmeanpoints)
print("ORL Opp. Avg. FG% ", ORL_oppfgp)
print("ORL Opp. Pts Deviation ", ORL_oppstdpts)
print("ORL Opp. FG% Deviation ", ORL_oppstdfgp)
```

```python
IND_meanpoints = IND1df.TeamPTS.mean()
IND_meanfgp = IND1df.TeamFGP.mean()
IND_sdpts = IND1df.TeamPTS.std()
IND_sdfgp = IND1df.TeamFGP.std()
IND_oppmeanpoints = IND1df.OpponentPTS.mean()
IND_oppfgp = IND1df.OpponentFGP.mean()
IND_oppstdpts = IND1df.OpponentPTS.std()
IND_oppstdfgp = IND1df.OpponentFGP.std()

print("IND Avg. Pts ", IND_meanpoints)
print("IND Avg. FG% ", IND_meanfgp)
print("IND Pts Deviation ", IND_sdpts)
print("IND FG% Deviation ", IND_sdfgp)
print("IND Opp. Avg. Pts ", IND_oppmeanpoints)
print("IND Opp. Avg. FG% ", IND_oppfgp)
print("IND Opp. Pts Deviation ", IND_oppstdpts)
print("IND Opp. FG% Deviation ", IND_oppstdfgp)
```

```python
def gameSim():
    ORLScore = (rnd.gauss(ORL_score,ORL_sdpts)+ rnd.gauss(ORL_oppmeanpoints,ORL_oppstdpts))/2
    INDScore = (rnd.gauss(IND_score,IND_sdpts)+ rnd.gauss(IND_oppmeanpoints,IND_oppstdpts))/2
    if int(round(ORLScore)) > int(round(INDScore)):
        return 1
    elif int(round(ORLScore)) < int(round(INDScore)):
        return -1
    else: return 0
```

```python
def gamesSim(ns):
    gamesout = []
    team1win = 0
    team2win = 0
    for i in range(ns):
        gm = gameSim()
        gamesout.append(gm)
        if gm == 1:
            team1win +=1 
        elif gm == -1:
            team2win +=1
    print('ORL Win ', round((team1win/(team1win+team2win))*100,2),'%')
    print('IND Win ', round((team2win/(team1win+team2win))*100,2),'%')
    return gamesout
```

```python
gamesSim(10000)
```
