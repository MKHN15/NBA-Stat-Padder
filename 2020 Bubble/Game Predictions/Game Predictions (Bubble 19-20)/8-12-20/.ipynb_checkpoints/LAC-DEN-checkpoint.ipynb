```python
LAC_winpct = LACdf.WINP.mean()
LAC_meanpts = LACdf.PTS.mean()
LAC_fgp = LACdf.FGP.mean()
LAC_tpp = LACdf.TPP.mean()
LAC_ftp = LACdf.FTP.mean()
LAC_dreb = LACdf.DREB.mean()
LAC_ast = LACdf.AST.mean()
LAC_tov = LACdf.TOV.mean()
LAC_stl = LACdf.STL.mean()
LAC_blk = LACdf.BLK.mean()
LAC_pfd = LACdf.PFD.mean()
LAC_offrtg = LACdf.ORTG.mean()
LAC_defrtg = LACdf.DRTG.mean()
LAC_clutchwinpct = LACdf.CWINP.mean()

print("LAC Win% ", LAC_winpct)
print("LAC Avg. Points ", LAC_meanpts)
print("LAC FG% ", LAC_fgp)
print("LAC 3P% ", LAC_tpp)
print("LAC FT% ", LAC_ftp)
print("LAC Def. Reb. ", LAC_dreb)
print("LAC Avg. Assists ", LAC_ast)
print("LAC Avg. Turnovers ", LAC_tov)
print("LAC Avg. Steals ", LAC_stl)
print("LAC Avg. Blocks ", LAC_blk)
print("LAC Avg. Personal Fouls Drawn ", LAC_pfd)
print("LAC Off. Rtg ", LAC_offrtg)
print("LAC Def. Rtg ", LAC_defrtg)
print("LAC Clutch Win% ", LAC_clutchwinpct)
```

```python
DEN_winpct = DENdf.WINP.mean()
DEN_meanpts = DENdf.PTS.mean()
DEN_fgp = DENdf.FGP.mean()
DEN_tpp = DENdf.TPP.mean()
DEN_ftp = DENdf.FTP.mean()
DEN_dreb = DENdf.DREB.mean()
DEN_ast = DENdf.AST.mean()
DEN_tov = DENdf.TOV.mean()
DEN_stl = DENdf.STL.mean()
DEN_blk = DENdf.BLK.mean()
DEN_pfd = DENdf.PFD.mean()
DEN_offrtg = LACdf.ORTG.mean()
DEN_defrtg = LACdf.DRTG.mean()
DEN_clutchwinpct = LACdf.CWINP.mean()

print("DEN Win% ", DEN_winpct)
print("DEN Avg. Points ", DEN_meanpts)
print("DEN FG% ", DEN_fgp)
print("DEN 3P% ", DEN_tpp)
print("DEN FT% ", DEN_ftp)
print("DEN Def. Reb. ", DEN_dreb)
print("DEN Avg. Assists ", DEN_ast)
print("DEN Avg. Turnovers ", DEN_tov)
print("DEN Avg. Steals ", DEN_stl)
print("DEN Avg. Blocks ", DEN_blk)
print("DEN Avg. Personal Fouls Drawn ", DEN_pfd)
print("DEN Off. Rtg ", DEN_offrtg)
print("DEN Def. Rtg ", DEN_defrtg)
print("DEN Clutch Win% ", DEN_clutchwinpct)
```

```python
LAC_scoring = LAC_meanpts*(LAC_fgp*LAC_tpp*LAC_ftp+LAC_offrtg)
LAC_misc = LAC_scoring+LAC_dreb+LAC_ast+LAC_stl+LAC_blk+LAC_pfd+LAC_defrtg
LAC_top = LAC_misc*LAC_winpct*LAC_clutchwinpct
LAC_score = LAC_top/(LAC_tov*10000)
LAC_score

print(LAC_score)
```

```python
DEN_scoring = DEN_meanpts*(DEN_fgp*DEN_tpp*DEN_ftp+DEN_offrtg)
DEN_misc = DEN_scoring+DEN_dreb+DEN_ast+DEN_stl+DEN_blk+DEN_pfd+DEN_defrtg
DEN_top = DEN_misc*DEN_winpct*DEN_clutchwinpct
DEN_score = DEN_top/(DEN_tov*10000)
DEN_score

print(DEN_score)
```

```python
LAC_meanpoints = LAC1df.TeamPTS.mean()
LAC_meanfgp = LAC1df.TeamFGP.mean()
LAC_sdpts = LAC1df.TeamPTS.std()
LAC_sdfgp = LAC1df.TeamFGP.std()
LAC_oppmeanpoints = LAC1df.OpponentPTS.mean()
LAC_oppfgp = LAC1df.OpponentFGP.mean()
LAC_oppstdpts = LAC1df.OpponentPTS.std()
LAC_oppstdfgp = LAC1df.OpponentFGP.std()

print("LAC Avg. Pts ", LAC_meanpoints)
print("LAC Avg. FG% ", LAC_meanfgp)
print("LAC Pts Deviation ", LAC_sdpts)
print("LAC FG% Deviation ", LAC_sdfgp)
print("LAC Opp. Avg. Pts ", LAC_oppmeanpoints)
print("LAC Opp. Avg. FG% ", LAC_oppfgp)
print("LAC Opp. Pts Deviation ", LAC_oppstdpts)
print("LAC Opp. FG% Deviation ", LAC_oppstdfgp)
```

```python
DEN_meanpoints = DEN1df.TeamPTS.mean()
DEN_meanfgp = DEN1df.TeamFGP.mean()
DEN_sdpts = DEN1df.TeamPTS.std()
DEN_sdfgp = DEN1df.TeamFGP.std()
DEN_oppmeanpoints = DEN1df.OpponentPTS.mean()
DEN_oppfgp = DEN1df.OpponentFGP.mean()
DEN_oppstdpts = DEN1df.OpponentPTS.std()
DEN_oppstdfgp = DEN1df.OpponentFGP.std()

print("DEN Avg. Pts ", DEN_meanpoints)
print("DEN Avg. FG% ", DEN_meanfgp)
print("DEN Pts Deviation ", DEN_sdpts)
print("DEN FG% Deviation ", DEN_sdfgp)
print("DEN Opp. Avg. Pts ", DEN_oppmeanpoints)
print("DEN Opp. Avg. FG% ", DEN_oppfgp)
print("DEN Opp. Pts Deviation ", DEN_oppstdpts)
print("DEN Opp. FG% Deviation ", DEN_oppstdfgp)
```

```python
def gameSim():
    LACScore = (rnd.gauss(LAC_score,LAC_sdpts)+ rnd.gauss(LAC_oppmeanpoints,LAC_oppstdpts))/2
    DENScore = (rnd.gauss(DEN_score,DEN_sdpts)+ rnd.gauss(DEN_oppmeanpoints,DEN_oppstdpts))/2
    if int(round(LACScore)) > int(round(DENScore)):
        return 1
    elif int(round(LACScore)) < int(round(DENScore)):
        return -1
    else: return 0
```

```python
def gamesSim(ns):
    gamesout = []
    team1win = 0
    team2win = 0
    for i in range(ns):
        gm = gameSim()
        gamesout.append(gm)
        if gm == 1:
            team1win +=1 
        elif gm == -1:
            team2win +=1
    print('LAC Win ', round((team1win/(team1win+team2win))*100,2),'%')
    print('DEN Win ', round((team2win/(team1win+team2win))*100,2),'%')
    return gamesout
```

```python
gamesSim(10000)
```
