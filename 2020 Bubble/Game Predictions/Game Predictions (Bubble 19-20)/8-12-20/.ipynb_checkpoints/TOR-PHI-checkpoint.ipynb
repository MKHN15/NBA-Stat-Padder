```python
PHI_winpct = PHIdf.WINP.mean()
PHI_meanpts = PHIdf.PTS.mean()
PHI_fgp = PHIdf.FGP.mean()
PHI_tpp = PHIdf.TPP.mean()
PHI_ftp = PHIdf.FTP.mean()
PHI_dreb = PHIdf.DREB.mean()
PHI_ast = PHIdf.AST.mean()
PHI_tov = PHIdf.TOV.mean()
PHI_stl = PHIdf.STL.mean()
PHI_blk = PHIdf.BLK.mean()
PHI_pfd = PHIdf.PFD.mean()
PHI_offrtg = TORdf.ORTG.mean()
PHI_defrtg = TORdf.DRTG.mean()
PHI_clutchwinpct = TORdf.CWINP.mean()

print("PHI Win% ", PHI_winpct)
print("PHI Avg. Points ", PHI_meanpts)
print("PHI FG% ", PHI_fgp)
print("PHI 3P% ", PHI_tpp)
print("PHI FT% ", PHI_ftp)
print("PHI Def. Reb. ", PHI_dreb)
print("PHI Avg. Assists ", PHI_ast)
print("PHI Avg. Turnovers ", PHI_tov)
print("PHI Avg. Steals ", PHI_stl)
print("PHI Avg. Blocks ", PHI_blk)
print("PHI Avg. Personal Fouls Drawn ", PHI_pfd)
print("PHI Off. Rtg ", PHI_offrtg)
print("PHI Def. Rtg ", PHI_defrtg)
print("PHI Clutch Win% ", PHI_clutchwinpct)
```

```python
TOR_scoring = TOR_meanpts*(TOR_fgp*TOR_tpp*TOR_ftp+TOR_offrtg)
TOR_misc = TOR_scoring+TOR_dreb+TOR_ast+TOR_stl+TOR_blk+TOR_pfd+TOR_defrtg
TOR_top = TOR_misc*TOR_winpct*TOR_clutchwinpct
TOR_score = TOR_top/(TOR_tov*10000)
TOR_score

print(TOR_score)
```

```python
PHI_scoring = PHI_meanpts*(PHI_fgp*PHI_tpp*PHI_ftp+PHI_offrtg)
PHI_misc = PHI_scoring+PHI_dreb+PHI_ast+PHI_stl+PHI_blk+PHI_pfd+PHI_defrtg
PHI_top = PHI_misc*PHI_winpct*PHI_clutchwinpct
PHI_score = PHI_top/(PHI_tov*10000)
PHI_score

print(PHI_score)
```

```python
TOR_meanpoints = TOR1df.TeamPTS.mean()
TOR_meanfgp = TOR1df.TeamFGP.mean()
TOR_sdpts = TOR1df.TeamPTS.std()
TOR_sdfgp = TOR1df.TeamFGP.std()
TOR_oppmeanpoints = TOR1df.OpponentPTS.mean()
TOR_oppfgp = TOR1df.OpponentFGP.mean()
TOR_oppstdpts = TOR1df.OpponentPTS.std()
TOR_oppstdfgp = TOR1df.OpponentFGP.std()

print("TOR Avg. Pts ", TOR_meanpoints)
print("TOR Avg. FG% ", TOR_meanfgp)
print("TOR Pts Deviation ", TOR_sdpts)
print("TOR FG% Deviation ", TOR_sdfgp)
print("TOR Opp. Avg. Pts ", TOR_oppmeanpoints)
print("TOR Opp. Avg. FG% ", TOR_oppfgp)
print("TOR Opp. Pts Deviation ", TOR_oppstdpts)
print("TOR Opp. FG% Deviation ", TOR_oppstdfgp)
```

```python
PHI_meanpoints = PHI1df.TeamPTS.mean()
PHI_meanfgp = PHI1df.TeamFGP.mean()
PHI_sdpts = PHI1df.TeamPTS.std()
PHI_sdfgp = PHI1df.TeamFGP.std()
PHI_oppmeanpoints = PHI1df.OpponentPTS.mean()
PHI_oppfgp = PHI1df.OpponentFGP.mean()
PHI_oppstdpts = PHI1df.OpponentPTS.std()
PHI_oppstdfgp = PHI1df.OpponentFGP.std()

print("PHI Avg. Pts ", PHI_meanpoints)
print("PHI Avg. FG% ", PHI_meanfgp)
print("PHI Pts Deviation ", PHI_sdpts)
print("PHI FG% Deviation ", PHI_sdfgp)
print("PHI Opp. Avg. Pts ", PHI_oppmeanpoints)
print("PHI Opp. Avg. FG% ", PHI_oppfgp)
print("PHI Opp. Pts Deviation ", PHI_oppstdpts)
print("PHI Opp. FG% Deviation ", PHI_oppstdfgp)
```

```python
def gameSim():
    TORScore = (rnd.gauss(TOR_score,TOR_sdpts)+ rnd.gauss(TOR_oppmeanpoints,TOR_oppstdpts))/2
    PHIScore = (rnd.gauss(PHI_score,PHI_sdpts)+ rnd.gauss(PHI_oppmeanpoints,PHI_oppstdpts))/2
    if int(round(TORScore)) > int(round(PHIScore)):
        return 1
    elif int(round(TORScore)) < int(round(PHIScore)):
        return -1
    else: return 0
```

```python
def gamesSim(ns):
    gamesout = []
    team1win = 0
    team2win = 0
    for i in range(ns):
        gm = gameSim()
        gamesout.append(gm)
        if gm == 1:
            team1win +=1 
        elif gm == -1:
            team2win +=1
    print('TOR Win ', round((team1win/(team1win+team2win))*100,2),'%')
    print('PHI Win ', round((team2win/(team1win+team2win))*100,2),'%')
    return gamesout
```

```python
gamesSim(10000)
```
