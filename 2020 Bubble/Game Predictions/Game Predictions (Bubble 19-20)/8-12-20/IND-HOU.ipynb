```python
IND_meanpoints = IND1df.TeamPTS.mean()
IND_meanfgp = IND1df.TeamFGP.mean()
IND_sdpts = IND1df.TeamPTS.std()
IND_sdfgp = IND1df.TeamFGP.std()
IND_oppmeanpoints = IND1df.OpponentPTS.mean()
IND_oppfgp = IND1df.OpponentFGP.mean()
IND_oppstdpts = IND1df.OpponentPTS.std()
IND_oppstdfgp = IND1df.OpponentFGP.std()

print("IND Avg. Pts ", IND_meanpoints)
print("IND Avg. FG% ", IND_meanfgp)
print("IND Pts Deviation ", IND_sdpts)
print("IND FG% Deviation ", IND_sdfgp)
print("IND Opp. Avg. Pts ", IND_oppmeanpoints)
print("IND Opp. Avg. FG% ", IND_oppfgp)
print("IND Opp. Pts Deviation ", IND_oppstdpts)
print("IND Opp. FG% Deviation ", IND_oppstdfgp)
```

```python
HOU_meanpoints = HOU1df.TeamPTS.mean()
HOU_meanfgp = HOU1df.TeamFGP.mean()
HOU_sdpts = HOU1df.TeamPTS.std()
HOU_sdfgp = HOU1df.TeamFGP.std()
HOU_oppmeanpoints = HOU1df.OpponentPTS.mean()
HOU_oppfgp = HOU1df.OpponentFGP.mean()
HOU_oppstdpts = HOU1df.OpponentPTS.std()
HOU_oppstdfgp = HOU1df.OpponentFGP.std()

print("HOU Avg. Pts ", HOU_meanpoints)
print("HOU Avg. FG% ", HOU_meanfgp)
print("HOU Pts Deviation ", HOU_sdpts)
print("HOU FG% Deviation ", HOU_sdfgp)
print("HOU Opp. Avg. Pts ", HOU_oppmeanpoints)
print("HOU Opp. Avg. FG% ", HOU_oppfgp)
print("HOU Opp. Pts Deviation ", HOU_oppstdpts)
print("HOU Opp. FG% Deviation ", HOU_oppstdfgp)
```

```python
def gameSim():
    INDScore = (rnd.gauss(IND_score,IND_sdpts)+ rnd.gauss(IND_oppmeanpoints,IND_oppstdpts))/2
    HOUScore = (rnd.gauss(HOU_score,HOU_sdpts)+ rnd.gauss(HOU_oppmeanpoints,HOU_oppstdpts))/2
    if int(round(INDScore)) > int(round(HOUScore)):
        return 1
    elif int(round(INDScore)) < int(round(HOUScore)):
        return -1
    else: return 0
```

```python
def gamesSim(ns):
    gamesout = []
    team1win = 0
    team2win = 0
    for i in range(ns):
        gm = gameSim()
        gamesout.append(gm)
        if gm == 1:
            team1win +=1 
        elif gm == -1:
            team2win +=1
    print('IND Win ', round((team1win/(team1win+team2win))*100,2),'%')
    print('HOU Win ', round((team2win/(team1win+team2win))*100,2),'%')
    return gamesout
```
