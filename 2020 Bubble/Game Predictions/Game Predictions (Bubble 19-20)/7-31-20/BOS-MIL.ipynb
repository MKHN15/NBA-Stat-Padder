```python
BOS_meanpoints = BOS1df.TeamPTS.mean()
BOS_meanfgp = BOS1df.TeamFGP.mean()
BOS_sdpts = BOS1df.TeamPTS.std()
BOS_sdfgp = BOS1df.TeamFGP.std()
BOS_oppmeanpoints = BOS1df.OpponentPTS.mean()
BOS_oppfgp = BOS1df.OpponentFGP.mean()
BOS_oppstdpts = BOS1df.OpponentPTS.std()
BOS_oppstdfgp = BOS1df.OpponentFGP.std()

print("BOS Avg. Pts ", BOS_meanpoints)
print("BOS Avg. FG% ", BOS_meanfgp)
print("BOS Pts Deviation ", BOS_sdpts)
print("BOS FG% Deviation ", BOS_sdfgp)
print("BOS Opp. Avg. Pts ", BOS_oppmeanpoints)
print("BOS Opp. Avg. FG% ", BOS_oppfgp)
print("BOS Opp. Pts Deviation ", BOS_oppstdpts)
print("BOS Opp. FG% Deviation ", BOS_oppstdfgp)
```

```python
MIL_meanpoints = MIL1df.TeamPTS.mean()
MIL_meanfgp = MIL1df.TeamFGP.mean()
MIL_sdpts = MIL1df.TeamPTS.std()
MIL_sdfgp = MIL1df.TeamFGP.std()
MIL_oppmeanpoints = MIL1df.OpponentPTS.mean()
MIL_oppfgp = MIL1df.OpponentFGP.mean()
MIL_oppstdpts = MIL1df.OpponentPTS.std()
MIL_oppstdfgp = MIL1df.OpponentFGP.std()

print("MIL Avg. Pts ", MIL_meanpoints)
print("MIL Avg. FG% ", MIL_meanfgp)
print("MIL Pts Deviation ", MIL_sdpts)
print("MIL FG% Deviation ", MIL_sdfgp)
print("MIL Opp. Avg. Pts ", MIL_oppmeanpoints)
print("MIL Opp. Avg. FG% ", MIL_oppfgp)
print("MIL Opp. Pts Deviation ", MIL_oppstdpts)
print("MIL Opp. FG% Deviation ", MIL_oppstdfgp)
```

```python
def gameSim():
    BOSScore = (rnd.gauss(BOS_score,BOS_sdpts)+ rnd.gauss(BOS_oppmeanpoints,BOS_oppstdpts))/2
    MILScore = (rnd.gauss(MIL_score,MIL_sdpts)+ rnd.gauss(MIL_oppmeanpoints,MIL_oppstdpts))/2
    if int(round(BOSScore)) > int(round(MILScore)):
        return 1
    elif int(round(BOSScore)) < int(round(MILScore)):
        return -1
    else: return 0
```

```python
def gamesSim(ns):
    gamesout = []
    team1win = 0
    team2win = 0
    for i in range(ns):
        gm = gameSim()
        gamesout.append(gm)
        if gm == 1:
            team1win +=1 
        elif gm == -1:
            team2win +=1
    print('BOS Win ', round((team1win/(team1win+team2win))*100,2),'%')
    print('MIL Win ', round((team2win/(team1win+team2win))*100,2),'%')
    return gamesout
```
