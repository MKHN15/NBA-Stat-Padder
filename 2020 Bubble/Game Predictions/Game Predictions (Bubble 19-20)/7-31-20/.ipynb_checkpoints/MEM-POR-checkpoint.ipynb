```python
MEMdf = gdf[gdf.TEAM == 'Memphis Grizzlies']
PORdf = gdf[gdf.TEAM == 'Portland Trail Blazers']
```

```python
MEM_winpct = MEMdf.WINP.mean()
MEM_meanpts = MEMdf.PTS.mean()
MEM_fgp = MEMdf.FGP.mean()
MEM_tpp = MEMdf.TPP.mean()
MEM_ftp = MEMdf.FTP.mean()
MEM_dreb = MEMdf.DREB.mean()
MEM_ast = MEMdf.AST.mean()
MEM_tov = MEMdf.TOV.mean()
MEM_stl = MEMdf.STL.mean()
MEM_blk = MEMdf.BLK.mean()
MEM_pfd = MEMdf.PFD.mean()
MEM_offrtg = MEMdf.ORTG.mean()
MEM_defrtg = MEMdf.DRTG.mean()
MEM_clutchwinpct = MEMdf.CWINP.mean()

print("MEM Win% ", MEM_winpct)
print("MEM Avg. Points ", MEM_meanpts)
print("MEM FG% ", MEM_fgp)
print("MEM 3P% ", MEM_tpp)
print("MEM FT% ", MEM_ftp)
print("MEM Def. Reb. ", MEM_dreb)
print("MEM Avg. Assists ", MEM_ast)
print("MEM Avg. Turnovers ", MEM_tov)
print("MEM Avg. Steals ", MEM_stl)
print("MEM Avg. Blocks ", MEM_blk)
print("MEM Avg. Personal Fouls Drawn ", MEM_pfd)
print("MEM Off. Rtg ", MEM_offrtg)
print("MEM Def. Rtg ", MEM_defrtg)
print("MEM Clutch Win% ", MEM_clutchwinpct)
```

```python
POR_winpct = PORdf.WINP.mean()
POR_meanpts = PORdf.PTS.mean()
POR_fgp = PORdf.FGP.mean()
POR_tpp = PORdf.TPP.mean()
POR_ftp = PORdf.FTP.mean()
POR_dreb = PORdf.DREB.mean()
POR_ast = PORdf.AST.mean()
POR_tov = PORdf.TOV.mean()
POR_stl = PORdf.STL.mean()
POR_blk = PORdf.BLK.mean()
POR_pfd = PORdf.PFD.mean()
POR_offrtg = MEMdf.ORTG.mean()
POR_defrtg = MEMdf.DRTG.mean()
POR_clutchwinpct = MEMdf.CWINP.mean()

print("POR Win% ", POR_winpct)
print("POR Avg. Points ", POR_meanpts)
print("POR FG% ", POR_fgp)
print("POR 3P% ", POR_tpp)
print("POR FT% ", POR_ftp)
print("POR Def. Reb. ", POR_dreb)
print("POR Avg. Assists ", POR_ast)
print("POR Avg. Turnovers ", POR_tov)
print("POR Avg. Steals ", POR_stl)
print("POR Avg. Blocks ", POR_blk)
print("POR Avg. Personal Fouls Drawn ", POR_pfd)
print("POR Off. Rtg ", POR_offrtg)
print("POR Def. Rtg ", POR_defrtg)
print("POR Clutch Win% ", POR_clutchwinpct)
```

```python
MEM_scoring = MEM_meanpts*(MEM_fgp*MEM_tpp*MEM_ftp+MEM_offrtg)
MEM_misc = MEM_scoring+MEM_dreb+MEM_ast+MEM_stl+MEM_blk+MEM_pfd+MEM_defrtg
MEM_top = MEM_misc*MEM_winpct*MEM_clutchwinpct
MEM_score = MEM_top/(MEM_tov*10000)
MEM_score

print(MEM_score)
```

```python
POR_scoring = POR_meanpts*(POR_fgp*POR_tpp*POR_ftp+POR_offrtg)
POR_misc = POR_scoring+POR_dreb+POR_ast+POR_stl+POR_blk+POR_pfd+POR_defrtg
POR_top = POR_misc*POR_winpct*POR_clutchwinpct
POR_score = POR_top/(POR_tov*10000)
POR_score

print(POR_score)
```

```python
MEM_meanpoints = MEM1df.TeamPTS.mean()
MEM_meanfgp = MEM1df.TeamFGP.mean()
MEM_sdpts = MEM1df.TeamPTS.std()
MEM_sdfgp = MEM1df.TeamFGP.std()
MEM_oppmeanpoints = MEM1df.OpponentPTS.mean()
MEM_oppfgp = MEM1df.OpponentFGP.mean()
MEM_oppstdpts = MEM1df.OpponentPTS.std()
MEM_oppstdfgp = MEM1df.OpponentFGP.std()

print("MEM Avg. Pts ", MEM_meanpoints)
print("MEM Avg. FG% ", MEM_meanfgp)
print("MEM Pts Deviation ", MEM_sdpts)
print("MEM FG% Deviation ", MEM_sdfgp)
print("MEM Opp. Avg. Pts ", MEM_oppmeanpoints)
print("MEM Opp. Avg. FG% ", MEM_oppfgp)
print("MEM Opp. Pts Deviation ", MEM_oppstdpts)
print("MEM Opp. FG% Deviation ", MEM_oppstdfgp)
```

```python
POR_meanpoints = POR1df.TeamPTS.mean()
POR_meanfgp = POR1df.TeamFGP.mean()
POR_sdpts = POR1df.TeamPTS.std()
POR_sdfgp = POR1df.TeamFGP.std()
POR_oppmeanpoints = POR1df.OpponentPTS.mean()
POR_oppfgp = POR1df.OpponentFGP.mean()
POR_oppstdpts = POR1df.OpponentPTS.std()
POR_oppstdfgp = POR1df.OpponentFGP.std()

print("POR Avg. Pts ", POR_meanpoints)
print("POR Avg. FG% ", POR_meanfgp)
print("POR Pts Deviation ", POR_sdpts)
print("POR FG% Deviation ", POR_sdfgp)
print("POR Opp. Avg. Pts ", POR_oppmeanpoints)
print("POR Opp. Avg. FG% ", POR_oppfgp)
print("POR Opp. Pts Deviation ", POR_oppstdpts)
print("POR Opp. FG% Deviation ", POR_oppstdfgp)
```

```python
def gameSim():
    MEMScore = (rnd.gauss(MEM_score,MEM_sdpts)+ rnd.gauss(MEM_oppmeanpoints,MEM_oppstdpts))/2
    PORScore = (rnd.gauss(POR_score,POR_sdpts)+ rnd.gauss(POR_oppmeanpoints,POR_oppstdpts))/2
    if int(round(MEMScore)) > int(round(PORScore)):
        return 1
    elif int(round(MEMScore)) < int(round(PORScore)):
        return -1
    else: return 0
```

```python
def gamesSim(ns):
    gamesout = []
    team1win = 0
    team2win = 0
    for i in range(ns):
        gm = gameSim()
        gamesout.append(gm)
        if gm == 1:
            team1win +=1 
        elif gm == -1:
            team2win +=1
    print('MEM Win ', round((team1win/(team1win+team2win))*100,2),'%')
    print('POR Win ', round((team2win/(team1win+team2win))*100,2),'%')
    return gamesout
```

```python
gamesSim(10000)
```
