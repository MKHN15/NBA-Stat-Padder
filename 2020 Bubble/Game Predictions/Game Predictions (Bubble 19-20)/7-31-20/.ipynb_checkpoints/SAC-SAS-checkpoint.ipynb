```python
SAC_meanpoints = SAC1df.TeamPTS.mean()
SAC_meanfgp = SAC1df.TeamFGP.mean()
SAC_sdpts = SAC1df.TeamPTS.std()
SAC_sdfgp = SAC1df.TeamFGP.std()
SAC_oppmeanpoints = SAC1df.OpponentPTS.mean()
SAC_oppfgp = SAC1df.OpponentFGP.mean()
SAC_oppstdpts = SAC1df.OpponentPTS.std()
SAC_oppstdfgp = SAC1df.OpponentFGP.std()

print("SAC Avg. Pts ", SAC_meanpoints)
print("SAC Avg. FG% ", SAC_meanfgp)
print("SAC Pts Deviation ", SAC_sdpts)
print("SAC FG% Deviation ", SAC_sdfgp)
print("SAC Opp. Avg. Pts ", SAC_oppmeanpoints)
print("SAC Opp. Avg. FG% ", SAC_oppfgp)
print("SAC Opp. Pts Deviation ", SAC_oppstdpts)
print("SAC Opp. FG% Deviation ", SAC_oppstdfgp)
```

```python
SAS_meanpoints = SAS1df.TeamPTS.mean()
SAS_meanfgp = SAS1df.TeamFGP.mean()
SAS_sdpts = SAS1df.TeamPTS.std()
SAS_sdfgp = SAS1df.TeamFGP.std()
SAS_oppmeanpoints = SAS1df.OpponentPTS.mean()
SAS_oppfgp = SAS1df.OpponentFGP.mean()
SAS_oppstdpts = SAS1df.OpponentPTS.std()
SAS_oppstdfgp = SAS1df.OpponentFGP.std()

print("SAS Avg. Pts ", SAS_meanpoints)
print("SAS Avg. FG% ", SAS_meanfgp)
print("SAS Pts Deviation ", SAS_sdpts)
print("SAS FG% Deviation ", SAS_sdfgp)
print("SAS Opp. Avg. Pts ", SAS_oppmeanpoints)
print("SAS Opp. Avg. FG% ", SAS_oppfgp)
print("SAS Opp. Pts Deviation ", SAS_oppstdpts)
print("SAS Opp. FG% Deviation ", SAS_oppstdfgp)
```

```python
def gameSim():
    SACScore = (rnd.gauss(SAC_score,SAC_sdpts)+ rnd.gauss(SAC_oppmeanpoints,SAC_oppstdpts))/2
    SASScore = (rnd.gauss(SAS_score,SAS_sdpts)+ rnd.gauss(SAS_oppmeanpoints,SAS_oppstdpts))/2
    if int(round(SACScore)) > int(round(SASScore)):
        return 1
    elif int(round(SACScore)) < int(round(SASScore)):
        return -1
    else: return 0
```

```python
def gamesSim(ns):
    gamesout = []
    team1win = 0
    team2win = 0
    for i in range(ns):
        gm = gameSim()
        gamesout.append(gm)
        if gm == 1:
            team1win +=1 
        elif gm == -1:
            team2win +=1
    print('SAC Win ', round((team1win/(team1win+team2win))*100,2),'%')
    print('SAS Win ', round((team2win/(team1win+team2win))*100,2),'%')
    return gamesout
```

```python
gamesSim(10000)
```
