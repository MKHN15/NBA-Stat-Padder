```python
PHXdf = gdf[gdf.TEAM == 'Phoenix Suns']
WASdf = gdf[gdf.TEAM == 'Washington Wizards']
```

```python
PHX_winpct = PHXdf.WINP.mean()
PHX_meanpts = PHXdf.PTS.mean()
PHX_fgp = PHXdf.FGP.mean()
PHX_tpp = PHXdf.TPP.mean()
PHX_ftp = PHXdf.FTP.mean()
PHX_dreb = PHXdf.DREB.mean()
PHX_ast = PHXdf.AST.mean()
PHX_tov = PHXdf.TOV.mean()
PHX_stl = PHXdf.STL.mean()
PHX_blk = PHXdf.BLK.mean()
PHX_pfd = PHXdf.PFD.mean()
PHX_offrtg = PHXdf.ORTG.mean()
PHX_defrtg = PHXdf.DRTG.mean()
PHX_clutchwinpct = PHXdf.CWINP.mean()

print("PHX Win% ", PHX_winpct)
print("PHX Avg. Points ", PHX_meanpts)
print("PHX FG% ", PHX_fgp)
print("PHX 3P% ", PHX_tpp)
print("PHX FT% ", PHX_ftp)
print("PHX Def. Reb. ", PHX_dreb)
print("PHX Avg. Assists ", PHX_ast)
print("PHX Avg. Turnovers ", PHX_tov)
print("PHX Avg. Steals ", PHX_stl)
print("PHX Avg. Blocks ", PHX_blk)
print("PHX Avg. Personal Fouls Drawn ", PHX_pfd)
print("PHX Off. Rtg ", PHX_offrtg)
print("PHX Def. Rtg ", PHX_defrtg)
print("PHX Clutch Win% ", PHX_clutchwinpct)
```

```python
WAS_winpct = WASdf.WINP.mean()
WAS_meanpts = WASdf.PTS.mean()
WAS_fgp = WASdf.FGP.mean()
WAS_tpp = WASdf.TPP.mean()
WAS_ftp = WASdf.FTP.mean()
WAS_dreb = WASdf.DREB.mean()
WAS_ast = WASdf.AST.mean()
WAS_tov = WASdf.TOV.mean()
WAS_stl = WASdf.STL.mean()
WAS_blk = WASdf.BLK.mean()
WAS_pfd = WASdf.PFD.mean()
WAS_offrtg = PHXdf.ORTG.mean()
WAS_defrtg = PHXdf.DRTG.mean()
WAS_clutchwinpct = PHXdf.CWINP.mean()

print("WAS Win% ", WAS_winpct)
print("WAS Avg. Points ", WAS_meanpts)
print("WAS FG% ", WAS_fgp)
print("WAS 3P% ", WAS_tpp)
print("WAS FT% ", WAS_ftp)
print("WAS Def. Reb. ", WAS_dreb)
print("WAS Avg. Assists ", WAS_ast)
print("WAS Avg. Turnovers ", WAS_tov)
print("WAS Avg. Steals ", WAS_stl)
print("WAS Avg. Blocks ", WAS_blk)
print("WAS Avg. Personal Fouls Drawn ", WAS_pfd)
print("WAS Off. Rtg ", WAS_offrtg)
print("WAS Def. Rtg ", WAS_defrtg)
print("WAS Clutch Win% ", WAS_clutchwinpct)
```

```python
PHX_scoring = PHX_meanpts*(PHX_fgp*PHX_tpp*PHX_ftp+PHX_offrtg)
PHX_misc = PHX_scoring+PHX_dreb+PHX_ast+PHX_stl+PHX_blk+PHX_pfd+PHX_defrtg
PHX_top = PHX_misc*PHX_winpct*PHX_clutchwinpct
PHX_score = PHX_top/(PHX_tov*10000)
PHX_score

print(PHX_score)
```

```python
WAS_scoring = WAS_meanpts*(WAS_fgp*WAS_tpp*WAS_ftp+WAS_offrtg)
WAS_misc = WAS_scoring+WAS_dreb+WAS_ast+WAS_stl+WAS_blk+WAS_pfd+WAS_defrtg
WAS_top = WAS_misc*WAS_winpct*WAS_clutchwinpct
WAS_score = WAS_top/(WAS_tov*10000)
WAS_score

print(WAS_score)
```

```python
PHX_meanpoints = PHX1df.TeamPTS.mean()
PHX_meanfgp = PHX1df.TeamFGP.mean()
PHX_sdpts = PHX1df.TeamPTS.std()
PHX_sdfgp = PHX1df.TeamFGP.std()
PHX_oppmeanpoints = PHX1df.OpponentPTS.mean()
PHX_oppfgp = PHX1df.OpponentFGP.mean()
PHX_oppstdpts = PHX1df.OpponentPTS.std()
PHX_oppstdfgp = PHX1df.OpponentFGP.std()

print("PHX Avg. Pts ", PHX_meanpoints)
print("PHX Avg. FG% ", PHX_meanfgp)
print("PHX Pts Deviation ", PHX_sdpts)
print("PHX FG% Deviation ", PHX_sdfgp)
print("PHX Opp. Avg. Pts ", PHX_oppmeanpoints)
print("PHX Opp. Avg. FG% ", PHX_oppfgp)
print("PHX Opp. Pts Deviation ", PHX_oppstdpts)
print("PHX Opp. FG% Deviation ", PHX_oppstdfgp)
```

```python
WAS_meanpoints = WAS1df.TeamPTS.mean()
WAS_meanfgp = WAS1df.TeamFGP.mean()
WAS_sdpts = WAS1df.TeamPTS.std()
WAS_sdfgp = WAS1df.TeamFGP.std()
WAS_oppmeanpoints = WAS1df.OpponentPTS.mean()
WAS_oppfgp = WAS1df.OpponentFGP.mean()
WAS_oppstdpts = WAS1df.OpponentPTS.std()
WAS_oppstdfgp = WAS1df.OpponentFGP.std()

print("WAS Avg. Pts ", WAS_meanpoints)
print("WAS Avg. FG% ", WAS_meanfgp)
print("WAS Pts Deviation ", WAS_sdpts)
print("WAS FG% Deviation ", WAS_sdfgp)
print("WAS Opp. Avg. Pts ", WAS_oppmeanpoints)
print("WAS Opp. Avg. FG% ", WAS_oppfgp)
print("WAS Opp. Pts Deviation ", WAS_oppstdpts)
print("WAS Opp. FG% Deviation ", WAS_oppstdfgp)
```

```python
def gameSim():
    PHXScore = (rnd.gauss(PHX_score,PHX_sdpts)+ rnd.gauss(PHX_oppmeanpoints,PHX_oppstdpts))/2
    WASScore = (rnd.gauss(WAS_score,WAS_sdpts)+ rnd.gauss(WAS_oppmeanpoints,WAS_oppstdpts))/2
    if int(round(PHXScore)) > int(round(WASScore)):
        return 1
    elif int(round(PHXScore)) < int(round(WASScore)):
        return -1
    else: return 0
```

```python
def gamesSim(ns):
    gamesout = []
    team1win = 0
    team2win = 0
    for i in range(ns):
        gm = gameSim()
        gamesout.append(gm)
        if gm == 1:
            team1win +=1 
        elif gm == -1:
            team2win +=1
    print('PHX Win ', round((team1win/(team1win+team2win))*100,2),'%')
    print('WAS Win ', round((team2win/(team1win+team2win))*100,2),'%')
    return gamesout
```

```python
gamesSim(10000)
```
