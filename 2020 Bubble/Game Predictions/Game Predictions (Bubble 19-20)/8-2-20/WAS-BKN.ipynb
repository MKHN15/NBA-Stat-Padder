```python
WAS_meanpoints = WAS1df.TeamPTS.mean()
WAS_meanfgp = WAS1df.TeamFGP.mean()
WAS_sdpts = WAS1df.TeamPTS.std()
WAS_sdfgp = WAS1df.TeamFGP.std()
WAS_oppmeanpoints = WAS1df.OpponentPTS.mean()
WAS_oppfgp = WAS1df.OpponentFGP.mean()
WAS_oppstdpts = WAS1df.OpponentPTS.std()
WAS_oppstdfgp = WAS1df.OpponentFGP.std()

print("WAS Avg. Pts ", WAS_meanpoints)
print("WAS Avg. FG% ", WAS_meanfgp)
print("WAS Pts Deviation ", WAS_sdpts)
print("WAS FG% Deviation ", WAS_sdfgp)
print("WAS Opp. Avg. Pts ", WAS_oppmeanpoints)
print("WAS Opp. Avg. FG% ", WAS_oppfgp)
print("WAS Opp. Pts Deviation ", WAS_oppstdpts)
print("WAS Opp. FG% Deviation ", WAS_oppstdfgp)
```

```python
BKN_meanpoints = BKN1df.TeamPTS.mean()
BKN_meanfgp = BKN1df.TeamFGP.mean()
BKN_sdpts = BKN1df.TeamPTS.std()
BKN_sdfgp = BKN1df.TeamFGP.std()
BKN_oppmeanpoints = BKN1df.OpponentPTS.mean()
BKN_oppfgp = BKN1df.OpponentFGP.mean()
BKN_oppstdpts = BKN1df.OpponentPTS.std()
BKN_oppstdfgp = BKN1df.OpponentFGP.std()

print("BKN Avg. Pts ", BKN_meanpoints)
print("BKN Avg. FG% ", BKN_meanfgp)
print("BKN Pts Deviation ", BKN_sdpts)
print("BKN FG% Deviation ", BKN_sdfgp)
print("BKN Opp. Avg. Pts ", BKN_oppmeanpoints)
print("BKN Opp. Avg. FG% ", BKN_oppfgp)
print("BKN Opp. Pts Deviation ", BKN_oppstdpts)
print("BKN Opp. FG% Deviation ", BKN_oppstdfgp)
```

```python
def gameSim():
    WASScore = (rnd.gauss(WAS_score,WAS_sdpts)+ rnd.gauss(WAS_oppmeanpoints,WAS_oppstdpts))/2
    BKNScore = (rnd.gauss(BKN_score,BKN_sdpts)+ rnd.gauss(BKN_oppmeanpoints,BKN_oppstdpts))/2
    if int(round(WASScore)) > int(round(BKNScore)):
        return 1
    elif int(round(WASScore)) < int(round(BKNScore)):
        return -1
    else: return 0
```

```python
def gamesSim(ns):
    gamesout = []
    team1win = 0
    team2win = 0
    for i in range(ns):
        gm = gameSim()
        gamesout.append(gm)
        if gm == 1:
            team1win +=1 
        elif gm == -1:
            team2win +=1
    print('WAS Win ', round((team1win/(team1win+team2win))*100,2),'%')
    print('BKN Win ', round((team2win/(team1win+team2win))*100,2),'%')
    return gamesout
```
