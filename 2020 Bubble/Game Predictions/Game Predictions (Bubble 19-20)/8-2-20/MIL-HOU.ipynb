```python
MIL_meanpoints = MIL1df.TeamPTS.mean()
MIL_meanfgp = MIL1df.TeamFGP.mean()
MIL_sdpts = MIL1df.TeamPTS.std()
MIL_sdfgp = MIL1df.TeamFGP.std()
MIL_oppmeanpoints = MIL1df.OpponentPTS.mean()
MIL_oppfgp = MIL1df.OpponentFGP.mean()
MIL_oppstdpts = MIL1df.OpponentPTS.std()
MIL_oppstdfgp = MIL1df.OpponentFGP.std()

print("MIL Avg. Pts ", MIL_meanpoints)
print("MIL Avg. FG% ", MIL_meanfgp)
print("MIL Pts Deviation ", MIL_sdpts)
print("MIL FG% Deviation ", MIL_sdfgp)
print("MIL Opp. Avg. Pts ", MIL_oppmeanpoints)
print("MIL Opp. Avg. FG% ", MIL_oppfgp)
print("MIL Opp. Pts Deviation ", MIL_oppstdpts)
print("MIL Opp. FG% Deviation ", MIL_oppstdfgp)
```

```python
HOU_meanpoints = HOU1df.TeamPTS.mean()
HOU_meanfgp = HOU1df.TeamFGP.mean()
HOU_sdpts = HOU1df.TeamPTS.std()
HOU_sdfgp = HOU1df.TeamFGP.std()
HOU_oppmeanpoints = HOU1df.OpponentPTS.mean()
HOU_oppfgp = HOU1df.OpponentFGP.mean()
HOU_oppstdpts = HOU1df.OpponentPTS.std()
HOU_oppstdfgp = HOU1df.OpponentFGP.std()

print("HOU Avg. Pts ", HOU_meanpoints)
print("HOU Avg. FG% ", HOU_meanfgp)
print("HOU Pts Deviation ", HOU_sdpts)
print("HOU FG% Deviation ", HOU_sdfgp)
print("HOU Opp. Avg. Pts ", HOU_oppmeanpoints)
print("HOU Opp. Avg. FG% ", HOU_oppfgp)
print("HOU Opp. Pts Deviation ", HOU_oppstdpts)
print("HOU Opp. FG% Deviation ", HOU_oppstdfgp)
```

```python
def gameSim():
    MILScore = (rnd.gauss(MIL_score,MIL_sdpts)+ rnd.gauss(MIL_oppmeanpoints,MIL_oppstdpts))/2
    HOUScore = (rnd.gauss(HOU_score,HOU_sdpts)+ rnd.gauss(HOU_oppmeanpoints,HOU_oppstdpts))/2
    if int(round(MILScore)) > int(round(HOUScore)):
        return 1
    elif int(round(MILScore)) < int(round(HOUScore)):
        return -1
    else: return 0
```

```python
def gamesSim(ns):
    gamesout = []
    team1win = 0
    team2win = 0
    for i in range(ns):
        gm = gameSim()
        gamesout.append(gm)
        if gm == 1:
            team1win +=1 
        elif gm == -1:
            team2win +=1
    print('MIL Win ', round((team1win/(team1win+team2win))*100,2),'%')
    print('HOU Win ', round((team2win/(team1win+team2win))*100,2),'%')
    return gamesout
```

```python
gamesSim(10000)
```
