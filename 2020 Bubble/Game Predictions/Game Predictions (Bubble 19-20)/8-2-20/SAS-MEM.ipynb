```python
SAS_meanpoints = SAS1df.TeamPTS.mean()
SAS_meanfgp = SAS1df.TeamFGP.mean()
SAS_sdpts = SAS1df.TeamPTS.std()
SAS_sdfgp = SAS1df.TeamFGP.std()
SAS_oppmeanpoints = SAS1df.OpponentPTS.mean()
SAS_oppfgp = SAS1df.OpponentFGP.mean()
SAS_oppstdpts = SAS1df.OpponentPTS.std()
SAS_oppstdfgp = SAS1df.OpponentFGP.std()

print("SAS Avg. Pts ", SAS_meanpoints)
print("SAS Avg. FG% ", SAS_meanfgp)
print("SAS Pts Deviation ", SAS_sdpts)
print("SAS FG% Deviation ", SAS_sdfgp)
print("SAS Opp. Avg. Pts ", SAS_oppmeanpoints)
print("SAS Opp. Avg. FG% ", SAS_oppfgp)
print("SAS Opp. Pts Deviation ", SAS_oppstdpts)
print("SAS Opp. FG% Deviation ", SAS_oppstdfgp)
```

```python
MEM_meanpoints = MEM1df.TeamPTS.mean()
MEM_meanfgp = MEM1df.TeamFGP.mean()
MEM_sdpts = MEM1df.TeamPTS.std()
MEM_sdfgp = MEM1df.TeamFGP.std()
MEM_oppmeanpoints = MEM1df.OpponentPTS.mean()
MEM_oppfgp = MEM1df.OpponentFGP.mean()
MEM_oppstdpts = MEM1df.OpponentPTS.std()
MEM_oppstdfgp = MEM1df.OpponentFGP.std()

print("MEM Avg. Pts ", MEM_meanpoints)
print("MEM Avg. FG% ", MEM_meanfgp)
print("MEM Pts Deviation ", MEM_sdpts)
print("MEM FG% Deviation ", MEM_sdfgp)
print("MEM Opp. Avg. Pts ", MEM_oppmeanpoints)
print("MEM Opp. Avg. FG% ", MEM_oppfgp)
print("MEM Opp. Pts Deviation ", MEM_oppstdpts)
print("MEM Opp. FG% Deviation ", MEM_oppstdfgp)
```

```python
def gameSim():
    SASScore = (rnd.gauss(SAS_score,SAS_sdpts)+ rnd.gauss(SAS_oppmeanpoints,SAS_oppstdpts))/2
    MEMScore = (rnd.gauss(MEM_score,MEM_sdpts)+ rnd.gauss(MEM_oppmeanpoints,MEM_oppstdpts))/2
    if int(round(SASScore)) > int(round(MEMScore)):
        return 1
    elif int(round(SASScore)) < int(round(MEMScore)):
        return -1
    else: return 0
```

```python
def gamesSim(ns):
    gamesout = []
    team1win = 0
    team2win = 0
    for i in range(ns):
        gm = gameSim()
        gamesout.append(gm)
        if gm == 1:
            team1win +=1 
        elif gm == -1:
            team2win +=1
    print('SAS Win ', round((team1win/(team1win+team2win))*100,2),'%')
    print('MEM Win ', round((team2win/(team1win+team2win))*100,2),'%')
    return gamesout
```
