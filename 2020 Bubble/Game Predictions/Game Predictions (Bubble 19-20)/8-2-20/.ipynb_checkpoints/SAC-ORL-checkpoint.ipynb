```python
SAC_meanpoints = SAC1df.TeamPTS.mean()
SAC_meanfgp = SAC1df.TeamFGP.mean()
SAC_sdpts = SAC1df.TeamPTS.std()
SAC_sdfgp = SAC1df.TeamFGP.std()
SAC_oppmeanpoints = SAC1df.OpponentPTS.mean()
SAC_oppfgp = SAC1df.OpponentFGP.mean()
SAC_oppstdpts = SAC1df.OpponentPTS.std()
SAC_oppstdfgp = SAC1df.OpponentFGP.std()

print("SAC Avg. Pts ", SAC_meanpoints)
print("SAC Avg. FG% ", SAC_meanfgp)
print("SAC Pts Deviation ", SAC_sdpts)
print("SAC FG% Deviation ", SAC_sdfgp)
print("SAC Opp. Avg. Pts ", SAC_oppmeanpoints)
print("SAC Opp. Avg. FG% ", SAC_oppfgp)
print("SAC Opp. Pts Deviation ", SAC_oppstdpts)
print("SAC Opp. FG% Deviation ", SAC_oppstdfgp)
```

```python
ORL_meanpoints = ORL1df.TeamPTS.mean()
ORL_meanfgp = ORL1df.TeamFGP.mean()
ORL_sdpts = ORL1df.TeamPTS.std()
ORL_sdfgp = ORL1df.TeamFGP.std()
ORL_oppmeanpoints = ORL1df.OpponentPTS.mean()
ORL_oppfgp = ORL1df.OpponentFGP.mean()
ORL_oppstdpts = ORL1df.OpponentPTS.std()
ORL_oppstdfgp = ORL1df.OpponentFGP.std()

print("ORL Avg. Pts ", ORL_meanpoints)
print("ORL Avg. FG% ", ORL_meanfgp)
print("ORL Pts Deviation ", ORL_sdpts)
print("ORL FG% Deviation ", ORL_sdfgp)
print("ORL Opp. Avg. Pts ", ORL_oppmeanpoints)
print("ORL Opp. Avg. FG% ", ORL_oppfgp)
print("ORL Opp. Pts Deviation ", ORL_oppstdpts)
print("ORL Opp. FG% Deviation ", ORL_oppstdfgp)
```

```python
def gameSim():
    SACScore = (rnd.gauss(SAC_score,SAC_sdpts)+ rnd.gauss(SAC_oppmeanpoints,SAC_oppstdpts))/2
    ORLScore = (rnd.gauss(ORL_score,ORL_sdpts)+ rnd.gauss(ORL_oppmeanpoints,ORL_oppstdpts))/2
    if int(round(SACScore)) > int(round(ORLScore)):
        return 1
    elif int(round(SACScore)) < int(round(ORLScore)):
        return -1
    else: return 0
```

```python
def gamesSim(ns):
    gamesout = []
    team1win = 0
    team2win = 0
    for i in range(ns):
        gm = gameSim()
        gamesout.append(gm)
        if gm == 1:
            team1win +=1 
        elif gm == -1:
            team2win +=1
    print('SAC Win ', round((team1win/(team1win+team2win))*100,2),'%')
    print('ORL Win ', round((team2win/(team1win+team2win))*100,2),'%')
    return gamesout
```

```python
gamesSim(10000)
```
