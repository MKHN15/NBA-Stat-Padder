```python
DAL_meanpoints = DAL1df.TeamPTS.mean()
DAL_meanfgp = DAL1df.TeamFGP.mean()
DAL_sdpts = DAL1df.TeamPTS.std()
DAL_sdfgp = DAL1df.TeamFGP.std()
DAL_oppmeanpoints = DAL1df.OpponentPTS.mean()
DAL_oppfgp = DAL1df.OpponentFGP.mean()
DAL_oppstdpts = DAL1df.OpponentPTS.std()
DAL_oppstdfgp = DAL1df.OpponentFGP.std()

print("DAL Avg. Pts ", DAL_meanpoints)
print("DAL Avg. FG% ", DAL_meanfgp)
print("DAL Pts Deviation ", DAL_sdpts)
print("DAL FG% Deviation ", DAL_sdfgp)
print("DAL Opp. Avg. Pts ", DAL_oppmeanpoints)
print("DAL Opp. Avg. FG% ", DAL_oppfgp)
print("DAL Opp. Pts Deviation ", DAL_oppstdpts)
print("DAL Opp. FG% Deviation ", DAL_oppstdfgp)
```

```python
PHX_meanpoints = PHX1df.TeamPTS.mean()
PHX_meanfgp = PHX1df.TeamFGP.mean()
PHX_sdpts = PHX1df.TeamPTS.std()
PHX_sdfgp = PHX1df.TeamFGP.std()
PHX_oppmeanpoints = PHX1df.OpponentPTS.mean()
PHX_oppfgp = PHX1df.OpponentFGP.mean()
PHX_oppstdpts = PHX1df.OpponentPTS.std()
PHX_oppstdfgp = PHX1df.OpponentFGP.std()

print("PHX Avg. Pts ", PHX_meanpoints)
print("PHX Avg. FG% ", PHX_meanfgp)
print("PHX Pts Deviation ", PHX_sdpts)
print("PHX FG% Deviation ", PHX_sdfgp)
print("PHX Opp. Avg. Pts ", PHX_oppmeanpoints)
print("PHX Opp. Avg. FG% ", PHX_oppfgp)
print("PHX Opp. Pts Deviation ", PHX_oppstdpts)
print("PHX Opp. FG% Deviation ", PHX_oppstdfgp)
```

```python
def gameSim():
    DALScore = (rnd.gauss(DAL_score,DAL_sdpts)+ rnd.gauss(DAL_oppmeanpoints,DAL_oppstdpts))/2
    PHXScore = (rnd.gauss(PHX_score,PHX_sdpts)+ rnd.gauss(PHX_oppmeanpoints,PHX_oppstdpts))/2
    if int(round(DALScore)) > int(round(PHXScore)):
        return 1
    elif int(round(DALScore)) < int(round(PHXScore)):
        return -1
    else: return 0
```

```python
def gamesSim(ns):
    gamesout = []
    team1win = 0
    team2win = 0
    for i in range(ns):
        gm = gameSim()
        gamesout.append(gm)
        if gm == 1:
            team1win +=1 
        elif gm == -1:
            team2win +=1
    print('DAL Win ', round((team1win/(team1win+team2win))*100,2),'%')
    print('PHX Win ', round((team2win/(team1win+team2win))*100,2),'%')
    return gamesout
```

```python
gamesSim(10000)
```
