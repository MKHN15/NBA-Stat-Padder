```python
LAC_meanpoints = LAC1df.TeamPTS.mean()
LAC_meanfgp = LAC1df.TeamFGP.mean()
LAC_sdpts = LAC1df.TeamPTS.std()
LAC_sdfgp = LAC1df.TeamFGP.std()
LAC_oppmeanpoints = LAC1df.OpponentPTS.mean()
LAC_oppfgp = LAC1df.OpponentFGP.mean()
LAC_oppstdpts = LAC1df.OpponentPTS.std()
LAC_oppstdfgp = LAC1df.OpponentFGP.std()

print("LAC Avg. Pts ", LAC_meanpoints)
print("LAC Avg. FG% ", LAC_meanfgp)
print("LAC Pts Deviation ", LAC_sdpts)
print("LAC FG% Deviation ", LAC_sdfgp)
print("LAC Opp. Avg. Pts ", LAC_oppmeanpoints)
print("LAC Opp. Avg. FG% ", LAC_oppfgp)
print("LAC Opp. Pts Deviation ", LAC_oppstdpts)
print("LAC Opp. FG% Deviation ", LAC_oppstdfgp)
```

```python
OKC_meanpoints = OKC1df.TeamPTS.mean()
OKC_meanfgp = OKC1df.TeamFGP.mean()
OKC_sdpts = OKC1df.TeamPTS.std()
OKC_sdfgp = OKC1df.TeamFGP.std()
OKC_oppmeanpoints = OKC1df.OpponentPTS.mean()
OKC_oppfgp = OKC1df.OpponentFGP.mean()
OKC_oppstdpts = OKC1df.OpponentPTS.std()
OKC_oppstdfgp = OKC1df.OpponentFGP.std()

print("OKC Avg. Pts ", OKC_meanpoints)
print("OKC Avg. FG% ", OKC_meanfgp)
print("OKC Pts Deviation ", OKC_sdpts)
print("OKC FG% Deviation ", OKC_sdfgp)
print("OKC Opp. Avg. Pts ", OKC_oppmeanpoints)
print("OKC Opp. Avg. FG% ", OKC_oppfgp)
print("OKC Opp. Pts Deviation ", OKC_oppstdpts)
print("OKC Opp. FG% Deviation ", OKC_oppstdfgp)
```

```python
def gameSim():
    LACScore = (rnd.gauss(LAC_score,LAC_sdpts)+ rnd.gauss(LAC_oppmeanpoints,LAC_oppstdpts))/2
    OKCScore = (rnd.gauss(OKC_score,OKC_sdpts)+ rnd.gauss(OKC_oppmeanpoints,OKC_oppstdpts))/2
    if int(round(LACScore)) > int(round(OKCScore)):
        return 1
    elif int(round(LACScore)) < int(round(OKCScore)):
        return -1
    else: return 0
```

```python
def gamesSim(ns):
    gamesout = []
    team1win = 0
    team2win = 0
    for i in range(ns):
        gm = gameSim()
        gamesout.append(gm)
        if gm == 1:
            team1win +=1 
        elif gm == -1:
            team2win +=1
    print('LAC Win ', round((team1win/(team1win+team2win))*100,2),'%')
    print('OKC Win ', round((team2win/(team1win+team2win))*100,2),'%')
    return gamesout
```

```python
gamesSim(10000)
```
