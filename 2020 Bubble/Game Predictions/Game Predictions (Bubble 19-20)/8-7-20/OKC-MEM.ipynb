```python
OKC_meanpoints = OKC1df.TeamPTS.mean()
OKC_meanfgp = OKC1df.TeamFGP.mean()
OKC_sdpts = OKC1df.TeamPTS.std()
OKC_sdfgp = OKC1df.TeamFGP.std()
OKC_oppmeanpoints = OKC1df.OpponentPTS.mean()
OKC_oppfgp = OKC1df.OpponentFGP.mean()
OKC_oppstdpts = OKC1df.OpponentPTS.std()
OKC_oppstdfgp = OKC1df.OpponentFGP.std()

print("OKC Avg. Pts ", OKC_meanpoints)
print("OKC Avg. FG% ", OKC_meanfgp)
print("OKC Pts Deviation ", OKC_sdpts)
print("OKC FG% Deviation ", OKC_sdfgp)
print("OKC Opp. Avg. Pts ", OKC_oppmeanpoints)
print("OKC Opp. Avg. FG% ", OKC_oppfgp)
print("OKC Opp. Pts Deviation ", OKC_oppstdpts)
print("OKC Opp. FG% Deviation ", OKC_oppstdfgp)
```

```python
MEM_meanpoints = MEM1df.TeamPTS.mean()
MEM_meanfgp = MEM1df.TeamFGP.mean()
MEM_sdpts = MEM1df.TeamPTS.std()
MEM_sdfgp = MEM1df.TeamFGP.std()
MEM_oppmeanpoints = MEM1df.OpponentPTS.mean()
MEM_oppfgp = MEM1df.OpponentFGP.mean()
MEM_oppstdpts = MEM1df.OpponentPTS.std()
MEM_oppstdfgp = MEM1df.OpponentFGP.std()

print("MEM Avg. Pts ", MEM_meanpoints)
print("MEM Avg. FG% ", MEM_meanfgp)
print("MEM Pts Deviation ", MEM_sdpts)
print("MEM FG% Deviation ", MEM_sdfgp)
print("MEM Opp. Avg. Pts ", MEM_oppmeanpoints)
print("MEM Opp. Avg. FG% ", MEM_oppfgp)
print("MEM Opp. Pts Deviation ", MEM_oppstdpts)
print("MEM Opp. FG% Deviation ", MEM_oppstdfgp)
```

```python
def gameSim():
    OKCScore = (rnd.gauss(OKC_score,OKC_sdpts)+ rnd.gauss(OKC_oppmeanpoints,OKC_oppstdpts))/2
    MEMScore = (rnd.gauss(MEM_score,MEM_sdpts)+ rnd.gauss(MEM_oppmeanpoints,MEM_oppstdpts))/2
    if int(round(OKCScore)) > int(round(MEMScore)):
        return 1
    elif int(round(OKCScore)) < int(round(MEMScore)):
        return -1
    else: return 0
```

```python
def gamesSim(ns):
    gamesout = []
    team1win = 0
    team2win = 0
    for i in range(ns):
        gm = gameSim()
        gamesout.append(gm)
        if gm == 1:
            team1win +=1 
        elif gm == -1:
            team2win +=1
    print('OKC Win ', round((team1win/(team1win+team2win))*100,2),'%')
    print('MEM Win ', round((team2win/(team1win+team2win))*100,2),'%')
    return gamesout
```
