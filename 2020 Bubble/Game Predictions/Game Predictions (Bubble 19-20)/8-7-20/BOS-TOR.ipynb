```python
BOS_meanpoints = BOS1df.TeamPTS.mean()
BOS_meanfgp = BOS1df.TeamFGP.mean()
BOS_sdpts = BOS1df.TeamPTS.std()
BOS_sdfgp = BOS1df.TeamFGP.std()
BOS_oppmeanpoints = BOS1df.OpponentPTS.mean()
BOS_oppfgp = BOS1df.OpponentFGP.mean()
BOS_oppstdpts = BOS1df.OpponentPTS.std()
BOS_oppstdfgp = BOS1df.OpponentFGP.std()

print("BOS Avg. Pts ", BOS_meanpoints)
print("BOS Avg. FG% ", BOS_meanfgp)
print("BOS Pts Deviation ", BOS_sdpts)
print("BOS FG% Deviation ", BOS_sdfgp)
print("BOS Opp. Avg. Pts ", BOS_oppmeanpoints)
print("BOS Opp. Avg. FG% ", BOS_oppfgp)
print("BOS Opp. Pts Deviation ", BOS_oppstdpts)
print("BOS Opp. FG% Deviation ", BOS_oppstdfgp)
```

```python
TOR_meanpoints = TOR1df.TeamPTS.mean()
TOR_meanfgp = TOR1df.TeamFGP.mean()
TOR_sdpts = TOR1df.TeamPTS.std()
TOR_sdfgp = TOR1df.TeamFGP.std()
TOR_oppmeanpoints = TOR1df.OpponentPTS.mean()
TOR_oppfgp = TOR1df.OpponentFGP.mean()
TOR_oppstdpts = TOR1df.OpponentPTS.std()
TOR_oppstdfgp = TOR1df.OpponentFGP.std()

print("TOR Avg. Pts ", TOR_meanpoints)
print("TOR Avg. FG% ", TOR_meanfgp)
print("TOR Pts Deviation ", TOR_sdpts)
print("TOR FG% Deviation ", TOR_sdfgp)
print("TOR Opp. Avg. Pts ", TOR_oppmeanpoints)
print("TOR Opp. Avg. FG% ", TOR_oppfgp)
print("TOR Opp. Pts Deviation ", TOR_oppstdpts)
print("TOR Opp. FG% Deviation ", TOR_oppstdfgp)
```

```python
def gameSim():
    BOSScore = (rnd.gauss(BOS_score,BOS_sdpts)+ rnd.gauss(BOS_oppmeanpoints,BOS_oppstdpts))/2
    TORScore = (rnd.gauss(TOR_score,TOR_sdpts)+ rnd.gauss(TOR_oppmeanpoints,TOR_oppstdpts))/2
    if int(round(BOSScore)) > int(round(TORScore)):
        return 1
    elif int(round(BOSScore)) < int(round(TORScore)):
        return -1
    else: return 0
```

```python
def gamesSim(ns):
    gamesout = []
    team1win = 0
    team2win = 0
    for i in range(ns):
        gm = gameSim()
        gamesout.append(gm)
        if gm == 1:
            team1win +=1 
        elif gm == -1:
            team2win +=1
    print('BOS Win ', round((team1win/(team1win+team2win))*100,2),'%')
    print('TOR Win ', round((team2win/(team1win+team2win))*100,2),'%')
    return gamesout
```
