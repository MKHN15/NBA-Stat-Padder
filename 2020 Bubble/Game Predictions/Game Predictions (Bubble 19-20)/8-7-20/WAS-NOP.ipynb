```python
WAS_meanpoints = WAS1df.TeamPTS.mean()
WAS_meanfgp = WAS1df.TeamFGP.mean()
WAS_sdpts = WAS1df.TeamPTS.std()
WAS_sdfgp = WAS1df.TeamFGP.std()
WAS_oppmeanpoints = WAS1df.OpponentPTS.mean()
WAS_oppfgp = WAS1df.OpponentFGP.mean()
WAS_oppstdpts = WAS1df.OpponentPTS.std()
WAS_oppstdfgp = WAS1df.OpponentFGP.std()

print("WAS Avg. Pts ", WAS_meanpoints)
print("WAS Avg. FG% ", WAS_meanfgp)
print("WAS Pts Deviation ", WAS_sdpts)
print("WAS FG% Deviation ", WAS_sdfgp)
print("WAS Opp. Avg. Pts ", WAS_oppmeanpoints)
print("WAS Opp. Avg. FG% ", WAS_oppfgp)
print("WAS Opp. Pts Deviation ", WAS_oppstdpts)
print("WAS Opp. FG% Deviation ", WAS_oppstdfgp)
```

```python
NOP_meanpoints = NOP1df.TeamPTS.mean()
NOP_meanfgp = NOP1df.TeamFGP.mean()
NOP_sdpts = NOP1df.TeamPTS.std()
NOP_sdfgp = NOP1df.TeamFGP.std()
NOP_oppmeanpoints = NOP1df.OpponentPTS.mean()
NOP_oppfgp = NOP1df.OpponentFGP.mean()
NOP_oppstdpts = NOP1df.OpponentPTS.std()
NOP_oppstdfgp = NOP1df.OpponentFGP.std()

print("NOP Avg. Pts ", NOP_meanpoints)
print("NOP Avg. FG% ", NOP_meanfgp)
print("NOP Pts Deviation ", NOP_sdpts)
print("NOP FG% Deviation ", NOP_sdfgp)
print("NOP Opp. Avg. Pts ", NOP_oppmeanpoints)
print("NOP Opp. Avg. FG% ", NOP_oppfgp)
print("NOP Opp. Pts Deviation ", NOP_oppstdpts)
print("NOP Opp. FG% Deviation ", NOP_oppstdfgp)
```

```python
def gameSim():
    WASScore = (rnd.gauss(WAS_score,WAS_sdpts)+ rnd.gauss(WAS_oppmeanpoints,WAS_oppstdpts))/2
    NOPScore = (rnd.gauss(NOP_score,NOP_sdpts)+ rnd.gauss(NOP_oppmeanpoints,NOP_oppstdpts))/2
    if int(round(WASScore)) > int(round(NOPScore)):
        return 1
    elif int(round(WASScore)) < int(round(NOPScore)):
        return -1
    else: return 0
```

```python
def gamesSim(ns):
    gamesout = []
    team1win = 0
    team2win = 0
    for i in range(ns):
        gm = gameSim()
        gamesout.append(gm)
        if gm == 1:
            team1win +=1 
        elif gm == -1:
            team2win +=1
    print('WAS Win ', round((team1win/(team1win+team2win))*100,2),'%')
    print('NOP Win ', round((team2win/(team1win+team2win))*100,2),'%')
    return gamesout
```
