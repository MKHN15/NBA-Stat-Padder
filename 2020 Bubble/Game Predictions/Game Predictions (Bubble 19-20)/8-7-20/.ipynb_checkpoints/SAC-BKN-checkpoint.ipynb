```python
SAC_meanpoints = SAC1df.TeamPTS.mean()
SAC_meanfgp = SAC1df.TeamFGP.mean()
SAC_sdpts = SAC1df.TeamPTS.std()
SAC_sdfgp = SAC1df.TeamFGP.std()
SAC_oppmeanpoints = SAC1df.OpponentPTS.mean()
SAC_oppfgp = SAC1df.OpponentFGP.mean()
SAC_oppstdpts = SAC1df.OpponentPTS.std()
SAC_oppstdfgp = SAC1df.OpponentFGP.std()

print("SAC Avg. Pts ", SAC_meanpoints)
print("SAC Avg. FG% ", SAC_meanfgp)
print("SAC Pts Deviation ", SAC_sdpts)
print("SAC FG% Deviation ", SAC_sdfgp)
print("SAC Opp. Avg. Pts ", SAC_oppmeanpoints)
print("SAC Opp. Avg. FG% ", SAC_oppfgp)
print("SAC Opp. Pts Deviation ", SAC_oppstdpts)
print("SAC Opp. FG% Deviation ", SAC_oppstdfgp)
```

```python
BKN_meanpoints = BKN1df.TeamPTS.mean()
BKN_meanfgp = BKN1df.TeamFGP.mean()
BKN_sdpts = BKN1df.TeamPTS.std()
BKN_sdfgp = BKN1df.TeamFGP.std()
BKN_oppmeanpoints = BKN1df.OpponentPTS.mean()
BKN_oppfgp = BKN1df.OpponentFGP.mean()
BKN_oppstdpts = BKN1df.OpponentPTS.std()
BKN_oppstdfgp = BKN1df.OpponentFGP.std()

print("BKN Avg. Pts ", BKN_meanpoints)
print("BKN Avg. FG% ", BKN_meanfgp)
print("BKN Pts Deviation ", BKN_sdpts)
print("BKN FG% Deviation ", BKN_sdfgp)
print("BKN Opp. Avg. Pts ", BKN_oppmeanpoints)
print("BKN Opp. Avg. FG% ", BKN_oppfgp)
print("BKN Opp. Pts Deviation ", BKN_oppstdpts)
print("BKN Opp. FG% Deviation ", BKN_oppstdfgp)
```

```python
def gameSim():
    SACScore = (rnd.gauss(SAC_score,SAC_sdpts)+ rnd.gauss(SAC_oppmeanpoints,SAC_oppstdpts))/2
    BKNScore = (rnd.gauss(BKN_score,BKN_sdpts)+ rnd.gauss(BKN_oppmeanpoints,BKN_oppstdpts))/2
    if int(round(SACScore)) > int(round(BKNScore)):
        return 1
    elif int(round(SACScore)) < int(round(BKNScore)):
        return -1
    else: return 0
```

```python
def gamesSim(ns):
    gamesout = []
    team1win = 0
    team2win = 0
    for i in range(ns):
        gm = gameSim()
        gamesout.append(gm)
        if gm == 1:
            team1win +=1 
        elif gm == -1:
            team2win +=1
    print('SAC Win ', round((team1win/(team1win+team2win))*100,2),'%')
    print('BKN Win ', round((team2win/(team1win+team2win))*100,2),'%')
    return gamesout
```

```python
gamesSim(10000)
```
