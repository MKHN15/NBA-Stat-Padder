```python
IND_scoring = IND_meanpts*(IND_fgp*IND_tpp*IND_ftp+IND_offrtg)
IND_misc = IND_scoring+IND_dreb+IND_ast+IND_stl+IND_blk+IND_pfd+IND_defrtg
IND_top = IND_misc*IND_winpct*IND_clutchwinpct
IND_score = IND_top/(IND_tov*10000)
IND_score

print(IND_score)
```

```python
MIA_scoring = MIA_meanpts*(MIA_fgp*MIA_tpp*MIA_ftp+MIA_offrtg)
MIA_misc = MIA_scoring+MIA_dreb+MIA_ast+MIA_stl+MIA_blk+MIA_pfd+MIA_defrtg
MIA_top = MIA_misc*MIA_winpct*MIA_clutchwinpct
MIA_score = MIA_top/(MIA_tov*10000)
MIA_score

print(MIA_score)
```

```python
IND_meanpoints = IND1df.TeamPTS.mean()
IND_meanfgp = IND1df.TeamFGP.mean()
IND_sdpts = IND1df.TeamPTS.std()
IND_sdfgp = IND1df.TeamFGP.std()
IND_oppmeanpoints = IND1df.OpponentPTS.mean()
IND_oppfgp = IND1df.OpponentFGP.mean()
IND_oppstdpts = IND1df.OpponentPTS.std()
IND_oppstdfgp = IND1df.OpponentFGP.std()

print("IND Avg. Pts ", IND_meanpoints)
print("IND Avg. FG% ", IND_meanfgp)
print("IND Pts Deviation ", IND_sdpts)
print("IND FG% Deviation ", IND_sdfgp)
print("IND Opp. Avg. Pts ", IND_oppmeanpoints)
print("IND Opp. Avg. FG% ", IND_oppfgp)
print("IND Opp. Pts Deviation ", IND_oppstdpts)
print("IND Opp. FG% Deviation ", IND_oppstdfgp)
```

```python
MIA_meanpoints = MIA1df.TeamPTS.mean()
MIA_meanfgp = MIA1df.TeamFGP.mean()
MIA_sdpts = MIA1df.TeamPTS.std()
MIA_sdfgp = MIA1df.TeamFGP.std()
MIA_oppmeanpoints = MIA1df.OpponentPTS.mean()
MIA_oppfgp = MIA1df.OpponentFGP.mean()
MIA_oppstdpts = MIA1df.OpponentPTS.std()
MIA_oppstdfgp = MIA1df.OpponentFGP.std()

print("MIA Avg. Pts ", MIA_meanpoints)
print("MIA Avg. FG% ", MIA_meanfgp)
print("MIA Pts Deviation ", MIA_sdpts)
print("MIA FG% Deviation ", MIA_sdfgp)
print("MIA Opp. Avg. Pts ", MIA_oppmeanpoints)
print("MIA Opp. Avg. FG% ", MIA_oppfgp)
print("MIA Opp. Pts Deviation ", MIA_oppstdpts)
print("MIA Opp. FG% Deviation ", MIA_oppstdfgp)
```

```python
def gameSim():
    INDScore = (rnd.gauss(IND_score,IND_sdpts)+ rnd.gauss(IND_oppmeanpoints,IND_oppstdpts))/2
    MIAScore = (rnd.gauss(MIA_score,MIA_sdpts)+ rnd.gauss(MIA_oppmeanpoints,MIA_oppstdpts))/2
    if int(round(INDScore)) > int(round(MIAScore)):
        return 1
    elif int(round(INDScore)) < int(round(MIAScore)):
        return -1
    else: return 0
```

```python
def gamesSim(ns):
    gamesout = []
    team1win = 0
    team2win = 0
    for i in range(ns):
        gm = gameSim()
        gamesout.append(gm)
        if gm == 1:
            team1win +=1 
        elif gm == -1:
            team2win +=1
    print('IND Win ', round((team1win/(team1win+team2win))*100,2),'%')
    print('MIA Win ', round((team2win/(team1win+team2win))*100,2),'%')
    return gamesout
```

```python
gamesSim(10000)
```
