```python
OKC_meanpoints = OKC1df.TeamPTS.mean()
OKC_meanfgp = OKC1df.TeamFGP.mean()
OKC_sdpts = OKC1df.TeamPTS.std()
OKC_sdfgp = OKC1df.TeamFGP.std()
OKC_oppmeanpoints = OKC1df.OpponentPTS.mean()
OKC_oppfgp = OKC1df.OpponentFGP.mean()
OKC_oppstdpts = OKC1df.OpponentPTS.std()
OKC_oppstdfgp = OKC1df.OpponentFGP.std()

print("OKC Avg. Pts ", OKC_meanpoints)
print("OKC Avg. FG% ", OKC_meanfgp)
print("OKC Pts Deviation ", OKC_sdpts)
print("OKC FG% Deviation ", OKC_sdfgp)
print("OKC Opp. Avg. Pts ", OKC_oppmeanpoints)
print("OKC Opp. Avg. FG% ", OKC_oppfgp)
print("OKC Opp. Pts Deviation ", OKC_oppstdpts)
print("OKC Opp. FG% Deviation ", OKC_oppstdfgp)
```

```python
PHX_meanpoints = PHX1df.TeamPTS.mean()
PHX_meanfgp = PHX1df.TeamFGP.mean()
PHX_sdpts = PHX1df.TeamPTS.std()
PHX_sdfgp = PHX1df.TeamFGP.std()
PHX_oppmeanpoints = PHX1df.OpponentPTS.mean()
PHX_oppfgp = PHX1df.OpponentFGP.mean()
PHX_oppstdpts = PHX1df.OpponentPTS.std()
PHX_oppstdfgp = PHX1df.OpponentFGP.std()

print("PHX Avg. Pts ", PHX_meanpoints)
print("PHX Avg. FG% ", PHX_meanfgp)
print("PHX Pts Deviation ", PHX_sdpts)
print("PHX FG% Deviation ", PHX_sdfgp)
print("PHX Opp. Avg. Pts ", PHX_oppmeanpoints)
print("PHX Opp. Avg. FG% ", PHX_oppfgp)
print("PHX Opp. Pts Deviation ", PHX_oppstdpts)
print("PHX Opp. FG% Deviation ", PHX_oppstdfgp)
```

```python
def gameSim():
    OKCScore = (rnd.gauss(OKC_score,OKC_sdpts)+ rnd.gauss(OKC_oppmeanpoints,OKC_oppstdpts))/2
    PHXScore = (rnd.gauss(PHX_score,PHX_sdpts)+ rnd.gauss(PHX_oppmeanpoints,PHX_oppstdpts))/2
    if int(round(OKCScore)) > int(round(PHXScore)):
        return 1
    elif int(round(OKCScore)) < int(round(PHXScore)):
        return -1
    else: return 0
```

```python
def gamesSim(ns):
    gamesout = []
    team1win = 0
    team2win = 0
    for i in range(ns):
        gm = gameSim()
        gamesout.append(gm)
        if gm == 1:
            team1win +=1 
        elif gm == -1:
            team2win +=1
    print('OKC Win ', round((team1win/(team1win+team2win))*100,2),'%')
    print('PHX Win ', round((team2win/(team1win+team2win))*100,2),'%')
    return gamesout
```

```python
gamesSim(10000)
```
