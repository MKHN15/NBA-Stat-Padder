```python
MIL_winpct = MILdf.WINP.mean()
MIL_meanpts = MILdf.PTS.mean()
MIL_fgp = MILdf.FGP.mean()
MIL_tpp = MILdf.TPP.mean()
MIL_ftp = MILdf.FTP.mean()
MIL_dreb = MILdf.DREB.mean()
MIL_ast = MILdf.AST.mean()
MIL_tov = MILdf.TOV.mean()
MIL_stl = MILdf.STL.mean()
MIL_blk = MILdf.BLK.mean()
MIL_pfd = MILdf.PFD.mean()
MIL_offrtg = TORdf.ORTG.mean()
MIL_defrtg = TORdf.DRTG.mean()
MIL_clutchwinpct = TORdf.CWINP.mean()

print("MIL Win% ", MIL_winpct)
print("MIL Avg. Points ", MIL_meanpts)
print("MIL FG% ", MIL_fgp)
print("MIL 3P% ", MIL_tpp)
print("MIL FT% ", MIL_ftp)
print("MIL Def. Reb. ", MIL_dreb)
print("MIL Avg. Assists ", MIL_ast)
print("MIL Avg. Turnovers ", MIL_tov)
print("MIL Avg. Steals ", MIL_stl)
print("MIL Avg. Blocks ", MIL_blk)
print("MIL Avg. Personal Fouls Drawn ", MIL_pfd)
print("MIL Off. Rtg ", MIL_offrtg)
print("MIL Def. Rtg ", MIL_defrtg)
print("MIL Clutch Win% ", MIL_clutchwinpct)
```

```python
TOR_scoring = TOR_meanpts*(TOR_fgp*TOR_tpp*TOR_ftp+TOR_offrtg)
TOR_misc = TOR_scoring+TOR_dreb+TOR_ast+TOR_stl+TOR_blk+TOR_pfd+TOR_defrtg
TOR_top = TOR_misc*TOR_winpct*TOR_clutchwinpct
TOR_score = TOR_top/(TOR_tov*10000)
TOR_score

print(TOR_score)
```

```python
MIL_scoring = MIL_meanpts*(MIL_fgp*MIL_tpp*MIL_ftp+MIL_offrtg)
MIL_misc = MIL_scoring+MIL_dreb+MIL_ast+MIL_stl+MIL_blk+MIL_pfd+MIL_defrtg
MIL_top = MIL_misc*MIL_winpct*MIL_clutchwinpct
MIL_score = MIL_top/(MIL_tov*10000)
MIL_score

print(MIL_score)
```

```python
TOR_meanpoints = TOR1df.TeamPTS.mean()
TOR_meanfgp = TOR1df.TeamFGP.mean()
TOR_sdpts = TOR1df.TeamPTS.std()
TOR_sdfgp = TOR1df.TeamFGP.std()
TOR_oppmeanpoints = TOR1df.OpponentPTS.mean()
TOR_oppfgp = TOR1df.OpponentFGP.mean()
TOR_oppstdpts = TOR1df.OpponentPTS.std()
TOR_oppstdfgp = TOR1df.OpponentFGP.std()

print("TOR Avg. Pts ", TOR_meanpoints)
print("TOR Avg. FG% ", TOR_meanfgp)
print("TOR Pts Deviation ", TOR_sdpts)
print("TOR FG% Deviation ", TOR_sdfgp)
print("TOR Opp. Avg. Pts ", TOR_oppmeanpoints)
print("TOR Opp. Avg. FG% ", TOR_oppfgp)
print("TOR Opp. Pts Deviation ", TOR_oppstdpts)
print("TOR Opp. FG% Deviation ", TOR_oppstdfgp)
```

```python
MIL_meanpoints = MIL1df.TeamPTS.mean()
MIL_meanfgp = MIL1df.TeamFGP.mean()
MIL_sdpts = MIL1df.TeamPTS.std()
MIL_sdfgp = MIL1df.TeamFGP.std()
MIL_oppmeanpoints = MIL1df.OpponentPTS.mean()
MIL_oppfgp = MIL1df.OpponentFGP.mean()
MIL_oppstdpts = MIL1df.OpponentPTS.std()
MIL_oppstdfgp = MIL1df.OpponentFGP.std()

print("MIL Avg. Pts ", MIL_meanpoints)
print("MIL Avg. FG% ", MIL_meanfgp)
print("MIL Pts Deviation ", MIL_sdpts)
print("MIL FG% Deviation ", MIL_sdfgp)
print("MIL Opp. Avg. Pts ", MIL_oppmeanpoints)
print("MIL Opp. Avg. FG% ", MIL_oppfgp)
print("MIL Opp. Pts Deviation ", MIL_oppstdpts)
print("MIL Opp. FG% Deviation ", MIL_oppstdfgp)
```

```python
def gameSim():
    TORScore = (rnd.gauss(TOR_score,TOR_sdpts)+ rnd.gauss(TOR_oppmeanpoints,TOR_oppstdpts))/2
    MILScore = (rnd.gauss(MIL_score,MIL_sdpts)+ rnd.gauss(MIL_oppmeanpoints,MIL_oppstdpts))/2
    if int(round(TORScore)) > int(round(MILScore)):
        return 1
    elif int(round(TORScore)) < int(round(MILScore)):
        return -1
    else: return 0
```

```python
def gamesSim(ns):
    gamesout = []
    team1win = 0
    team2win = 0
    for i in range(ns):
        gm = gameSim()
        gamesout.append(gm)
        if gm == 1:
            team1win +=1 
        elif gm == -1:
            team2win +=1
    print('TOR Win ', round((team1win/(team1win+team2win))*100,2),'%')
    print('MIL Win ', round((team2win/(team1win+team2win))*100,2),'%')
    return gamesout
```

```python
gamesSim(10000)
```
