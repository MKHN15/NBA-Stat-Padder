```python
DAL_meanpoints = DAL1df.TeamPTS.mean()
DAL_meanfgp = DAL1df.TeamFGP.mean()
DAL_sdpts = DAL1df.TeamPTS.std()
DAL_sdfgp = DAL1df.TeamFGP.std()
DAL_oppmeanpoints = DAL1df.OpponentPTS.mean()
DAL_oppfgp = DAL1df.OpponentFGP.mean()
DAL_oppstdpts = DAL1df.OpponentPTS.std()
DAL_oppstdfgp = DAL1df.OpponentFGP.std()

print("DAL Avg. Pts ", DAL_meanpoints)
print("DAL Avg. FG% ", DAL_meanfgp)
print("DAL Pts Deviation ", DAL_sdpts)
print("DAL FG% Deviation ", DAL_sdfgp)
print("DAL Opp. Avg. Pts ", DAL_oppmeanpoints)
print("DAL Opp. Avg. FG% ", DAL_oppfgp)
print("DAL Opp. Pts Deviation ", DAL_oppstdpts)
print("DAL Opp. FG% Deviation ", DAL_oppstdfgp)
```

```python
UTA_meanpoints = UTA1df.TeamPTS.mean()
UTA_meanfgp = UTA1df.TeamFGP.mean()
UTA_sdpts = UTA1df.TeamPTS.std()
UTA_sdfgp = UTA1df.TeamFGP.std()
UTA_oppmeanpoints = UTA1df.OpponentPTS.mean()
UTA_oppfgp = UTA1df.OpponentFGP.mean()
UTA_oppstdpts = UTA1df.OpponentPTS.std()
UTA_oppstdfgp = UTA1df.OpponentFGP.std()

print("UTA Avg. Pts ", UTA_meanpoints)
print("UTA Avg. FG% ", UTA_meanfgp)
print("UTA Pts Deviation ", UTA_sdpts)
print("UTA FG% Deviation ", UTA_sdfgp)
print("UTA Opp. Avg. Pts ", UTA_oppmeanpoints)
print("UTA Opp. Avg. FG% ", UTA_oppfgp)
print("UTA Opp. Pts Deviation ", UTA_oppstdpts)
print("UTA Opp. FG% Deviation ", UTA_oppstdfgp)
```

```python
def gameSim():
    DALScore = (rnd.gauss(DAL_score,DAL_sdpts)+ rnd.gauss(DAL_oppmeanpoints,DAL_oppstdpts))/2
    UTAScore = (rnd.gauss(UTA_score,UTA_sdpts)+ rnd.gauss(UTA_oppmeanpoints,UTA_oppstdpts))/2
    if int(round(DALScore)) > int(round(UTAScore)):
        return 1
    elif int(round(DALScore)) < int(round(UTAScore)):
        return -1
    else: return 0
```

```python
def gamesSim(ns):
    gamesout = []
    team1win = 0
    team2win = 0
    for i in range(ns):
        gm = gameSim()
        gamesout.append(gm)
        if gm == 1:
            team1win +=1 
        elif gm == -1:
            team2win +=1
    print('DAL Win ', round((team1win/(team1win+team2win))*100,2),'%')
    print('UTA Win ', round((team2win/(team1win+team2win))*100,2),'%')
    return gamesout
```

```python
gamesSim(10000)
```
