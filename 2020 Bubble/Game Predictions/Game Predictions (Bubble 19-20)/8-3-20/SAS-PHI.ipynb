```python
SAS_meanpoints = SAS1df.TeamPTS.mean()
SAS_meanfgp = SAS1df.TeamFGP.mean()
SAS_sdpts = SAS1df.TeamPTS.std()
SAS_sdfgp = SAS1df.TeamFGP.std()
SAS_oppmeanpoints = SAS1df.OpponentPTS.mean()
SAS_oppfgp = SAS1df.OpponentFGP.mean()
SAS_oppstdpts = SAS1df.OpponentPTS.std()
SAS_oppstdfgp = SAS1df.OpponentFGP.std()

print("SAS Avg. Pts ", SAS_meanpoints)
print("SAS Avg. FG% ", SAS_meanfgp)
print("SAS Pts Deviation ", SAS_sdpts)
print("SAS FG% Deviation ", SAS_sdfgp)
print("SAS Opp. Avg. Pts ", SAS_oppmeanpoints)
print("SAS Opp. Avg. FG% ", SAS_oppfgp)
print("SAS Opp. Pts Deviation ", SAS_oppstdpts)
print("SAS Opp. FG% Deviation ", SAS_oppstdfgp)
```

```python
PHI_meanpoints = PHI1df.TeamPTS.mean()
PHI_meanfgp = PHI1df.TeamFGP.mean()
PHI_sdpts = PHI1df.TeamPTS.std()
PHI_sdfgp = PHI1df.TeamFGP.std()
PHI_oppmeanpoints = PHI1df.OpponentPTS.mean()
PHI_oppfgp = PHI1df.OpponentFGP.mean()
PHI_oppstdpts = PHI1df.OpponentPTS.std()
PHI_oppstdfgp = PHI1df.OpponentFGP.std()

print("PHI Avg. Pts ", PHI_meanpoints)
print("PHI Avg. FG% ", PHI_meanfgp)
print("PHI Pts Deviation ", PHI_sdpts)
print("PHI FG% Deviation ", PHI_sdfgp)
print("PHI Opp. Avg. Pts ", PHI_oppmeanpoints)
print("PHI Opp. Avg. FG% ", PHI_oppfgp)
print("PHI Opp. Pts Deviation ", PHI_oppstdpts)
print("PHI Opp. FG% Deviation ", PHI_oppstdfgp)
```

```python
def gameSim():
    SASScore = (rnd.gauss(SAS_score,SAS_sdpts)+ rnd.gauss(SAS_oppmeanpoints,SAS_oppstdpts))/2
    PHIScore = (rnd.gauss(PHI_score,PHI_sdpts)+ rnd.gauss(PHI_oppmeanpoints,PHI_oppstdpts))/2
    if int(round(SASScore)) > int(round(PHIScore)):
        return 1
    elif int(round(SASScore)) < int(round(PHIScore)):
        return -1
    else: return 0
```

```python
def gamesSim(ns):
    gamesout = []
    team1win = 0
    team2win = 0
    for i in range(ns):
        gm = gameSim()
        gamesout.append(gm)
        if gm == 1:
            team1win +=1 
        elif gm == -1:
            team2win +=1
    print('SAS Win ', round((team1win/(team1win+team2win))*100,2),'%')
    print('PHI Win ', round((team2win/(team1win+team2win))*100,2),'%')
    return gamesout
```
