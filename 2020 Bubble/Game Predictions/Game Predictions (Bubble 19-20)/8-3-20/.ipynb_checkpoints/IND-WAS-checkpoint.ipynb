```python
IND_meanpoints = IND1df.TeamPTS.mean()
IND_meanfgp = IND1df.TeamFGP.mean()
IND_sdpts = IND1df.TeamPTS.std()
IND_sdfgp = IND1df.TeamFGP.std()
IND_oppmeanpoints = IND1df.OpponentPTS.mean()
IND_oppfgp = IND1df.OpponentFGP.mean()
IND_oppstdpts = IND1df.OpponentPTS.std()
IND_oppstdfgp = IND1df.OpponentFGP.std()

print("IND Avg. Pts ", IND_meanpoints)
print("IND Avg. FG% ", IND_meanfgp)
print("IND Pts Deviation ", IND_sdpts)
print("IND FG% Deviation ", IND_sdfgp)
print("IND Opp. Avg. Pts ", IND_oppmeanpoints)
print("IND Opp. Avg. FG% ", IND_oppfgp)
print("IND Opp. Pts Deviation ", IND_oppstdpts)
print("IND Opp. FG% Deviation ", IND_oppstdfgp)
```

```python
WAS_meanpoints = WAS1df.TeamPTS.mean()
WAS_meanfgp = WAS1df.TeamFGP.mean()
WAS_sdpts = WAS1df.TeamPTS.std()
WAS_sdfgp = WAS1df.TeamFGP.std()
WAS_oppmeanpoints = WAS1df.OpponentPTS.mean()
WAS_oppfgp = WAS1df.OpponentFGP.mean()
WAS_oppstdpts = WAS1df.OpponentPTS.std()
WAS_oppstdfgp = WAS1df.OpponentFGP.std()

print("WAS Avg. Pts ", WAS_meanpoints)
print("WAS Avg. FG% ", WAS_meanfgp)
print("WAS Pts Deviation ", WAS_sdpts)
print("WAS FG% Deviation ", WAS_sdfgp)
print("WAS Opp. Avg. Pts ", WAS_oppmeanpoints)
print("WAS Opp. Avg. FG% ", WAS_oppfgp)
print("WAS Opp. Pts Deviation ", WAS_oppstdpts)
print("WAS Opp. FG% Deviation ", WAS_oppstdfgp)
```

```python
def gameSim():
    INDScore = (rnd.gauss(IND_score,IND_sdpts)+ rnd.gauss(IND_oppmeanpoints,IND_oppstdpts))/2
    WASScore = (rnd.gauss(WAS_score,WAS_sdpts)+ rnd.gauss(WAS_oppmeanpoints,WAS_oppstdpts))/2
    if int(round(INDScore)) > int(round(WASScore)):
        return 1
    elif int(round(INDScore)) < int(round(WASScore)):
        return -1
    else: return 0
```

```python
def gamesSim(ns):
    gamesout = []
    team1win = 0
    team2win = 0
    for i in range(ns):
        gm = gameSim()
        gamesout.append(gm)
        if gm == 1:
            team1win +=1 
        elif gm == -1:
            team2win +=1
    print('IND Win ', round((team1win/(team1win+team2win))*100,2),'%')
    print('WAS Win ', round((team2win/(team1win+team2win))*100,2),'%')
    return gamesout
```

```python
gamesSim(10000)
```
