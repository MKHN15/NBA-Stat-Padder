```python
import pandas as pd 
import random as rnd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
MEMdf = gdf[gdf.TEAM == 'Memphis Grizzlies']
NOPdf = gdf[gdf.TEAM == 'New Orleans Pelicans']
```

```python
MEM_winpct = MEMdf.WINP.mean()
MEM_meanpts = MEMdf.PTS.mean()
MEM_fgp = MEMdf.FGP.mean()
MEM_tpp = MEMdf.TPP.mean()
MEM_ftp = MEMdf.FTP.mean()
MEM_dreb = MEMdf.DREB.mean()
MEM_ast = MEMdf.AST.mean()
MEM_tov = MEMdf.TOV.mean()
MEM_stl = MEMdf.STL.mean()
MEM_blk = MEMdf.BLK.mean()
MEM_pfd = MEMdf.PFD.mean()
MEM_offrtg = MEMdf.ORTG.mean()
MEM_defrtg = MEMdf.DRTG.mean()
MEM_clutchwinpct = MEMdf.CWINP.mean()

print("MEM Win% ", MEM_winpct)
print("MEM Avg. Points ", MEM_meanpts)
print("MEM FG% ", MEM_fgp)
print("MEM 3P% ", MEM_tpp)
print("MEM FT% ", MEM_ftp)
print("MEM Def. Reb. ", MEM_dreb)
print("MEM Avg. Assists ", MEM_ast)
print("MEM Avg. Turnovers ", MEM_tov)
print("MEM Avg. Steals ", MEM_stl)
print("MEM Avg. Blocks ", MEM_blk)
print("MEM Avg. Personal Fouls Drawn ", MEM_pfd)
print("MEM Off. Rtg ", MEM_offrtg)
print("MEM Def. Rtg ", MEM_defrtg)
print("MEM Clutch Win% ", MEM_clutchwinpct)
```

```python
NOP_winpct = NOPdf.WINP.mean()
NOP_meanpts = NOPdf.PTS.mean()
NOP_fgp = NOPdf.FGP.mean()
NOP_tpp = NOPdf.TPP.mean()
NOP_ftp = NOPdf.FTP.mean()
NOP_dreb = NOPdf.DREB.mean()
NOP_ast = NOPdf.AST.mean()
NOP_tov = NOPdf.TOV.mean()
NOP_stl = NOPdf.STL.mean()
NOP_blk = NOPdf.BLK.mean()
NOP_pfd = NOPdf.PFD.mean()
NOP_offrtg = MEMdf.ORTG.mean()
NOP_defrtg = MEMdf.DRTG.mean()
NOP_clutchwinpct = MEMdf.CWINP.mean()

print("NOP Win% ", NOP_winpct)
print("NOP Avg. Points ", NOP_meanpts)
print("NOP FG% ", NOP_fgp)
print("NOP 3P% ", NOP_tpp)
print("NOP FT% ", NOP_ftp)
print("NOP Def. Reb. ", NOP_dreb)
print("NOP Avg. Assists ", NOP_ast)
print("NOP Avg. Turnovers ", NOP_tov)
print("NOP Avg. Steals ", NOP_stl)
print("NOP Avg. Blocks ", NOP_blk)
print("NOP Avg. Personal Fouls Drawn ", NOP_pfd)
print("NOP Off. Rtg ", NOP_offrtg)
print("NOP Def. Rtg ", NOP_defrtg)
print("NOP Clutch Win% ", NOP_clutchwinpct)
```

```python
MEM_scoring = MEM_meanpts*(MEM_fgp*MEM_tpp*MEM_ftp+MEM_offrtg)
MEM_misc = MEM_scoring+MEM_dreb+MEM_ast+MEM_stl+MEM_blk+MEM_pfd+MEM_defrtg
MEM_top = MEM_misc*MEM_winpct*MEM_clutchwinpct
MEM_score = MEM_top/(MEM_tov*10000)
MEM_score

print(MEM_score)
```

```python
NOP_scoring = NOP_meanpts*(NOP_fgp*NOP_tpp*NOP_ftp+NOP_offrtg)
NOP_misc = NOP_scoring+NOP_dreb+NOP_ast+NOP_stl+NOP_blk+NOP_pfd+NOP_defrtg
NOP_top = NOP_misc*NOP_winpct*NOP_clutchwinpct
NOP_score = NOP_top/(NOP_tov*10000)
NOP_score

print(NOP_score)
```

```python
MEM_meanpoints = MEM1df.TeamPTS.mean()
MEM_meanfgp = MEM1df.TeamFGP.mean()
MEM_sdpts = MEM1df.TeamPTS.std()
MEM_sdfgp = MEM1df.TeamFGP.std()
MEM_oppmeanpoints = MEM1df.OpponentPTS.mean()
MEM_oppfgp = MEM1df.OpponentFGP.mean()
MEM_oppstdpts = MEM1df.OpponentPTS.std()
MEM_oppstdfgp = MEM1df.OpponentFGP.std()

print("MEM Avg. Pts ", MEM_meanpoints)
print("MEM Avg. FG% ", MEM_meanfgp)
print("MEM Pts Deviation ", MEM_sdpts)
print("MEM FG% Deviation ", MEM_sdfgp)
print("MEM Opp. Avg. Pts ", MEM_oppmeanpoints)
print("MEM Opp. Avg. FG% ", MEM_oppfgp)
print("MEM Opp. Pts Deviation ", MEM_oppstdpts)
print("MEM Opp. FG% Deviation ", MEM_oppstdfgp)
```

```python
NOP_meanpoints = NOP1df.TeamPTS.mean()
NOP_meanfgp = NOP1df.TeamFGP.mean()
NOP_sdpts = NOP1df.TeamPTS.std()
NOP_sdfgp = NOP1df.TeamFGP.std()
NOP_oppmeanpoints = NOP1df.OpponentPTS.mean()
NOP_oppfgp = NOP1df.OpponentFGP.mean()
NOP_oppstdpts = NOP1df.OpponentPTS.std()
NOP_oppstdfgp = NOP1df.OpponentFGP.std()

print("NOP Avg. Pts ", NOP_meanpoints)
print("NOP Avg. FG% ", NOP_meanfgp)
print("NOP Pts Deviation ", NOP_sdpts)
print("NOP FG% Deviation ", NOP_sdfgp)
print("NOP Opp. Avg. Pts ", NOP_oppmeanpoints)
print("NOP Opp. Avg. FG% ", NOP_oppfgp)
print("NOP Opp. Pts Deviation ", NOP_oppstdpts)
print("NOP Opp. FG% Deviation ", NOP_oppstdfgp)
```

```python
def gameSim():
    MEMScore = (rnd.gauss(MEM_score,MEM_sdpts)+ rnd.gauss(MEM_oppmeanpoints,MEM_oppstdpts))/2
    NOPScore = (rnd.gauss(NOP_score,NOP_sdpts)+ rnd.gauss(NOP_oppmeanpoints,NOP_oppstdpts))/2
    if int(round(MEMScore)) > int(round(NOPScore)):
        return 1
    elif int(round(MEMScore)) < int(round(NOPScore)):
        return -1
    else: return 0
```

```python
def gamesSim(ns):
    gamesout = []
    team1win = 0
    team2win = 0
    for i in range(ns):
        gm = gameSim()
        gamesout.append(gm)
        if gm == 1:
            team1win +=1 
        elif gm == -1:
            team2win +=1
    print('MEM Win ', round((team1win/(team1win+team2win))*100,2),'%')
    print('NOP Win ', round((team2win/(team1win+team2win))*100,2),'%')
    return gamesout
```

```python
gamesSim(10000)
```
