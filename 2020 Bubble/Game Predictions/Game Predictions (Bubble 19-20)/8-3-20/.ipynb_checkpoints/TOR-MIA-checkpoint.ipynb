```python
TOR_winpct = TORdf.WINP.mean()
TOR_meanpts = TORdf.PTS.mean()
TOR_fgp = TORdf.FGP.mean()
TOR_tpp = TORdf.TPP.mean()
TOR_ftp = TORdf.FTP.mean()
TOR_dreb = TORdf.DREB.mean()
TOR_ast = TORdf.AST.mean()
TOR_tov = TORdf.TOV.mean()
TOR_stl = TORdf.STL.mean()
TOR_blk = TORdf.BLK.mean()
TOR_pfd = TORdf.PFD.mean()
TOR_offrtg = TORdf.ORTG.mean()
TOR_defrtg = TORdf.DRTG.mean()
TOR_clutchwinpct = TORdf.CWINP.mean()

print("TOR Win% ", TOR_winpct)
print("TOR Avg. Points ", TOR_meanpts)
print("TOR FG% ", TOR_fgp)
print("TOR 3P% ", TOR_tpp)
print("TOR FT% ", TOR_ftp)
print("TOR Def. Reb. ", TOR_dreb)
print("TOR Avg. Assists ", TOR_ast)
print("TOR Avg. Turnovers ", TOR_tov)
print("TOR Avg. Steals ", TOR_stl)
print("TOR Avg. Blocks ", TOR_blk)
print("TOR Avg. Personal Fouls Drawn ", TOR_pfd)
print("TOR Off. Rtg ", TOR_offrtg)
print("TOR Def. Rtg ", TOR_defrtg)
print("TOR Clutch Win% ", TOR_clutchwinpct)
```

```python
MIA_winpct = MIAdf.WINP.mean()
MIA_meanpts = MIAdf.PTS.mean()
MIA_fgp = MIAdf.FGP.mean()
MIA_tpp = MIAdf.TPP.mean()
MIA_ftp = MIAdf.FTP.mean()
MIA_dreb = MIAdf.DREB.mean()
MIA_ast = MIAdf.AST.mean()
MIA_tov = MIAdf.TOV.mean()
MIA_stl = MIAdf.STL.mean()
MIA_blk = MIAdf.BLK.mean()
MIA_pfd = MIAdf.PFD.mean()
MIA_offrtg = TORdf.ORTG.mean()
MIA_defrtg = TORdf.DRTG.mean()
MIA_clutchwinpct = TORdf.CWINP.mean()

print("MIA Win% ", MIA_winpct)
print("MIA Avg. Points ", MIA_meanpts)
print("MIA FG% ", MIA_fgp)
print("MIA 3P% ", MIA_tpp)
print("MIA FT% ", MIA_ftp)
print("MIA Def. Reb. ", MIA_dreb)
print("MIA Avg. Assists ", MIA_ast)
print("MIA Avg. Turnovers ", MIA_tov)
print("MIA Avg. Steals ", MIA_stl)
print("MIA Avg. Blocks ", MIA_blk)
print("MIA Avg. Personal Fouls Drawn ", MIA_pfd)
print("MIA Off. Rtg ", MIA_offrtg)
print("MIA Def. Rtg ", MIA_defrtg)
print("MIA Clutch Win% ", MIA_clutchwinpct)
```

```python
TOR_scoring = TOR_meanpts*(TOR_fgp*TOR_tpp*TOR_ftp+TOR_offrtg)
TOR_misc = TOR_scoring+TOR_dreb+TOR_ast+TOR_stl+TOR_blk+TOR_pfd+TOR_defrtg
TOR_top = TOR_misc*TOR_winpct*TOR_clutchwinpct
TOR_score = TOR_top/(TOR_tov*10000)
TOR_score

print(TOR_score)
```

```python
MIA_scoring = MIA_meanpts*(MIA_fgp*MIA_tpp*MIA_ftp+MIA_offrtg)
MIA_misc = MIA_scoring+MIA_dreb+MIA_ast+MIA_stl+MIA_blk+MIA_pfd+MIA_defrtg
MIA_top = MIA_misc*MIA_winpct*MIA_clutchwinpct
MIA_score = MIA_top/(MIA_tov*10000)
MIA_score

print(MIA_score)
```

```python
TOR_meanpoints = TOR1df.TeamPTS.mean()
TOR_meanfgp = TOR1df.TeamFGP.mean()
TOR_sdpts = TOR1df.TeamPTS.std()
TOR_sdfgp = TOR1df.TeamFGP.std()
TOR_oppmeanpoints = TOR1df.OpponentPTS.mean()
TOR_oppfgp = TOR1df.OpponentFGP.mean()
TOR_oppstdpts = TOR1df.OpponentPTS.std()
TOR_oppstdfgp = TOR1df.OpponentFGP.std()

print("TOR Avg. Pts ", TOR_meanpoints)
print("TOR Avg. FG% ", TOR_meanfgp)
print("TOR Pts Deviation ", TOR_sdpts)
print("TOR FG% Deviation ", TOR_sdfgp)
print("TOR Opp. Avg. Pts ", TOR_oppmeanpoints)
print("TOR Opp. Avg. FG% ", TOR_oppfgp)
print("TOR Opp. Pts Deviation ", TOR_oppstdpts)
print("TOR Opp. FG% Deviation ", TOR_oppstdfgp)
```

```python
MIA_meanpoints = MIA1df.TeamPTS.mean()
MIA_meanfgp = MIA1df.TeamFGP.mean()
MIA_sdpts = MIA1df.TeamPTS.std()
MIA_sdfgp = MIA1df.TeamFGP.std()
MIA_oppmeanpoints = MIA1df.OpponentPTS.mean()
MIA_oppfgp = MIA1df.OpponentFGP.mean()
MIA_oppstdpts = MIA1df.OpponentPTS.std()
MIA_oppstdfgp = MIA1df.OpponentFGP.std()

print("MIA Avg. Pts ", MIA_meanpoints)
print("MIA Avg. FG% ", MIA_meanfgp)
print("MIA Pts Deviation ", MIA_sdpts)
print("MIA FG% Deviation ", MIA_sdfgp)
print("MIA Opp. Avg. Pts ", MIA_oppmeanpoints)
print("MIA Opp. Avg. FG% ", MIA_oppfgp)
print("MIA Opp. Pts Deviation ", MIA_oppstdpts)
print("MIA Opp. FG% Deviation ", MIA_oppstdfgp)
```

```python
def gameSim():
    TORScore = (rnd.gauss(TOR_score,TOR_sdpts)+ rnd.gauss(TOR_oppmeanpoints,TOR_oppstdpts))/2
    MIAScore = (rnd.gauss(MIA_score,MIA_sdpts)+ rnd.gauss(MIA_oppmeanpoints,MIA_oppstdpts))/2
    if int(round(TORScore)) > int(round(MIAScore)):
        return 1
    elif int(round(TORScore)) < int(round(MIAScore)):
        return -1
    else: return 0
```

```python
def gamesSim(ns):
    gamesout = []
    team1win = 0
    team2win = 0
    for i in range(ns):
        gm = gameSim()
        gamesout.append(gm)
        if gm == 1:
            team1win +=1 
        elif gm == -1:
            team2win +=1
    print('TOR Win ', round((team1win/(team1win+team2win))*100,2),'%')
    print('MIA Win ', round((team2win/(team1win+team2win))*100,2),'%')
    return gamesout
```

```python
gamesSim(10000)
```
