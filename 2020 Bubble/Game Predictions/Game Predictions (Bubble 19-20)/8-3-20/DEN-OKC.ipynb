```python
DEN_meanpoints = DEN1df.TeamPTS.mean()
DEN_meanfgp = DEN1df.TeamFGP.mean()
DEN_sdpts = DEN1df.TeamPTS.std()
DEN_sdfgp = DEN1df.TeamFGP.std()
DEN_oppmeanpoints = DEN1df.OpponentPTS.mean()
DEN_oppfgp = DEN1df.OpponentFGP.mean()
DEN_oppstdpts = DEN1df.OpponentPTS.std()
DEN_oppstdfgp = DEN1df.OpponentFGP.std()

print("DEN Avg. Pts ", DEN_meanpoints)
print("DEN Avg. FG% ", DEN_meanfgp)
print("DEN Pts Deviation ", DEN_sdpts)
print("DEN FG% Deviation ", DEN_sdfgp)
print("DEN Opp. Avg. Pts ", DEN_oppmeanpoints)
print("DEN Opp. Avg. FG% ", DEN_oppfgp)
print("DEN Opp. Pts Deviation ", DEN_oppstdpts)
print("DEN Opp. FG% Deviation ", DEN_oppstdfgp)
```

```python
OKC_meanpoints = OKC1df.TeamPTS.mean()
OKC_meanfgp = OKC1df.TeamFGP.mean()
OKC_sdpts = OKC1df.TeamPTS.std()
OKC_sdfgp = OKC1df.TeamFGP.std()
OKC_oppmeanpoints = OKC1df.OpponentPTS.mean()
OKC_oppfgp = OKC1df.OpponentFGP.mean()
OKC_oppstdpts = OKC1df.OpponentPTS.std()
OKC_oppstdfgp = OKC1df.OpponentFGP.std()

print("OKC Avg. Pts ", OKC_meanpoints)
print("OKC Avg. FG% ", OKC_meanfgp)
print("OKC Pts Deviation ", OKC_sdpts)
print("OKC FG% Deviation ", OKC_sdfgp)
print("OKC Opp. Avg. Pts ", OKC_oppmeanpoints)
print("OKC Opp. Avg. FG% ", OKC_oppfgp)
print("OKC Opp. Pts Deviation ", OKC_oppstdpts)
print("OKC Opp. FG% Deviation ", OKC_oppstdfgp)
```

```python
def gameSim():
    DENScore = (rnd.gauss(DEN_score,DEN_sdpts)+ rnd.gauss(DEN_oppmeanpoints,DEN_oppstdpts))/2
    OKCScore = (rnd.gauss(OKC_score,OKC_sdpts)+ rnd.gauss(OKC_oppmeanpoints,OKC_oppstdpts))/2
    if int(round(DENScore)) > int(round(OKCScore)):
        return 1
    elif int(round(DENScore)) < int(round(OKCScore)):
        return -1
    else: return 0
```

```python
def gamesSim(ns):
    gamesout = []
    team1win = 0
    team2win = 0
    for i in range(ns):
        gm = gameSim()
        gamesout.append(gm)
        if gm == 1:
            team1win +=1 
        elif gm == -1:
            team2win +=1
    print('DEN Win ', round((team1win/(team1win+team2win))*100,2),'%')
    print('OKC Win ', round((team2win/(team1win+team2win))*100,2),'%')
    return gamesout
```
