```python
SASdf = gdf[gdf.TEAM == 'San Antonio Spurs']
NOPdf = gdf[gdf.TEAM == 'New Orleans Pelicans']
```

```python
SAS_winpct = SASdf.WINP.mean()
SAS_meanpts = SASdf.PTS.mean()
SAS_fgp = SASdf.FGP.mean()
SAS_tpp = SASdf.TPP.mean()
SAS_ftp = SASdf.FTP.mean()
SAS_dreb = SASdf.DREB.mean()
SAS_ast = SASdf.AST.mean()
SAS_tov = SASdf.TOV.mean()
SAS_stl = SASdf.STL.mean()
SAS_blk = SASdf.BLK.mean()
SAS_pfd = SASdf.PFD.mean()
SAS_offrtg = SASdf.ORTG.mean()
SAS_defrtg = SASdf.DRTG.mean()
SAS_clutchwinpct = SASdf.CWINP.mean()

print("SAS Win% ", SAS_winpct)
print("SAS Avg. Points ", SAS_meanpts)
print("SAS FG% ", SAS_fgp)
print("SAS 3P% ", SAS_tpp)
print("SAS FT% ", SAS_ftp)
print("SAS Def. Reb. ", SAS_dreb)
print("SAS Avg. Assists ", SAS_ast)
print("SAS Avg. Turnovers ", SAS_tov)
print("SAS Avg. Steals ", SAS_stl)
print("SAS Avg. Blocks ", SAS_blk)
print("SAS Avg. Personal Fouls Drawn ", SAS_pfd)
print("SAS Off. Rtg ", SAS_offrtg)
print("SAS Def. Rtg ", SAS_defrtg)
print("SAS Clutch Win% ", SAS_clutchwinpct)
```

```python
NOP_winpct = NOPdf.WINP.mean()
NOP_meanpts = NOPdf.PTS.mean()
NOP_fgp = NOPdf.FGP.mean()
NOP_tpp = NOPdf.TPP.mean()
NOP_ftp = NOPdf.FTP.mean()
NOP_dreb = NOPdf.DREB.mean()
NOP_ast = NOPdf.AST.mean()
NOP_tov = NOPdf.TOV.mean()
NOP_stl = NOPdf.STL.mean()
NOP_blk = NOPdf.BLK.mean()
NOP_pfd = NOPdf.PFD.mean()
NOP_offrtg = SASdf.ORTG.mean()
NOP_defrtg = SASdf.DRTG.mean()
NOP_clutchwinpct = SASdf.CWINP.mean()

print("NOP Win% ", NOP_winpct)
print("NOP Avg. Points ", NOP_meanpts)
print("NOP FG% ", NOP_fgp)
print("NOP 3P% ", NOP_tpp)
print("NOP FT% ", NOP_ftp)
print("NOP Def. Reb. ", NOP_dreb)
print("NOP Avg. Assists ", NOP_ast)
print("NOP Avg. Turnovers ", NOP_tov)
print("NOP Avg. Steals ", NOP_stl)
print("NOP Avg. Blocks ", NOP_blk)
print("NOP Avg. Personal Fouls Drawn ", NOP_pfd)
print("NOP Off. Rtg ", NOP_offrtg)
print("NOP Def. Rtg ", NOP_defrtg)
print("NOP Clutch Win% ", NOP_clutchwinpct)
```

```python
SAS_scoring = SAS_meanpts*(SAS_fgp*SAS_tpp*SAS_ftp+SAS_offrtg)
SAS_misc = SAS_scoring+SAS_dreb+SAS_ast+SAS_stl+SAS_blk+SAS_pfd+SAS_defrtg
SAS_top = SAS_misc*SAS_winpct*SAS_clutchwinpct
SAS_score = SAS_top/(SAS_tov*10000)
SAS_score

print(SAS_score)
```

```python
NOP_scoring = NOP_meanpts*(NOP_fgp*NOP_tpp*NOP_ftp+NOP_offrtg)
NOP_misc = NOP_scoring+NOP_dreb+NOP_ast+NOP_stl+NOP_blk+NOP_pfd+NOP_defrtg
NOP_top = NOP_misc*NOP_winpct*NOP_clutchwinpct
NOP_score = NOP_top/(NOP_tov*10000)
NOP_score

print(NOP_score)
```

```python
SAS_meanpoints = SAS1df.TeamPTS.mean()
SAS_meanfgp = SAS1df.TeamFGP.mean()
SAS_sdpts = SAS1df.TeamPTS.std()
SAS_sdfgp = SAS1df.TeamFGP.std()
SAS_oppmeanpoints = SAS1df.OpponentPTS.mean()
SAS_oppfgp = SAS1df.OpponentFGP.mean()
SAS_oppstdpts = SAS1df.OpponentPTS.std()
SAS_oppstdfgp = SAS1df.OpponentFGP.std()

print("SAS Avg. Pts ", SAS_meanpoints)
print("SAS Avg. FG% ", SAS_meanfgp)
print("SAS Pts Deviation ", SAS_sdpts)
print("SAS FG% Deviation ", SAS_sdfgp)
print("SAS Opp. Avg. Pts ", SAS_oppmeanpoints)
print("SAS Opp. Avg. FG% ", SAS_oppfgp)
print("SAS Opp. Pts Deviation ", SAS_oppstdpts)
print("SAS Opp. FG% Deviation ", SAS_oppstdfgp)
```

```python
NOP_meanpoints = NOP1df.TeamPTS.mean()
NOP_meanfgp = NOP1df.TeamFGP.mean()
NOP_sdpts = NOP1df.TeamPTS.std()
NOP_sdfgp = NOP1df.TeamFGP.std()
NOP_oppmeanpoints = NOP1df.OpponentPTS.mean()
NOP_oppfgp = NOP1df.OpponentFGP.mean()
NOP_oppstdpts = NOP1df.OpponentPTS.std()
NOP_oppstdfgp = NOP1df.OpponentFGP.std()

print("NOP Avg. Pts ", NOP_meanpoints)
print("NOP Avg. FG% ", NOP_meanfgp)
print("NOP Pts Deviation ", NOP_sdpts)
print("NOP FG% Deviation ", NOP_sdfgp)
print("NOP Opp. Avg. Pts ", NOP_oppmeanpoints)
print("NOP Opp. Avg. FG% ", NOP_oppfgp)
print("NOP Opp. Pts Deviation ", NOP_oppstdpts)
print("NOP Opp. FG% Deviation ", NOP_oppstdfgp)
```

```python
def gameSim():
    SASScore = (rnd.gauss(SAS_score,SAS_sdpts)+ rnd.gauss(SAS_oppmeanpoints,SAS_oppstdpts))/2
    NOPScore = (rnd.gauss(NOP_score,NOP_sdpts)+ rnd.gauss(NOP_oppmeanpoints,NOP_oppstdpts))/2
    if int(round(SASScore)) > int(round(NOPScore)):
        return 1
    elif int(round(SASScore)) < int(round(NOPScore)):
        return -1
    else: return 0
```

```python
def gamesSim(ns):
    gamesout = []
    team1win = 0
    team2win = 0
    for i in range(ns):
        gm = gameSim()
        gamesout.append(gm)
        if gm == 1:
            team1win +=1 
        elif gm == -1:
            team2win +=1
    print('SAS Win ', round((team1win/(team1win+team2win))*100,2),'%')
    print('NOP Win ', round((team2win/(team1win+team2win))*100,2),'%')
    return gamesout
```

```python
gamesSim(10000)
```
