```python
import pandas as pd 
import random as rnd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
MEMdf = gdf[gdf.TEAM == 'Memphis Grizzlies']
TORdf = gdf[gdf.TEAM == 'Toronto Raptors']
```

```python
MEM_winpct = MEMdf.WINP.mean()
MEM_meanpts = MEMdf.PTS.mean()
MEM_fgp = MEMdf.FGP.mean()
MEM_tpp = MEMdf.TPP.mean()
MEM_ftp = MEMdf.FTP.mean()
MEM_dreb = MEMdf.DREB.mean()
MEM_ast = MEMdf.AST.mean()
MEM_tov = MEMdf.TOV.mean()
MEM_stl = MEMdf.STL.mean()
MEM_blk = MEMdf.BLK.mean()
MEM_pfd = MEMdf.PFD.mean()
MEM_offrtg = MEMdf.ORTG.mean()
MEM_defrtg = MEMdf.DRTG.mean()
MEM_clutchwinpct = MEMdf.CWINP.mean()

print("MEM Win% ", MEM_winpct)
print("MEM Avg. Points ", MEM_meanpts)
print("MEM FG% ", MEM_fgp)
print("MEM 3P% ", MEM_tpp)
print("MEM FT% ", MEM_ftp)
print("MEM Def. Reb. ", MEM_dreb)
print("MEM Avg. Assists ", MEM_ast)
print("MEM Avg. Turnovers ", MEM_tov)
print("MEM Avg. Steals ", MEM_stl)
print("MEM Avg. Blocks ", MEM_blk)
print("MEM Avg. Personal Fouls Drawn ", MEM_pfd)
print("MEM Off. Rtg ", MEM_offrtg)
print("MEM Def. Rtg ", MEM_defrtg)
print("MEM Clutch Win% ", MEM_clutchwinpct)
```

```python
TOR_winpct = TORdf.WINP.mean()
TOR_meanpts = TORdf.PTS.mean()
TOR_fgp = TORdf.FGP.mean()
TOR_tpp = TORdf.TPP.mean()
TOR_ftp = TORdf.FTP.mean()
TOR_dreb = TORdf.DREB.mean()
TOR_ast = TORdf.AST.mean()
TOR_tov = TORdf.TOV.mean()
TOR_stl = TORdf.STL.mean()
TOR_blk = TORdf.BLK.mean()
TOR_pfd = TORdf.PFD.mean()
TOR_offrtg = MEMdf.ORTG.mean()
TOR_defrtg = MEMdf.DRTG.mean()
TOR_clutchwinpct = MEMdf.CWINP.mean()

print("TOR Win% ", TOR_winpct)
print("TOR Avg. Points ", TOR_meanpts)
print("TOR FG% ", TOR_fgp)
print("TOR 3P% ", TOR_tpp)
print("TOR FT% ", TOR_ftp)
print("TOR Def. Reb. ", TOR_dreb)
print("TOR Avg. Assists ", TOR_ast)
print("TOR Avg. Turnovers ", TOR_tov)
print("TOR Avg. Steals ", TOR_stl)
print("TOR Avg. Blocks ", TOR_blk)
print("TOR Avg. Personal Fouls Drawn ", TOR_pfd)
print("TOR Off. Rtg ", TOR_offrtg)
print("TOR Def. Rtg ", TOR_defrtg)
print("TOR Clutch Win% ", TOR_clutchwinpct)
```

```python
MEM_scoring = MEM_meanpts*(MEM_fgp*MEM_tpp*MEM_ftp+MEM_offrtg)
MEM_misc = MEM_scoring+MEM_dreb+MEM_ast+MEM_stl+MEM_blk+MEM_pfd+MEM_defrtg
MEM_top = MEM_misc*MEM_winpct*MEM_clutchwinpct
MEM_score = MEM_top/(MEM_tov*10000)
MEM_score

print(MEM_score)
```

```python
TOR_scoring = TOR_meanpts*(TOR_fgp*TOR_tpp*TOR_ftp+TOR_offrtg)
TOR_misc = TOR_scoring+TOR_dreb+TOR_ast+TOR_stl+TOR_blk+TOR_pfd+TOR_defrtg
TOR_top = TOR_misc*TOR_winpct*TOR_clutchwinpct
TOR_score = TOR_top/(TOR_tov*10000)
TOR_score

print(TOR_score)
```

```python
MEM_meanpoints = MEM1df.TeamPTS.mean()
MEM_meanfgp = MEM1df.TeamFGP.mean()
MEM_sdpts = MEM1df.TeamPTS.std()
MEM_sdfgp = MEM1df.TeamFGP.std()
MEM_oppmeanpoints = MEM1df.OpponentPTS.mean()
MEM_oppfgp = MEM1df.OpponentFGP.mean()
MEM_oppstdpts = MEM1df.OpponentPTS.std()
MEM_oppstdfgp = MEM1df.OpponentFGP.std()

print("MEM Avg. Pts ", MEM_meanpoints)
print("MEM Avg. FG% ", MEM_meanfgp)
print("MEM Pts Deviation ", MEM_sdpts)
print("MEM FG% Deviation ", MEM_sdfgp)
print("MEM Opp. Avg. Pts ", MEM_oppmeanpoints)
print("MEM Opp. Avg. FG% ", MEM_oppfgp)
print("MEM Opp. Pts Deviation ", MEM_oppstdpts)
print("MEM Opp. FG% Deviation ", MEM_oppstdfgp)
```

```python
TOR_meanpoints = TOR1df.TeamPTS.mean()
TOR_meanfgp = TOR1df.TeamFGP.mean()
TOR_sdpts = TOR1df.TeamPTS.std()
TOR_sdfgp = TOR1df.TeamFGP.std()
TOR_oppmeanpoints = TOR1df.OpponentPTS.mean()
TOR_oppfgp = TOR1df.OpponentFGP.mean()
TOR_oppstdpts = TOR1df.OpponentPTS.std()
TOR_oppstdfgp = TOR1df.OpponentFGP.std()

print("TOR Avg. Pts ", TOR_meanpoints)
print("TOR Avg. FG% ", TOR_meanfgp)
print("TOR Pts Deviation ", TOR_sdpts)
print("TOR FG% Deviation ", TOR_sdfgp)
print("TOR Opp. Avg. Pts ", TOR_oppmeanpoints)
print("TOR Opp. Avg. FG% ", TOR_oppfgp)
print("TOR Opp. Pts Deviation ", TOR_oppstdpts)
print("TOR Opp. FG% Deviation ", TOR_oppstdfgp)
```

```python
def gameSim():
    MEMScore = (rnd.gauss(MEM_score,MEM_sdpts)+ rnd.gauss(MEM_oppmeanpoints,MEM_oppstdpts))/2
    TORScore = (rnd.gauss(TOR_score,TOR_sdpts)+ rnd.gauss(TOR_oppmeanpoints,TOR_oppstdpts))/2
    if int(round(MEMScore)) > int(round(TORScore)):
        return 1
    elif int(round(MEMScore)) < int(round(TORScore)):
        return -1
    else: return 0
```

```python
def gamesSim(ns):
    gamesout = []
    team1win = 0
    team2win = 0
    for i in range(ns):
        gm = gameSim()
        gamesout.append(gm)
        if gm == 1:
            team1win +=1 
        elif gm == -1:
            team2win +=1
    print('MEM Win ', round((team1win/(team1win+team2win))*100,2),'%')
    print('TOR Win ', round((team2win/(team1win+team2win))*100,2),'%')
    return gamesout
```

```python
gamesSim(10000)
```
