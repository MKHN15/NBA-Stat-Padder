```python
ORL_meanpoints = ORL1df.TeamPTS.mean()
ORL_meanfgp = ORL1df.TeamFGP.mean()
ORL_sdpts = ORL1df.TeamPTS.std()
ORL_sdfgp = ORL1df.TeamFGP.std()
ORL_oppmeanpoints = ORL1df.OpponentPTS.mean()
ORL_oppfgp = ORL1df.OpponentFGP.mean()
ORL_oppstdpts = ORL1df.OpponentPTS.std()
ORL_oppstdfgp = ORL1df.OpponentFGP.std()

print("ORL Avg. Pts ", ORL_meanpoints)
print("ORL Avg. FG% ", ORL_meanfgp)
print("ORL Pts Deviation ", ORL_sdpts)
print("ORL FG% Deviation ", ORL_sdfgp)
print("ORL Opp. Avg. Pts ", ORL_oppmeanpoints)
print("ORL Opp. Avg. FG% ", ORL_oppfgp)
print("ORL Opp. Pts Deviation ", ORL_oppstdpts)
print("ORL Opp. FG% Deviation ", ORL_oppstdfgp)
```

```python
BOS_meanpoints = BOS1df.TeamPTS.mean()
BOS_meanfgp = BOS1df.TeamFGP.mean()
BOS_sdpts = BOS1df.TeamPTS.std()
BOS_sdfgp = BOS1df.TeamFGP.std()
BOS_oppmeanpoints = BOS1df.OpponentPTS.mean()
BOS_oppfgp = BOS1df.OpponentFGP.mean()
BOS_oppstdpts = BOS1df.OpponentPTS.std()
BOS_oppstdfgp = BOS1df.OpponentFGP.std()

print("BOS Avg. Pts ", BOS_meanpoints)
print("BOS Avg. FG% ", BOS_meanfgp)
print("BOS Pts Deviation ", BOS_sdpts)
print("BOS FG% Deviation ", BOS_sdfgp)
print("BOS Opp. Avg. Pts ", BOS_oppmeanpoints)
print("BOS Opp. Avg. FG% ", BOS_oppfgp)
print("BOS Opp. Pts Deviation ", BOS_oppstdpts)
print("BOS Opp. FG% Deviation ", BOS_oppstdfgp)
```

```python
def gameSim():
    ORLScore = (rnd.gauss(ORL_score,ORL_sdpts)+ rnd.gauss(ORL_oppmeanpoints,ORL_oppstdpts))/2
    BOSScore = (rnd.gauss(BOS_score,BOS_sdpts)+ rnd.gauss(BOS_oppmeanpoints,BOS_oppstdpts))/2
    if int(round(ORLScore)) > int(round(BOSScore)):
        return 1
    elif int(round(ORLScore)) < int(round(BOSScore)):
        return -1
    else: return 0
```

```python
def gamesSim(ns):
    gamesout = []
    team1win = 0
    team2win = 0
    for i in range(ns):
        gm = gameSim()
        gamesout.append(gm)
        if gm == 1:
            team1win +=1 
        elif gm == -1:
            team2win +=1
    print('ORL Win ', round((team1win/(team1win+team2win))*100,2),'%')
    print('BOS Win ', round((team2win/(team1win+team2win))*100,2),'%')
    return gamesout
```

```python
gamesSim(10000)
```
