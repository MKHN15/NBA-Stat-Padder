```python
SAC_meanpoints = SAC1df.TeamPTS.mean()
SAC_meanfgp = SAC1df.TeamFGP.mean()
SAC_sdpts = SAC1df.TeamPTS.std()
SAC_sdfgp = SAC1df.TeamFGP.std()
SAC_oppmeanpoints = SAC1df.OpponentPTS.mean()
SAC_oppfgp = SAC1df.OpponentFGP.mean()
SAC_oppstdpts = SAC1df.OpponentPTS.std()
SAC_oppstdfgp = SAC1df.OpponentFGP.std()

print("SAC Avg. Pts ", SAC_meanpoints)
print("SAC Avg. FG% ", SAC_meanfgp)
print("SAC Pts Deviation ", SAC_sdpts)
print("SAC FG% Deviation ", SAC_sdfgp)
print("SAC Opp. Avg. Pts ", SAC_oppmeanpoints)
print("SAC Opp. Avg. FG% ", SAC_oppfgp)
print("SAC Opp. Pts Deviation ", SAC_oppstdpts)
print("SAC Opp. FG% Deviation ", SAC_oppstdfgp)
```

```python
LAL_meanpoints = LAL1df.TeamPTS.mean()
LAL_meanfgp = LAL1df.TeamFGP.mean()
LAL_sdpts = LAL1df.TeamPTS.std()
LAL_sdfgp = LAL1df.TeamFGP.std()
LAL_oppmeanpoints = LAL1df.OpponentPTS.mean()
LAL_oppfgp = LAL1df.OpponentFGP.mean()
LAL_oppstdpts = LAL1df.OpponentPTS.std()
LAL_oppstdfgp = LAL1df.OpponentFGP.std()

print("LAL Avg. Pts ", LAL_meanpoints)
print("LAL Avg. FG% ", LAL_meanfgp)
print("LAL Pts Deviation ", LAL_sdpts)
print("LAL FG% Deviation ", LAL_sdfgp)
print("LAL Opp. Avg. Pts ", LAL_oppmeanpoints)
print("LAL Opp. Avg. FG% ", LAL_oppfgp)
print("LAL Opp. Pts Deviation ", LAL_oppstdpts)
print("LAL Opp. FG% Deviation ", LAL_oppstdfgp)
```

```python
def gameSim():
    SACScore = (rnd.gauss(SAC_score,SAC_sdpts)+ rnd.gauss(SAC_oppmeanpoints,SAC_oppstdpts))/2
    LALScore = (rnd.gauss(LAL_score,LAL_sdpts)+ rnd.gauss(LAL_oppmeanpoints,LAL_oppstdpts))/2
    if int(round(SACScore)) > int(round(LALScore)):
        return 1
    elif int(round(SACScore)) < int(round(LALScore)):
        return -1
    else: return 0
```

```python
def gamesSim(ns):
    gamesout = []
    team1win = 0
    team2win = 0
    for i in range(ns):
        gm = gameSim()
        gamesout.append(gm)
        if gm == 1:
            team1win +=1 
        elif gm == -1:
            team2win +=1
    print('SAC Win ', round((team1win/(team1win+team2win))*100,2),'%')
    print('LAL Win ', round((team2win/(team1win+team2win))*100,2),'%')
    return gamesout
```
