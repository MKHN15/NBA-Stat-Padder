```python
POR_meanpoints = POR1df.TeamPTS.mean()
POR_meanfgp = POR1df.TeamFGP.mean()
POR_sdpts = POR1df.TeamPTS.std()
POR_sdfgp = POR1df.TeamFGP.std()
POR_oppmeanpoints = POR1df.OpponentPTS.mean()
POR_oppfgp = POR1df.OpponentFGP.mean()
POR_oppstdpts = POR1df.OpponentPTS.std()
POR_oppstdfgp = POR1df.OpponentFGP.std()

print("POR Avg. Pts ", POR_meanpoints)
print("POR Avg. FG% ", POR_meanfgp)
print("POR Pts Deviation ", POR_sdpts)
print("POR FG% Deviation ", POR_sdfgp)
print("POR Opp. Avg. Pts ", POR_oppmeanpoints)
print("POR Opp. Avg. FG% ", POR_oppfgp)
print("POR Opp. Pts Deviation ", POR_oppstdpts)
print("POR Opp. FG% Deviation ", POR_oppstdfgp)
```

```python
BKN_meanpoints = BKN1df.TeamPTS.mean()
BKN_meanfgp = BKN1df.TeamFGP.mean()
BKN_sdpts = BKN1df.TeamPTS.std()
BKN_sdfgp = BKN1df.TeamFGP.std()
BKN_oppmeanpoints = BKN1df.OpponentPTS.mean()
BKN_oppfgp = BKN1df.OpponentFGP.mean()
BKN_oppstdpts = BKN1df.OpponentPTS.std()
BKN_oppstdfgp = BKN1df.OpponentFGP.std()

print("BKN Avg. Pts ", BKN_meanpoints)
print("BKN Avg. FG% ", BKN_meanfgp)
print("BKN Pts Deviation ", BKN_sdpts)
print("BKN FG% Deviation ", BKN_sdfgp)
print("BKN Opp. Avg. Pts ", BKN_oppmeanpoints)
print("BKN Opp. Avg. FG% ", BKN_oppfgp)
print("BKN Opp. Pts Deviation ", BKN_oppstdpts)
print("BKN Opp. FG% Deviation ", BKN_oppstdfgp)
```

```python
def gameSim():
    PORScore = (rnd.gauss(POR_score,POR_sdpts)+ rnd.gauss(POR_oppmeanpoints,POR_oppstdpts))/2
    BKNScore = (rnd.gauss(BKN_score,BKN_sdpts)+ rnd.gauss(BKN_oppmeanpoints,BKN_oppstdpts))/2
    if int(round(PORScore)) > int(round(BKNScore)):
        return 1
    elif int(round(PORScore)) < int(round(BKNScore)):
        return -1
    else: return 0
```

```python
def gamesSim(ns):
    gamesout = []
    team1win = 0
    team2win = 0
    for i in range(ns):
        gm = gameSim()
        gamesout.append(gm)
        if gm == 1:
            team1win +=1 
        elif gm == -1:
            team2win +=1
    print('POR Win ', round((team1win/(team1win+team2win))*100,2),'%')
    print('BKN Win ', round((team2win/(team1win+team2win))*100,2),'%')
    return gamesout
```

```python
gamesSim(10000)
```
