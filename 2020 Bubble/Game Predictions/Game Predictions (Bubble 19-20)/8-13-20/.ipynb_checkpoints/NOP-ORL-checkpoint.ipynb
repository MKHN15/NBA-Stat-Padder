```python
NOP_meanpoints = NOP1df.TeamPTS.mean()
NOP_meanfgp = NOP1df.TeamFGP.mean()
NOP_sdpts = NOP1df.TeamPTS.std()
NOP_sdfgp = NOP1df.TeamFGP.std()
NOP_oppmeanpoints = NOP1df.OpponentPTS.mean()
NOP_oppfgp = NOP1df.OpponentFGP.mean()
NOP_oppstdpts = NOP1df.OpponentPTS.std()
NOP_oppstdfgp = NOP1df.OpponentFGP.std()

print("NOP Avg. Pts ", NOP_meanpoints)
print("NOP Avg. FG% ", NOP_meanfgp)
print("NOP Pts Deviation ", NOP_sdpts)
print("NOP FG% Deviation ", NOP_sdfgp)
print("NOP Opp. Avg. Pts ", NOP_oppmeanpoints)
print("NOP Opp. Avg. FG% ", NOP_oppfgp)
print("NOP Opp. Pts Deviation ", NOP_oppstdpts)
print("NOP Opp. FG% Deviation ", NOP_oppstdfgp)
```

```python
ORL_meanpoints = ORL1df.TeamPTS.mean()
ORL_meanfgp = ORL1df.TeamFGP.mean()
ORL_sdpts = ORL1df.TeamPTS.std()
ORL_sdfgp = ORL1df.TeamFGP.std()
ORL_oppmeanpoints = ORL1df.OpponentPTS.mean()
ORL_oppfgp = ORL1df.OpponentFGP.mean()
ORL_oppstdpts = ORL1df.OpponentPTS.std()
ORL_oppstdfgp = ORL1df.OpponentFGP.std()

print("ORL Avg. Pts ", ORL_meanpoints)
print("ORL Avg. FG% ", ORL_meanfgp)
print("ORL Pts Deviation ", ORL_sdpts)
print("ORL FG% Deviation ", ORL_sdfgp)
print("ORL Opp. Avg. Pts ", ORL_oppmeanpoints)
print("ORL Opp. Avg. FG% ", ORL_oppfgp)
print("ORL Opp. Pts Deviation ", ORL_oppstdpts)
print("ORL Opp. FG% Deviation ", ORL_oppstdfgp)
```

```python
def gameSim():
    NOPScore = (rnd.gauss(NOP_score,NOP_sdpts)+ rnd.gauss(NOP_oppmeanpoints,NOP_oppstdpts))/2
    ORLScore = (rnd.gauss(ORL_score,ORL_sdpts)+ rnd.gauss(ORL_oppmeanpoints,ORL_oppstdpts))/2
    if int(round(NOPScore)) > int(round(ORLScore)):
        return 1
    elif int(round(NOPScore)) < int(round(ORLScore)):
        return -1
    else: return 0
```

```python
def gamesSim(ns):
    gamesout = []
    team1win = 0
    team2win = 0
    for i in range(ns):
        gm = gameSim()
        gamesout.append(gm)
        if gm == 1:
            team1win +=1 
        elif gm == -1:
            team2win +=1
    print('NOP Win ', round((team1win/(team1win+team2win))*100,2),'%')
    print('ORL Win ', round((team2win/(team1win+team2win))*100,2),'%')
    return gamesout
```

```python
gamesSim(10000)
```
