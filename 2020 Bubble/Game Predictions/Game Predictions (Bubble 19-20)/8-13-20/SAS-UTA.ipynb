```python
SAS_meanpoints = SAS1df.TeamPTS.mean()
SAS_meanfgp = SAS1df.TeamFGP.mean()
SAS_sdpts = SAS1df.TeamPTS.std()
SAS_sdfgp = SAS1df.TeamFGP.std()
SAS_oppmeanpoints = SAS1df.OpponentPTS.mean()
SAS_oppfgp = SAS1df.OpponentFGP.mean()
SAS_oppstdpts = SAS1df.OpponentPTS.std()
SAS_oppstdfgp = SAS1df.OpponentFGP.std()

print("SAS Avg. Pts ", SAS_meanpoints)
print("SAS Avg. FG% ", SAS_meanfgp)
print("SAS Pts Deviation ", SAS_sdpts)
print("SAS FG% Deviation ", SAS_sdfgp)
print("SAS Opp. Avg. Pts ", SAS_oppmeanpoints)
print("SAS Opp. Avg. FG% ", SAS_oppfgp)
print("SAS Opp. Pts Deviation ", SAS_oppstdpts)
print("SAS Opp. FG% Deviation ", SAS_oppstdfgp)
```

```python
UTA_meanpoints = UTA1df.TeamPTS.mean()
UTA_meanfgp = UTA1df.TeamFGP.mean()
UTA_sdpts = UTA1df.TeamPTS.std()
UTA_sdfgp = UTA1df.TeamFGP.std()
UTA_oppmeanpoints = UTA1df.OpponentPTS.mean()
UTA_oppfgp = UTA1df.OpponentFGP.mean()
UTA_oppstdpts = UTA1df.OpponentPTS.std()
UTA_oppstdfgp = UTA1df.OpponentFGP.std()

print("UTA Avg. Pts ", UTA_meanpoints)
print("UTA Avg. FG% ", UTA_meanfgp)
print("UTA Pts Deviation ", UTA_sdpts)
print("UTA FG% Deviation ", UTA_sdfgp)
print("UTA Opp. Avg. Pts ", UTA_oppmeanpoints)
print("UTA Opp. Avg. FG% ", UTA_oppfgp)
print("UTA Opp. Pts Deviation ", UTA_oppstdpts)
print("UTA Opp. FG% Deviation ", UTA_oppstdfgp)
```

```python
def gameSim():
    SASScore = (rnd.gauss(SAS_score,SAS_sdpts)+ rnd.gauss(SAS_oppmeanpoints,SAS_oppstdpts))/2
    UTAScore = (rnd.gauss(UTA_score,UTA_sdpts)+ rnd.gauss(UTA_oppmeanpoints,UTA_oppstdpts))/2
    if int(round(SASScore)) > int(round(UTAScore)):
        return 1
    elif int(round(SASScore)) < int(round(UTAScore)):
        return -1
    else: return 0
```

```python
def gamesSim(ns):
    gamesout = []
    team1win = 0
    team2win = 0
    for i in range(ns):
        gm = gameSim()
        gamesout.append(gm)
        if gm == 1:
            team1win +=1 
        elif gm == -1:
            team2win +=1
    print('SAS Win ', round((team1win/(team1win+team2win))*100,2),'%')
    print('UTA Win ', round((team2win/(team1win+team2win))*100,2),'%')
    return gamesout
```
