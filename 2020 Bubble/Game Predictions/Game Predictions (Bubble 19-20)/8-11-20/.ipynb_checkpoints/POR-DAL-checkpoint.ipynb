```python
DAL_winpct = DALdf.WINP.mean()
DAL_meanpts = DALdf.PTS.mean()
DAL_fgp = DALdf.FGP.mean()
DAL_tpp = DALdf.TPP.mean()
DAL_ftp = DALdf.FTP.mean()
DAL_dreb = DALdf.DREB.mean()
DAL_ast = DALdf.AST.mean()
DAL_tov = DALdf.TOV.mean()
DAL_stl = DALdf.STL.mean()
DAL_blk = DALdf.BLK.mean()
DAL_pfd = DALdf.PFD.mean()
DAL_offrtg = PORdf.ORTG.mean()
DAL_defrtg = PORdf.DRTG.mean()
DAL_clutchwinpct = PORdf.CWINP.mean()

print("DAL Win% ", DAL_winpct)
print("DAL Avg. Points ", DAL_meanpts)
print("DAL FG% ", DAL_fgp)
print("DAL 3P% ", DAL_tpp)
print("DAL FT% ", DAL_ftp)
print("DAL Def. Reb. ", DAL_dreb)
print("DAL Avg. Assists ", DAL_ast)
print("DAL Avg. Turnovers ", DAL_tov)
print("DAL Avg. Steals ", DAL_stl)
print("DAL Avg. Blocks ", DAL_blk)
print("DAL Avg. Personal Fouls Drawn ", DAL_pfd)
print("DAL Off. Rtg ", DAL_offrtg)
print("DAL Def. Rtg ", DAL_defrtg)
print("DAL Clutch Win% ", DAL_clutchwinpct)
```

```python
POR_scoring = POR_meanpts*(POR_fgp*POR_tpp*POR_ftp+POR_offrtg)
POR_misc = POR_scoring+POR_dreb+POR_ast+POR_stl+POR_blk+POR_pfd+POR_defrtg
POR_top = POR_misc*POR_winpct*POR_clutchwinpct
POR_score = POR_top/(POR_tov*10000)
POR_score

print(POR_score)
```

```python
DAL_scoring = DAL_meanpts*(DAL_fgp*DAL_tpp*DAL_ftp+DAL_offrtg)
DAL_misc = DAL_scoring+DAL_dreb+DAL_ast+DAL_stl+DAL_blk+DAL_pfd+DAL_defrtg
DAL_top = DAL_misc*DAL_winpct*DAL_clutchwinpct
DAL_score = DAL_top/(DAL_tov*10000)
DAL_score

print(DAL_score)
```

```python
POR_meanpoints = POR1df.TeamPTS.mean()
POR_meanfgp = POR1df.TeamFGP.mean()
POR_sdpts = POR1df.TeamPTS.std()
POR_sdfgp = POR1df.TeamFGP.std()
POR_oppmeanpoints = POR1df.OpponentPTS.mean()
POR_oppfgp = POR1df.OpponentFGP.mean()
POR_oppstdpts = POR1df.OpponentPTS.std()
POR_oppstdfgp = POR1df.OpponentFGP.std()

print("POR Avg. Pts ", POR_meanpoints)
print("POR Avg. FG% ", POR_meanfgp)
print("POR Pts Deviation ", POR_sdpts)
print("POR FG% Deviation ", POR_sdfgp)
print("POR Opp. Avg. Pts ", POR_oppmeanpoints)
print("POR Opp. Avg. FG% ", POR_oppfgp)
print("POR Opp. Pts Deviation ", POR_oppstdpts)
print("POR Opp. FG% Deviation ", POR_oppstdfgp)
```

```python
DAL_meanpoints = DAL1df.TeamPTS.mean()
DAL_meanfgp = DAL1df.TeamFGP.mean()
DAL_sdpts = DAL1df.TeamPTS.std()
DAL_sdfgp = DAL1df.TeamFGP.std()
DAL_oppmeanpoints = DAL1df.OpponentPTS.mean()
DAL_oppfgp = DAL1df.OpponentFGP.mean()
DAL_oppstdpts = DAL1df.OpponentPTS.std()
DAL_oppstdfgp = DAL1df.OpponentFGP.std()

print("DAL Avg. Pts ", DAL_meanpoints)
print("DAL Avg. FG% ", DAL_meanfgp)
print("DAL Pts Deviation ", DAL_sdpts)
print("DAL FG% Deviation ", DAL_sdfgp)
print("DAL Opp. Avg. Pts ", DAL_oppmeanpoints)
print("DAL Opp. Avg. FG% ", DAL_oppfgp)
print("DAL Opp. Pts Deviation ", DAL_oppstdpts)
print("DAL Opp. FG% Deviation ", DAL_oppstdfgp)
```

```python
def gameSim():
    PORScore = (rnd.gauss(POR_score,POR_sdpts)+ rnd.gauss(POR_oppmeanpoints,POR_oppstdpts))/2
    DALScore = (rnd.gauss(DAL_score,DAL_sdpts)+ rnd.gauss(DAL_oppmeanpoints,DAL_oppstdpts))/2
    if int(round(PORScore)) > int(round(DALScore)):
        return 1
    elif int(round(PORScore)) < int(round(DALScore)):
        return -1
    else: return 0
```

```python
def gamesSim(ns):
    gamesout = []
    team1win = 0
    team2win = 0
    for i in range(ns):
        gm = gameSim()
        gamesout.append(gm)
        if gm == 1:
            team1win +=1 
        elif gm == -1:
            team2win +=1
    print('POR Win ', round((team1win/(team1win+team2win))*100,2),'%')
    print('DAL Win ', round((team2win/(team1win+team2win))*100,2),'%')
    return gamesout
```

```python
gamesSim(10000)
```
