```python
BOS_meanpoints = BOS1df.TeamPTS.mean()
BOS_meanfgp = BOS1df.TeamFGP.mean()
BOS_sdpts = BOS1df.TeamPTS.std()
BOS_sdfgp = BOS1df.TeamFGP.std()
BOS_oppmeanpoints = BOS1df.OpponentPTS.mean()
BOS_oppfgp = BOS1df.OpponentFGP.mean()
BOS_oppstdpts = BOS1df.OpponentPTS.std()
BOS_oppstdfgp = BOS1df.OpponentFGP.std()

print("BOS Avg. Pts ", BOS_meanpoints)
print("BOS Avg. FG% ", BOS_meanfgp)
print("BOS Pts Deviation ", BOS_sdpts)
print("BOS FG% Deviation ", BOS_sdfgp)
print("BOS Opp. Avg. Pts ", BOS_oppmeanpoints)
print("BOS Opp. Avg. FG% ", BOS_oppfgp)
print("BOS Opp. Pts Deviation ", BOS_oppstdpts)
print("BOS Opp. FG% Deviation ", BOS_oppstdfgp)
```

```python
MEM_meanpoints = MEM1df.TeamPTS.mean()
MEM_meanfgp = MEM1df.TeamFGP.mean()
MEM_sdpts = MEM1df.TeamPTS.std()
MEM_sdfgp = MEM1df.TeamFGP.std()
MEM_oppmeanpoints = MEM1df.OpponentPTS.mean()
MEM_oppfgp = MEM1df.OpponentFGP.mean()
MEM_oppstdpts = MEM1df.OpponentPTS.std()
MEM_oppstdfgp = MEM1df.OpponentFGP.std()

print("MEM Avg. Pts ", MEM_meanpoints)
print("MEM Avg. FG% ", MEM_meanfgp)
print("MEM Pts Deviation ", MEM_sdpts)
print("MEM FG% Deviation ", MEM_sdfgp)
print("MEM Opp. Avg. Pts ", MEM_oppmeanpoints)
print("MEM Opp. Avg. FG% ", MEM_oppfgp)
print("MEM Opp. Pts Deviation ", MEM_oppstdpts)
print("MEM Opp. FG% Deviation ", MEM_oppstdfgp)
```

```python
def gameSim():
    BOSScore = (rnd.gauss(BOS_score,BOS_sdpts)+ rnd.gauss(BOS_oppmeanpoints,BOS_oppstdpts))/2
    MEMScore = (rnd.gauss(MEM_score,MEM_sdpts)+ rnd.gauss(MEM_oppmeanpoints,MEM_oppstdpts))/2
    if int(round(BOSScore)) > int(round(MEMScore)):
        return 1
    elif int(round(BOSScore)) < int(round(MEMScore)):
        return -1
    else: return 0
```

```python
def gamesSim(ns):
    gamesout = []
    team1win = 0
    team2win = 0
    for i in range(ns):
        gm = gameSim()
        gamesout.append(gm)
        if gm == 1:
            team1win +=1 
        elif gm == -1:
            team2win +=1
    print('BOS Win ', round((team1win/(team1win+team2win))*100,2),'%')
    print('MEM Win ', round((team2win/(team1win+team2win))*100,2),'%')
    return gamesout
```

```python
gamesSim(10000)
```
