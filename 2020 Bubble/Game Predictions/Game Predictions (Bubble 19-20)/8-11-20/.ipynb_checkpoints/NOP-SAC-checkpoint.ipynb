```python
NOP_meanpoints = NOP1df.TeamPTS.mean()
NOP_meanfgp = NOP1df.TeamFGP.mean()
NOP_sdpts = NOP1df.TeamPTS.std()
NOP_sdfgp = NOP1df.TeamFGP.std()
NOP_oppmeanpoints = NOP1df.OpponentPTS.mean()
NOP_oppfgp = NOP1df.OpponentFGP.mean()
NOP_oppstdpts = NOP1df.OpponentPTS.std()
NOP_oppstdfgp = NOP1df.OpponentFGP.std()

print("NOP Avg. Pts ", NOP_meanpoints)
print("NOP Avg. FG% ", NOP_meanfgp)
print("NOP Pts Deviation ", NOP_sdpts)
print("NOP FG% Deviation ", NOP_sdfgp)
print("NOP Opp. Avg. Pts ", NOP_oppmeanpoints)
print("NOP Opp. Avg. FG% ", NOP_oppfgp)
print("NOP Opp. Pts Deviation ", NOP_oppstdpts)
print("NOP Opp. FG% Deviation ", NOP_oppstdfgp)
```

```python
SAC_meanpoints = SAC1df.TeamPTS.mean()
SAC_meanfgp = SAC1df.TeamFGP.mean()
SAC_sdpts = SAC1df.TeamPTS.std()
SAC_sdfgp = SAC1df.TeamFGP.std()
SAC_oppmeanpoints = SAC1df.OpponentPTS.mean()
SAC_oppfgp = SAC1df.OpponentFGP.mean()
SAC_oppstdpts = SAC1df.OpponentPTS.std()
SAC_oppstdfgp = SAC1df.OpponentFGP.std()

print("SAC Avg. Pts ", SAC_meanpoints)
print("SAC Avg. FG% ", SAC_meanfgp)
print("SAC Pts Deviation ", SAC_sdpts)
print("SAC FG% Deviation ", SAC_sdfgp)
print("SAC Opp. Avg. Pts ", SAC_oppmeanpoints)
print("SAC Opp. Avg. FG% ", SAC_oppfgp)
print("SAC Opp. Pts Deviation ", SAC_oppstdpts)
print("SAC Opp. FG% Deviation ", SAC_oppstdfgp)
```

```python
def gameSim():
    NOPScore = (rnd.gauss(NOP_score,NOP_sdpts)+ rnd.gauss(NOP_oppmeanpoints,NOP_oppstdpts))/2
    SACScore = (rnd.gauss(SAC_score,SAC_sdpts)+ rnd.gauss(SAC_oppmeanpoints,SAC_oppstdpts))/2
    if int(round(NOPScore)) > int(round(SACScore)):
        return 1
    elif int(round(NOPScore)) < int(round(SACScore)):
        return -1
    else: return 0
```

```python
def gamesSim(ns):
    gamesout = []
    team1win = 0
    team2win = 0
    for i in range(ns):
        gm = gameSim()
        gamesout.append(gm)
        if gm == 1:
            team1win +=1 
        elif gm == -1:
            team2win +=1
    print('NOP Win ', round((team1win/(team1win+team2win))*100,2),'%')
    print('SAC Win ', round((team2win/(team1win+team2win))*100,2),'%')
    return gamesout
```

```python
gamesSim(10000)
```
