```python
UTA_meanpoints = UTA1df.TeamPTS.mean()
UTA_meanfgp = UTA1df.TeamFGP.mean()
UTA_sdpts = UTA1df.TeamPTS.std()
UTA_sdfgp = UTA1df.TeamFGP.std()
UTA_oppmeanpoints = UTA1df.OpponentPTS.mean()
UTA_oppfgp = UTA1df.OpponentFGP.mean()
UTA_oppstdpts = UTA1df.OpponentPTS.std()
UTA_oppstdfgp = UTA1df.OpponentFGP.std()

print("UTA Avg. Pts ", UTA_meanpoints)
print("UTA Avg. FG% ", UTA_meanfgp)
print("UTA Pts Deviation ", UTA_sdpts)
print("UTA FG% Deviation ", UTA_sdfgp)
print("UTA Opp. Avg. Pts ", UTA_oppmeanpoints)
print("UTA Opp. Avg. FG% ", UTA_oppfgp)
print("UTA Opp. Pts Deviation ", UTA_oppstdpts)
print("UTA Opp. FG% Deviation ", UTA_oppstdfgp)
```

```python
OKC_meanpoints = OKC1df.TeamPTS.mean()
OKC_meanfgp = OKC1df.TeamFGP.mean()
OKC_sdpts = OKC1df.TeamPTS.std()
OKC_sdfgp = OKC1df.TeamFGP.std()
OKC_oppmeanpoints = OKC1df.OpponentPTS.mean()
OKC_oppfgp = OKC1df.OpponentFGP.mean()
OKC_oppstdpts = OKC1df.OpponentPTS.std()
OKC_oppstdfgp = OKC1df.OpponentFGP.std()

print("OKC Avg. Pts ", OKC_meanpoints)
print("OKC Avg. FG% ", OKC_meanfgp)
print("OKC Pts Deviation ", OKC_sdpts)
print("OKC FG% Deviation ", OKC_sdfgp)
print("OKC Opp. Avg. Pts ", OKC_oppmeanpoints)
print("OKC Opp. Avg. FG% ", OKC_oppfgp)
print("OKC Opp. Pts Deviation ", OKC_oppstdpts)
print("OKC Opp. FG% Deviation ", OKC_oppstdfgp)
```

```python
def gameSim():
    UTAScore = (rnd.gauss(UTA_score,UTA_sdpts)+ rnd.gauss(UTA_oppmeanpoints,UTA_oppstdpts))/2
    OKCScore = (rnd.gauss(OKC_score,OKC_sdpts)+ rnd.gauss(OKC_oppmeanpoints,OKC_oppstdpts))/2
    if int(round(UTAScore)) > int(round(OKCScore)):
        return 1
    elif int(round(UTAScore)) < int(round(OKCScore)):
        return -1
    else: return 0
```

```python
def gamesSim(ns):
    gamesout = []
    team1win = 0
    team2win = 0
    for i in range(ns):
        gm = gameSim()
        gamesout.append(gm)
        if gm == 1:
            team1win +=1 
        elif gm == -1:
            team2win +=1
    print('UTA Win ', round((team1win/(team1win+team2win))*100,2),'%')
    print('OKC Win ', round((team2win/(team1win+team2win))*100,2),'%')
    return gamesout
```

```python
gamesSim(10000)
```
