```python
PHI_meanpoints = PHI1df.TeamPTS.mean()
PHI_meanfgp = PHI1df.TeamFGP.mean()
PHI_sdpts = PHI1df.TeamPTS.std()
PHI_sdfgp = PHI1df.TeamFGP.std()
PHI_oppmeanpoints = PHI1df.OpponentPTS.mean()
PHI_oppfgp = PHI1df.OpponentFGP.mean()
PHI_oppstdpts = PHI1df.OpponentPTS.std()
PHI_oppstdfgp = PHI1df.OpponentFGP.std()

print("PHI Avg. Pts ", PHI_meanpoints)
print("PHI Avg. FG% ", PHI_meanfgp)
print("PHI Pts Deviation ", PHI_sdpts)
print("PHI FG% Deviation ", PHI_sdfgp)
print("PHI Opp. Avg. Pts ", PHI_oppmeanpoints)
print("PHI Opp. Avg. FG% ", PHI_oppfgp)
print("PHI Opp. Pts Deviation ", PHI_oppstdpts)
print("PHI Opp. FG% Deviation ", PHI_oppstdfgp)
```

```python
IND_meanpoints = IND1df.TeamPTS.mean()
IND_meanfgp = IND1df.TeamFGP.mean()
IND_sdpts = IND1df.TeamPTS.std()
IND_sdfgp = IND1df.TeamFGP.std()
IND_oppmeanpoints = IND1df.OpponentPTS.mean()
IND_oppfgp = IND1df.OpponentFGP.mean()
IND_oppstdpts = IND1df.OpponentPTS.std()
IND_oppstdfgp = IND1df.OpponentFGP.std()

print("IND Avg. Pts ", IND_meanpoints)
print("IND Avg. FG% ", IND_meanfgp)
print("IND Pts Deviation ", IND_sdpts)
print("IND FG% Deviation ", IND_sdfgp)
print("IND Opp. Avg. Pts ", IND_oppmeanpoints)
print("IND Opp. Avg. FG% ", IND_oppfgp)
print("IND Opp. Pts Deviation ", IND_oppstdpts)
print("IND Opp. FG% Deviation ", IND_oppstdfgp)
```

```python
def gameSim():
    PHIScore = (rnd.gauss(PHI_score,PHI_sdpts)+ rnd.gauss(PHI_oppmeanpoints,PHI_oppstdpts))/2
    INDScore = (rnd.gauss(IND_score,IND_sdpts)+ rnd.gauss(IND_oppmeanpoints,IND_oppstdpts))/2
    if int(round(PHIScore)) > int(round(INDScore)):
        return 1
    elif int(round(PHIScore)) < int(round(INDScore)):
        return -1
    else: return 0
```

```python
def gamesSim(ns):
    gamesout = []
    team1win = 0
    team2win = 0
    for i in range(ns):
        gm = gameSim()
        gamesout.append(gm)
        if gm == 1:
            team1win +=1 
        elif gm == -1:
            team2win +=1
    print('PHI Win ', round((team1win/(team1win+team2win))*100,2),'%')
    print('IND Win ', round((team2win/(team1win+team2win))*100,2),'%')
    return gamesout
```

```python
gamesSim(10000)
```
