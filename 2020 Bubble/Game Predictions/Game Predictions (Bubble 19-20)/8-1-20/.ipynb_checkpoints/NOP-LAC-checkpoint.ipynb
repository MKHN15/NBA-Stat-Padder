```python
LAC_meanpoints = LAC1df.TeamPTS.mean()
LAC_meanfgp = LAC1df.TeamFGP.mean()
LAC_sdpts = LAC1df.TeamPTS.std()
LAC_sdfgp = LAC1df.TeamFGP.std()
LAC_oppmeanpoints = LAC1df.OpponentPTS.mean()
LAC_oppfgp = LAC1df.OpponentFGP.mean()
LAC_oppstdpts = LAC1df.OpponentPTS.std()
LAC_oppstdfgp = LAC1df.OpponentFGP.std()

print("LAC Avg. Pts ", LAC_meanpoints)
print("LAC Avg. FG% ", LAC_meanfgp)
print("LAC Pts Deviation ", LAC_sdpts)
print("LAC FG% Deviation ", LAC_sdfgp)
print("LAC Opp. Avg. Pts ", LAC_oppmeanpoints)
print("LAC Opp. Avg. FG% ", LAC_oppfgp)
print("LAC Opp. Pts Deviation ", LAC_oppstdpts)
print("LAC Opp. FG% Deviation ", LAC_oppstdfgp)
```

```python
NOP_meanpoints = NOP1df.TeamPTS.mean()
NOP_meanfgp = NOP1df.TeamFGP.mean()
NOP_sdpts = NOP1df.TeamPTS.std()
NOP_sdfgp = NOP1df.TeamFGP.std()
NOP_oppmeanpoints = NOP1df.OpponentPTS.mean()
NOP_oppfgp = NOP1df.OpponentFGP.mean()
NOP_oppstdpts = NOP1df.OpponentPTS.std()
NOP_oppstdfgp = NOP1df.OpponentFGP.std()

print("NOP Avg. Pts ", NOP_meanpoints)
print("NOP Avg. FG% ", NOP_meanfgp)
print("NOP Pts Deviation ", NOP_sdpts)
print("NOP FG% Deviation ", NOP_sdfgp)
print("NOP Opp. Avg. Pts ", NOP_oppmeanpoints)
print("NOP Opp. Avg. FG% ", NOP_oppfgp)
print("NOP Opp. Pts Deviation ", NOP_oppstdpts)
print("NOP Opp. FG% Deviation ", NOP_oppstdfgp)
```

```python
def gameSim():
    LACScore = (rnd.gauss(LAC_score,LAC_sdpts)+ rnd.gauss(LAC_oppmeanpoints,LAC_oppstdpts))/2
    NOPScore = (rnd.gauss(NOP_score,NOP_sdpts)+ rnd.gauss(NOP_oppmeanpoints,NOP_oppstdpts))/2
    if int(round(LACScore)) > int(round(NOPScore)):
        return 1
    elif int(round(LACScore)) < int(round(NOPScore)):
        return -1
    else: return 0
```

```python
def gamesSim(ns):
    gamesout = []
    team1win = 0
    team2win = 0
    for i in range(ns):
        gm = gameSim()
        gamesout.append(gm)
        if gm == 1:
            team1win +=1 
        elif gm == -1:
            team2win +=1
    print('LAC Win ', round((team1win/(team1win+team2win))*100,2),'%')
    print('NOP Win ', round((team2win/(team1win+team2win))*100,2),'%')
    return gamesout
```

```python
gamesSim(10000)
```
