```python
MIA_winpct = MIAdf.WINP.mean()
MIA_meanpts = MIAdf.PTS.mean()
MIA_fgp = MIAdf.FGP.mean()
MIA_tpp = MIAdf.TPP.mean()
MIA_ftp = MIAdf.FTP.mean()
MIA_dreb = MIAdf.DREB.mean()
MIA_ast = MIAdf.AST.mean()
MIA_tov = MIAdf.TOV.mean()
MIA_stl = MIAdf.STL.mean()
MIA_blk = MIAdf.BLK.mean()
MIA_pfd = MIAdf.PFD.mean()
MIA_offrtg = MIAdf.ORTG.mean()
MIA_defrtg = MIAdf.DRTG.mean()
MIA_clutchwinpct = MIAdf.CWINP.mean()

print("MIA Win% ", MIA_winpct)
print("MIA Avg. Points ", MIA_meanpts)
print("MIA FG% ", MIA_fgp)
print("MIA 3P% ", MIA_tpp)
print("MIA FT% ", MIA_ftp)
print("MIA Def. Reb. ", MIA_dreb)
print("MIA Avg. Assists ", MIA_ast)
print("MIA Avg. Turnovers ", MIA_tov)
print("MIA Avg. Steals ", MIA_stl)
print("MIA Avg. Blocks ", MIA_blk)
print("MIA Avg. Personal Fouls Drawn ", MIA_pfd)
print("MIA Off. Rtg ", MIA_offrtg)
print("MIA Def. Rtg ", MIA_defrtg)
print("MIA Clutch Win% ", MIA_clutchwinpct)
```

```python
DEN_winpct = DENdf.WINP.mean()
DEN_meanpts = DENdf.PTS.mean()
DEN_fgp = DENdf.FGP.mean()
DEN_tpp = DENdf.TPP.mean()
DEN_ftp = DENdf.FTP.mean()
DEN_dreb = DENdf.DREB.mean()
DEN_ast = DENdf.AST.mean()
DEN_tov = DENdf.TOV.mean()
DEN_stl = DENdf.STL.mean()
DEN_blk = DENdf.BLK.mean()
DEN_pfd = DENdf.PFD.mean()
DEN_offrtg = MIAdf.ORTG.mean()
DEN_defrtg = MIAdf.DRTG.mean()
DEN_clutchwinpct = MIAdf.CWINP.mean()

print("DEN Win% ", DEN_winpct)
print("DEN Avg. Points ", DEN_meanpts)
print("DEN FG% ", DEN_fgp)
print("DEN 3P% ", DEN_tpp)
print("DEN FT% ", DEN_ftp)
print("DEN Def. Reb. ", DEN_dreb)
print("DEN Avg. Assists ", DEN_ast)
print("DEN Avg. Turnovers ", DEN_tov)
print("DEN Avg. Steals ", DEN_stl)
print("DEN Avg. Blocks ", DEN_blk)
print("DEN Avg. Personal Fouls Drawn ", DEN_pfd)
print("DEN Off. Rtg ", DEN_offrtg)
print("DEN Def. Rtg ", DEN_defrtg)
print("DEN Clutch Win% ", DEN_clutchwinpct)
```

```python
MIA_scoring = MIA_meanpts*(MIA_fgp*MIA_tpp*MIA_ftp+MIA_offrtg)
MIA_misc = MIA_scoring+MIA_dreb+MIA_ast+MIA_stl+MIA_blk+MIA_pfd+MIA_defrtg
MIA_top = MIA_misc*MIA_winpct*MIA_clutchwinpct
MIA_score = MIA_top/(MIA_tov*10000)
MIA_score

print(MIA_score)
```

```python
DEN_scoring = DEN_meanpts*(DEN_fgp*DEN_tpp*DEN_ftp+DEN_offrtg)
DEN_misc = DEN_scoring+DEN_dreb+DEN_ast+DEN_stl+DEN_blk+DEN_pfd+DEN_defrtg
DEN_top = DEN_misc*DEN_winpct*DEN_clutchwinpct
DEN_score = DEN_top/(DEN_tov*10000)
DEN_score

print(DEN_score)
```

```python
MIA_meanpoints = MIA1df.TeamPTS.mean()
MIA_meanfgp = MIA1df.TeamFGP.mean()
MIA_sdpts = MIA1df.TeamPTS.std()
MIA_sdfgp = MIA1df.TeamFGP.std()
MIA_oppmeanpoints = MIA1df.OpponentPTS.mean()
MIA_oppfgp = MIA1df.OpponentFGP.mean()
MIA_oppstdpts = MIA1df.OpponentPTS.std()
MIA_oppstdfgp = MIA1df.OpponentFGP.std()

print("MIA Avg. Pts ", MIA_meanpoints)
print("MIA Avg. FG% ", MIA_meanfgp)
print("MIA Pts Deviation ", MIA_sdpts)
print("MIA FG% Deviation ", MIA_sdfgp)
print("MIA Opp. Avg. Pts ", MIA_oppmeanpoints)
print("MIA Opp. Avg. FG% ", MIA_oppfgp)
print("MIA Opp. Pts Deviation ", MIA_oppstdpts)
print("MIA Opp. FG% Deviation ", MIA_oppstdfgp)
```

```python
DEN_meanpoints = DEN1df.TeamPTS.mean()
DEN_meanfgp = DEN1df.TeamFGP.mean()
DEN_sdpts = DEN1df.TeamPTS.std()
DEN_sdfgp = DEN1df.TeamFGP.std()
DEN_oppmeanpoints = DEN1df.OpponentPTS.mean()
DEN_oppfgp = DEN1df.OpponentFGP.mean()
DEN_oppstdpts = DEN1df.OpponentPTS.std()
DEN_oppstdfgp = DEN1df.OpponentFGP.std()

print("DEN Avg. Pts ", DEN_meanpoints)
print("DEN Avg. FG% ", DEN_meanfgp)
print("DEN Pts Deviation ", DEN_sdpts)
print("DEN FG% Deviation ", DEN_sdfgp)
print("DEN Opp. Avg. Pts ", DEN_oppmeanpoints)
print("DEN Opp. Avg. FG% ", DEN_oppfgp)
print("DEN Opp. Pts Deviation ", DEN_oppstdpts)
print("DEN Opp. FG% Deviation ", DEN_oppstdfgp)
```

```python
def gameSim():
    MIAScore = (rnd.gauss(MIA_score,MIA_sdpts)+ rnd.gauss(MIA_oppmeanpoints,MIA_oppstdpts))/2
    DENScore = (rnd.gauss(DEN_score,DEN_sdpts)+ rnd.gauss(DEN_oppmeanpoints,DEN_oppstdpts))/2
    if int(round(MIAScore)) > int(round(DENScore)):
        return 1
    elif int(round(MIAScore)) < int(round(DENScore)):
        return -1
    else: return 0
```

```python
def gamesSim(ns):
    gamesout = []
    team1win = 0
    team2win = 0
    for i in range(ns):
        gm = gameSim()
        gamesout.append(gm)
        if gm == 1:
            team1win +=1 
        elif gm == -1:
            team2win +=1
    print('MIA Win ', round((team1win/(team1win+team2win))*100,2),'%')
    print('DEN Win ', round((team2win/(team1win+team2win))*100,2),'%')
    return gamesout
```

```python
gamesSim(10000)
```
