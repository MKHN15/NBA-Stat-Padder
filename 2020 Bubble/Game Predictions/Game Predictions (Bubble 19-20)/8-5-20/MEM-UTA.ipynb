```python
MEM_meanpoints = MEM1df.TeamPTS.mean()
MEM_meanfgp = MEM1df.TeamFGP.mean()
MEM_sdpts = MEM1df.TeamPTS.std()
MEM_sdfgp = MEM1df.TeamFGP.std()
MEM_oppmeanpoints = MEM1df.OpponentPTS.mean()
MEM_oppfgp = MEM1df.OpponentFGP.mean()
MEM_oppstdpts = MEM1df.OpponentPTS.std()
MEM_oppstdfgp = MEM1df.OpponentFGP.std()

print("MEM Avg. Pts ", MEM_meanpoints)
print("MEM Avg. FG% ", MEM_meanfgp)
print("MEM Pts Deviation ", MEM_sdpts)
print("MEM FG% Deviation ", MEM_sdfgp)
print("MEM Opp. Avg. Pts ", MEM_oppmeanpoints)
print("MEM Opp. Avg. FG% ", MEM_oppfgp)
print("MEM Opp. Pts Deviation ", MEM_oppstdpts)
print("MEM Opp. FG% Deviation ", MEM_oppstdfgp)
```

```python
UTA_meanpoints = UTA1df.TeamPTS.mean()
UTA_meanfgp = UTA1df.TeamFGP.mean()
UTA_sdpts = UTA1df.TeamPTS.std()
UTA_sdfgp = UTA1df.TeamFGP.std()
UTA_oppmeanpoints = UTA1df.OpponentPTS.mean()
UTA_oppfgp = UTA1df.OpponentFGP.mean()
UTA_oppstdpts = UTA1df.OpponentPTS.std()
UTA_oppstdfgp = UTA1df.OpponentFGP.std()

print("UTA Avg. Pts ", UTA_meanpoints)
print("UTA Avg. FG% ", UTA_meanfgp)
print("UTA Pts Deviation ", UTA_sdpts)
print("UTA FG% Deviation ", UTA_sdfgp)
print("UTA Opp. Avg. Pts ", UTA_oppmeanpoints)
print("UTA Opp. Avg. FG% ", UTA_oppfgp)
print("UTA Opp. Pts Deviation ", UTA_oppstdpts)
print("UTA Opp. FG% Deviation ", UTA_oppstdfgp)
```

```python
def gameSim():
    MEMScore = (rnd.gauss(MEM_score,MEM_sdpts)+ rnd.gauss(MEM_oppmeanpoints,MEM_oppstdpts))/2
    UTAScore = (rnd.gauss(UTA_score,UTA_sdpts)+ rnd.gauss(UTA_oppmeanpoints,UTA_oppstdpts))/2
    if int(round(MEMScore)) > int(round(UTAScore)):
        return 1
    elif int(round(MEMScore)) < int(round(UTAScore)):
        return -1
    else: return 0
```

```python
def gamesSim(ns):
    gamesout = []
    team1win = 0
    team2win = 0
    for i in range(ns):
        gm = gameSim()
        gamesout.append(gm)
        if gm == 1:
            team1win +=1 
        elif gm == -1:
            team2win +=1
    print('MEM Win ', round((team1win/(team1win+team2win))*100,2),'%')
    print('UTA Win ', round((team2win/(team1win+team2win))*100,2),'%')
    return gamesout
```
