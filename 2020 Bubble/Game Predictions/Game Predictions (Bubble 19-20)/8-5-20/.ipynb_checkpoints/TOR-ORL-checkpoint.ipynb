```python
TOR_meanpoints = TOR1df.TeamPTS.mean()
TOR_meanfgp = TOR1df.TeamFGP.mean()
TOR_sdpts = TOR1df.TeamPTS.std()
TOR_sdfgp = TOR1df.TeamFGP.std()
TOR_oppmeanpoints = TOR1df.OpponentPTS.mean()
TOR_oppfgp = TOR1df.OpponentFGP.mean()
TOR_oppstdpts = TOR1df.OpponentPTS.std()
TOR_oppstdfgp = TOR1df.OpponentFGP.std()

print("TOR Avg. Pts ", TOR_meanpoints)
print("TOR Avg. FG% ", TOR_meanfgp)
print("TOR Pts Deviation ", TOR_sdpts)
print("TOR FG% Deviation ", TOR_sdfgp)
print("TOR Opp. Avg. Pts ", TOR_oppmeanpoints)
print("TOR Opp. Avg. FG% ", TOR_oppfgp)
print("TOR Opp. Pts Deviation ", TOR_oppstdpts)
print("TOR Opp. FG% Deviation ", TOR_oppstdfgp)
```

```python
ORL_meanpoints = ORL1df.TeamPTS.mean()
ORL_meanfgp = ORL1df.TeamFGP.mean()
ORL_sdpts = ORL1df.TeamPTS.std()
ORL_sdfgp = ORL1df.TeamFGP.std()
ORL_oppmeanpoints = ORL1df.OpponentPTS.mean()
ORL_oppfgp = ORL1df.OpponentFGP.mean()
ORL_oppstdpts = ORL1df.OpponentPTS.std()
ORL_oppstdfgp = ORL1df.OpponentFGP.std()

print("ORL Avg. Pts ", ORL_meanpoints)
print("ORL Avg. FG% ", ORL_meanfgp)
print("ORL Pts Deviation ", ORL_sdpts)
print("ORL FG% Deviation ", ORL_sdfgp)
print("ORL Opp. Avg. Pts ", ORL_oppmeanpoints)
print("ORL Opp. Avg. FG% ", ORL_oppfgp)
print("ORL Opp. Pts Deviation ", ORL_oppstdpts)
print("ORL Opp. FG% Deviation ", ORL_oppstdfgp)
```

```python
def gameSim():
    TORScore = (rnd.gauss(TOR_score,TOR_sdpts)+ rnd.gauss(TOR_oppmeanpoints,TOR_oppstdpts))/2
    ORLScore = (rnd.gauss(ORL_score,ORL_sdpts)+ rnd.gauss(ORL_oppmeanpoints,ORL_oppstdpts))/2
    if int(round(TORScore)) > int(round(ORLScore)):
        return 1
    elif int(round(TORScore)) < int(round(ORLScore)):
        return -1
    else: return 0
```

```python
def gamesSim(ns):
    gamesout = []
    team1win = 0
    team2win = 0
    for i in range(ns):
        gm = gameSim()
        gamesout.append(gm)
        if gm == 1:
            team1win +=1 
        elif gm == -1:
            team2win +=1
    print('TOR Win ', round((team1win/(team1win+team2win))*100,2),'%')
    print('ORL Win ', round((team2win/(team1win+team2win))*100,2),'%')
    return gamesout
```

```python
gamesSim(10000)
```
