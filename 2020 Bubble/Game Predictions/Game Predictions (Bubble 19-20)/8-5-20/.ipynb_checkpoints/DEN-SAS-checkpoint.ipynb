```python
DEN_winpct = DENdf.WINP.mean()
DEN_meanpts = DENdf.PTS.mean()
DEN_fgp = DENdf.FGP.mean()
DEN_tpp = DENdf.TPP.mean()
DEN_ftp = DENdf.FTP.mean()
DEN_dreb = DENdf.DREB.mean()
DEN_ast = DENdf.AST.mean()
DEN_tov = DENdf.TOV.mean()
DEN_stl = DENdf.STL.mean()
DEN_blk = DENdf.BLK.mean()
DEN_pfd = DENdf.PFD.mean()
DEN_offrtg = DENdf.ORTG.mean()
DEN_defrtg = DENdf.DRTG.mean()
DEN_clutchwinpct = DENdf.CWINP.mean()

print("DEN Win% ", DEN_winpct)
print("DEN Avg. Points ", DEN_meanpts)
print("DEN FG% ", DEN_fgp)
print("DEN 3P% ", DEN_tpp)
print("DEN FT% ", DEN_ftp)
print("DEN Def. Reb. ", DEN_dreb)
print("DEN Avg. Assists ", DEN_ast)
print("DEN Avg. Turnovers ", DEN_tov)
print("DEN Avg. Steals ", DEN_stl)
print("DEN Avg. Blocks ", DEN_blk)
print("DEN Avg. Personal Fouls Drawn ", DEN_pfd)
print("DEN Off. Rtg ", DEN_offrtg)
print("DEN Def. Rtg ", DEN_defrtg)
print("DEN Clutch Win% ", DEN_clutchwinpct)
```

```python
SAS_winpct = SASdf.WINP.mean()
SAS_meanpts = SASdf.PTS.mean()
SAS_fgp = SASdf.FGP.mean()
SAS_tpp = SASdf.TPP.mean()
SAS_ftp = SASdf.FTP.mean()
SAS_dreb = SASdf.DREB.mean()
SAS_ast = SASdf.AST.mean()
SAS_tov = SASdf.TOV.mean()
SAS_stl = SASdf.STL.mean()
SAS_blk = SASdf.BLK.mean()
SAS_pfd = SASdf.PFD.mean()
SAS_offrtg = DENdf.ORTG.mean()
SAS_defrtg = DENdf.DRTG.mean()
SAS_clutchwinpct = DENdf.CWINP.mean()

print("SAS Win% ", SAS_winpct)
print("SAS Avg. Points ", SAS_meanpts)
print("SAS FG% ", SAS_fgp)
print("SAS 3P% ", SAS_tpp)
print("SAS FT% ", SAS_ftp)
print("SAS Def. Reb. ", SAS_dreb)
print("SAS Avg. Assists ", SAS_ast)
print("SAS Avg. Turnovers ", SAS_tov)
print("SAS Avg. Steals ", SAS_stl)
print("SAS Avg. Blocks ", SAS_blk)
print("SAS Avg. Personal Fouls Drawn ", SAS_pfd)
print("SAS Off. Rtg ", SAS_offrtg)
print("SAS Def. Rtg ", SAS_defrtg)
print("SAS Clutch Win% ", SAS_clutchwinpct)
```

```python
DEN_scoring = DEN_meanpts*(DEN_fgp*DEN_tpp*DEN_ftp+DEN_offrtg)
DEN_misc = DEN_scoring+DEN_dreb+DEN_ast+DEN_stl+DEN_blk+DEN_pfd+DEN_defrtg
DEN_top = DEN_misc*DEN_winpct*DEN_clutchwinpct
DEN_score = DEN_top/(DEN_tov*10000)
DEN_score

print(DEN_score)
```

```python
SAS_scoring = SAS_meanpts*(SAS_fgp*SAS_tpp*SAS_ftp+SAS_offrtg)
SAS_misc = SAS_scoring+SAS_dreb+SAS_ast+SAS_stl+SAS_blk+SAS_pfd+SAS_defrtg
SAS_top = SAS_misc*SAS_winpct*SAS_clutchwinpct
SAS_score = SAS_top/(SAS_tov*10000)
SAS_score

print(SAS_score)
```

```python
DEN_meanpoints = DEN1df.TeamPTS.mean()
DEN_meanfgp = DEN1df.TeamFGP.mean()
DEN_sdpts = DEN1df.TeamPTS.std()
DEN_sdfgp = DEN1df.TeamFGP.std()
DEN_oppmeanpoints = DEN1df.OpponentPTS.mean()
DEN_oppfgp = DEN1df.OpponentFGP.mean()
DEN_oppstdpts = DEN1df.OpponentPTS.std()
DEN_oppstdfgp = DEN1df.OpponentFGP.std()

print("DEN Avg. Pts ", DEN_meanpoints)
print("DEN Avg. FG% ", DEN_meanfgp)
print("DEN Pts Deviation ", DEN_sdpts)
print("DEN FG% Deviation ", DEN_sdfgp)
print("DEN Opp. Avg. Pts ", DEN_oppmeanpoints)
print("DEN Opp. Avg. FG% ", DEN_oppfgp)
print("DEN Opp. Pts Deviation ", DEN_oppstdpts)
print("DEN Opp. FG% Deviation ", DEN_oppstdfgp)
```

```python
SAS_meanpoints = SAS1df.TeamPTS.mean()
SAS_meanfgp = SAS1df.TeamFGP.mean()
SAS_sdpts = SAS1df.TeamPTS.std()
SAS_sdfgp = SAS1df.TeamFGP.std()
SAS_oppmeanpoints = SAS1df.OpponentPTS.mean()
SAS_oppfgp = SAS1df.OpponentFGP.mean()
SAS_oppstdpts = SAS1df.OpponentPTS.std()
SAS_oppstdfgp = SAS1df.OpponentFGP.std()

print("SAS Avg. Pts ", SAS_meanpoints)
print("SAS Avg. FG% ", SAS_meanfgp)
print("SAS Pts Deviation ", SAS_sdpts)
print("SAS FG% Deviation ", SAS_sdfgp)
print("SAS Opp. Avg. Pts ", SAS_oppmeanpoints)
print("SAS Opp. Avg. FG% ", SAS_oppfgp)
print("SAS Opp. Pts Deviation ", SAS_oppstdpts)
print("SAS Opp. FG% Deviation ", SAS_oppstdfgp)
```

```python
def gameSim():
    DENScore = (rnd.gauss(DEN_score,DEN_sdpts)+ rnd.gauss(DEN_oppmeanpoints,DEN_oppstdpts))/2
    SASScore = (rnd.gauss(SAS_score,SAS_sdpts)+ rnd.gauss(SAS_oppmeanpoints,SAS_oppstdpts))/2
    if int(round(DENScore)) > int(round(SASScore)):
        return 1
    elif int(round(DENScore)) < int(round(SASScore)):
        return -1
    else: return 0
```

```python
def gamesSim(ns):
    gamesout = []
    team1win = 0
    team2win = 0
    for i in range(ns):
        gm = gameSim()
        gamesout.append(gm)
        if gm == 1:
            team1win +=1 
        elif gm == -1:
            team2win +=1
    print('DEN Win ', round((team1win/(team1win+team2win))*100,2),'%')
    print('SAS Win ', round((team2win/(team1win+team2win))*100,2),'%')
    return gamesout
```

```python
gamesSim(10000)
```
