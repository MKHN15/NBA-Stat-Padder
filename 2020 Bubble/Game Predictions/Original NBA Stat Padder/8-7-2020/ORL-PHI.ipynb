```python
import pandas as pd 
import random as rnd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
gdf = pd.read_csv('regssdatacsv.csv')
```

```python
gdf.columns
```

```python
ORLdf = gdf[gdf.Tm == 'ORL']
PHIdf = gdf[gdf.Tm == 'PHI']
```

```python
ORLdf.Date = ORLdf.Date.apply(lambda x: pd.to_datetime(x, format='%Y-%m-%d', errors='ignore'))
ORLdf = ORLdf[(ORLdf['Date'] > pd.to_datetime('20191022', format='%Y%m%d', errors='ignore')) & (ORLdf['Date'] <= pd.to_datetime('20200311', format='%Y%m%d', errors='ignore'))]

PHIdf.Date = PHIdf.Date.apply(lambda x: pd.to_datetime(x, format='%Y-%m-%d', errors='ignore'))
PHIdf = PHIdf[(PHIdf['Date'] > pd.to_datetime('20191022', format='%Y%m%d', errors='ignore'))& (PHIdf['Date'] <= pd.to_datetime('20200311', format='%Y%m%d', errors='ignore'))]
```

```python
ORLmeanpts = ORLdf.TeamPTS.mean()
PHImeanpts = PHIdf.TeamPTS.mean()
ORLfgp = ORLdf.TeamFGP.mean()
PHIfgp = PHIdf.TeamFGP.mean()
ORLsdpts = ORLdf.TeamPTS.std()
PHIsdpts = PHIdf.TeamPTS.std()

ORLmeaPHIp = ORLdf.OpponentPTS.mean()
PHImeaPHIp = PHIdf.OpponentPTS.mean()
ORLfgpopp = ORLdf.OpponentFGP.mean()
PHIfgpopp = PHIdf.OpponentFGP.mean()
ORLsdopp = ORLdf.OpponentPTS.std()
PHIsdopp = PHIdf.OpponentPTS.std()

print("ORL Mean ", ORLmeanpts)
print("ORL FG% ", ORLfgp)
print("ORL Points Standard Deviation ", ORLsdpts)
print("PHI Points Mean ", PHImeanpts)
print("PHI FG% ", PHIfgp)
print("PHI Points Standard Deviation ", PHIsdpts)

print("ORL OppPoints Mean ", ORLmeaPHIp)
print("ORL OppFG% Mean ", ORLfgpopp)
print("ORL OppPoints SD ", ORLsdopp)
print("PHI OppPoints Mean ", PHImeaPHIp)
print("PHI OppFG% Mean ", PHIfgpopp)
print("PHI OppPoints SD ", PHIsdopp)
```

```python
def gameSim():
    ORLScore = (rnd.gauss(ORLmeanpts,ORLsdpts)+ rnd.gauss(PHImeaPHIp,PHIsdopp))/2
    PHIScore = (rnd.gauss(PHImeanpts,PHIsdpts)+ rnd.gauss(PHImeaPHIp,PHIsdopp))/2
    if int(round(ORLScore)) > int(round(PHIScore)):
        return 1
    elif int(round(ORLScore)) < int(round(PHIScore)):
        return -1
    else: return 0
```

```python
def gAmesSim(ns):
    gamesout = []
    team1win = 0
    team2win = 0
    for i in range(ns):
        gm = gameSim()
        gamesout.append(gm)
        if gm == 1:
            team1win +=1 
        elif gm == -1:
            team2win +=1
    print('ORL Win ', team1win/(team1win+team2win),'%')
    print('PHI Win ', team2win/(team1win+team2win),'%')
    return gamesout
        
```

```python
gAmesSim(10000)
```
