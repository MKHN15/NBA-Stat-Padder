```python
import pandas as pd 
import random as rnd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
gdf = pd.read_csv('regssdatacsv.csv')
```

```python
gdf.columns
```

```python
PHIdf = gdf[gdf.Tm == 'PHI']
WASdf = gdf[gdf.Tm == 'WAS']
```

```python
PHIdf.Date = PHIdf.Date.apply(lambda x: pd.to_datetime(x, format='%Y-%m-%d', errors='ignore'))
PHIdf = PHIdf[(PHIdf['Date'] > pd.to_datetime('20191022', format='%Y%m%d', errors='ignore')) & (PHIdf['Date'] <= pd.to_datetime('20200311', format='%Y%m%d', errors='ignore'))]

WASdf.Date = WASdf.Date.apply(lambda x: pd.to_datetime(x, format='%Y-%m-%d', errors='ignore'))
WASdf = WASdf[(WASdf['Date'] > pd.to_datetime('20191022', format='%Y%m%d', errors='ignore'))& (WASdf['Date'] <= pd.to_datetime('20200311', format='%Y%m%d', errors='ignore'))]
```

```python
PHIdf.TeamFGP.hist()
WASdf.TeamFGP.hist()
```

```python
PHImeanpts = PHIdf.TeamPTS.mean()
WASmeanpts = WASdf.TeamPTS.mean()
PHIfgp = PHIdf.TeamFGP.mean()
WASfgp = WASdf.TeamFGP.mean()
PHIsdpts = PHIdf.TeamPTS.std()
WASsdpts = WASdf.TeamPTS.std()

PHImeaWASp = PHIdf.OpponentPTS.mean()
WASmeaWASp = WASdf.OpponentPTS.mean()
PHIfgpopp = PHIdf.OpponentFGP.mean()
WASfgpopp = WASdf.OpponentFGP.mean()
PHIsdopp = PHIdf.OpponentPTS.std()
WASsdopp = WASdf.OpponentPTS.std()

print("PHI Mean ", PHImeanpts)
print("PHI FG% ", PHIfgp)
print("PHI Points Standard Deviation ", PHIsdpts)
print("WAS Points Mean ", WASmeanpts)
print("WAS FG% ", WASfgp)
print("WAS Points Standard Deviation ", WASsdpts)

print("PHI OppPoints Mean ", PHImeaWASp)
print("PHI OppFG% Mean ", PHIfgpopp)
print("PHI OppPoints SD ", PHIsdopp)
print("WAS OppPoints Mean ", WASmeaWASp)
print("WAS OppFG% Mean ", WASfgpopp)
print("WAS OppPoints SD ", WASsdopp)
```

```python
def gameSim():
    PHIScore = (rnd.gauss(PHImeanpts,PHIsdpts)+ rnd.gauss(WASmeaWASp,WASsdopp))/2
    WASScore = (rnd.gauss(WASmeanpts,WASsdpts)+ rnd.gauss(WASmeaWASp,WASsdopp))/2
    if int(round(PHIScore)) > int(round(WASScore)):
        return 1
    elif int(round(PHIScore)) < int(round(WASScore)):
        return -1
    else: return 0
```

```python
def gAmesSim(ns):
    gamesout = []
    team1win = 0
    team2win = 0
    for i in range(ns):
        gm = gameSim()
        gamesout.append(gm)
        if gm == 1:
            team1win +=1 
        elif gm == -1:
            team2win +=1
    print('PHI Win ', team1win/(team1win+team2win),'%')
    print('WAS Win ', team2win/(team1win+team2win),'%')
    return gamesout
        
```

```python
gAmesSim(10000)
```
