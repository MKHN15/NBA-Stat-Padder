```python
import pandas as pd 
import random as rnd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
gdf = pd.read_csv('regssdatacsv.csv')
```

```python
gdf.columns
```

```python
DENdf = gdf[gdf.Tm == 'DEN']
SASdf = gdf[gdf.Tm == 'SAS']
```

```python
DENdf.Date = DENdf.Date.apply(lambda x: pd.to_datetime(x, format='%Y-%m-%d', errors='ignore'))
DENdf = DENdf[(DENdf['Date'] > pd.to_datetime('20191022', format='%Y%m%d', errors='ignore')) & (DENdf['Date'] <= pd.to_datetime('20200311', format='%Y%m%d', errors='ignore'))]

SASdf.Date = SASdf.Date.apply(lambda x: pd.to_datetime(x, format='%Y-%m-%d', errors='ignore'))
SASdf = SASdf[(SASdf['Date'] > pd.to_datetime('20191022', format='%Y%m%d', errors='ignore'))& (SASdf['Date'] <= pd.to_datetime('20200311', format='%Y%m%d', errors='ignore'))]
```

```python
DENmeanpts = DENdf.TeamPTS.mean()
SASmeanpts = SASdf.TeamPTS.mean()
DENfgp = DENdf.TeamFGP.mean()
SASfgp = SASdf.TeamFGP.mean()
DENsdpts = DENdf.TeamPTS.std()
SASsdpts = SASdf.TeamPTS.std()

DENmeaSASp = DENdf.OpponentPTS.mean()
SASmeaSASp = SASdf.OpponentPTS.mean()
DENfgpopp = DENdf.OpponentFGP.mean()
SASfgpopp = SASdf.OpponentFGP.mean()
DENsdopp = DENdf.OpponentPTS.std()
SASsdopp = SASdf.OpponentPTS.std()

print("DEN Mean ", DENmeanpts)
print("DEN FG% ", DENfgp)
print("DEN Points Standard Deviation ", DENsdpts)
print("SAS Points Mean ", SASmeanpts)
print("SAS FG% ", SASfgp)
print("SAS Points Standard Deviation ", SASsdpts)

print("DEN OppPoints Mean ", DENmeaSASp)
print("DEN OppFG% Mean ", DENfgpopp)
print("DEN OppPoints SD ", DENsdopp)
print("SAS OppPoints Mean ", SASmeaSASp)
print("SAS OppFG% Mean ", SASfgpopp)
print("SAS OppPoints SD ", SASsdopp)
```

```python
def gameSim():
    DENScore = (rnd.gauss(DENmeanpts,DENsdpts)+ rnd.gauss(SASmeaSASp,SASsdopp))/2
    SASScore = (rnd.gauss(SASmeanpts,SASsdpts)+ rnd.gauss(SASmeaSASp,SASsdopp))/2
    if int(round(DENScore)) > int(round(SASScore)):
        return 1
    elif int(round(DENScore)) < int(round(SASScore)):
        return -1
    else: return 0
```

```python
def gAmesSim(ns):
    gamesout = []
    team1win = 0
    team2win = 0
    for i in range(ns):
        gm = gameSim()
        gamesout.append(gm)
        if gm == 1:
            team1win +=1 
        elif gm == -1:
            team2win +=1
    print('DEN Win ', team1win/(team1win+team2win),'%')
    print('SAS Win ', team2win/(team1win+team2win),'%')
    return gamesout
        
```

```python
gAmesSim(10000)
```
