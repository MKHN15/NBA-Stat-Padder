```python
import pandas as pd 
import random as rnd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
gdf = pd.read_csv('regssdatacsv.csv')
```

```python
gdf.columns
```

```python
INDdf = gdf[gdf.Tm == 'IND']
HOUdf = gdf[gdf.Tm == 'HOU']
```

```python
INDdf.Date = INDdf.Date.apply(lambda x: pd.to_datetime(x, format='%Y-%m-%d', errors='ignore'))
INDdf = INDdf[(INDdf['Date'] > pd.to_datetime('20191022', format='%Y%m%d', errors='ignore')) & (INDdf['Date'] <= pd.to_datetime('20200311', format='%Y%m%d', errors='ignore'))]

HOUdf.Date = HOUdf.Date.apply(lambda x: pd.to_datetime(x, format='%Y-%m-%d', errors='ignore'))
HOUdf = HOUdf[(HOUdf['Date'] > pd.to_datetime('20191022', format='%Y%m%d', errors='ignore'))& (HOUdf['Date'] <= pd.to_datetime('20200311', format='%Y%m%d', errors='ignore'))]
```

```python
INDmeanpts = INDdf.TeamPTS.mean()
HOUmeanpts = HOUdf.TeamPTS.mean()
INDfgp = INDdf.TeamFGP.mean()
HOUfgp = HOUdf.TeamFGP.mean()
INDsdpts = INDdf.TeamPTS.std()
HOUsdpts = HOUdf.TeamPTS.std()

INDmeaHOUp = INDdf.OpponentPTS.mean()
HOUmeaHOUp = HOUdf.OpponentPTS.mean()
INDfgpopp = INDdf.OpponentFGP.mean()
HOUfgpopp = HOUdf.OpponentFGP.mean()
INDsdopp = INDdf.OpponentPTS.std()
HOUsdopp = HOUdf.OpponentPTS.std()

print("IND Mean ", INDmeanpts)
print("IND FG% ", INDfgp)
print("IND Points Standard Deviation ", INDsdpts)
print("HOU Points Mean ", HOUmeanpts)
print("HOU FG% ", HOUfgp)
print("HOU Points Standard Deviation ", HOUsdpts)

print("IND OppPoints Mean ", INDmeaHOUp)
print("IND OppFG% Mean ", INDfgpopp)
print("IND OppPoints SD ", INDsdopp)
print("HOU OppPoints Mean ", HOUmeaHOUp)
print("HOU OppFG% Mean ", HOUfgpopp)
print("HOU OppPoints SD ", HOUsdopp)
```

```python
def gameSim():
    INDScore = (rnd.gauss(INDmeanpts,INDsdpts)+ rnd.gauss(HOUmeaHOUp,HOUsdopp))/2
    HOUScore = (rnd.gauss(HOUmeanpts,HOUsdpts)+ rnd.gauss(HOUmeaHOUp,HOUsdopp))/2
    if int(round(INDScore)) > int(round(HOUScore)):
        return 1
    elif int(round(INDScore)) < int(round(HOUScore)):
        return -1
    else: return 0
```

```python
def gAmesSim(ns):
    gamesout = []
    team1win = 0
    team2win = 0
    for i in range(ns):
        gm = gameSim()
        gamesout.append(gm)
        if gm == 1:
            team1win +=1 
        elif gm == -1:
            team2win +=1
    print('IND Win ', team1win/(team1win+team2win),'%')
    print('HOU Win ', team2win/(team1win+team2win),'%')
    return gamesout
        
```

```python
gAmesSim(10000)
```
