```python
import pandas as pd 
import random as rnd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
gdf = pd.read_csv('regssdatacsv.csv')
```

```python
gdf.columns
```

```python
TORdf = gdf[gdf.Tm == 'TOR']
PHIdf = gdf[gdf.Tm == 'PHI']
```

```python
TORdf.Date = TORdf.Date.apply(lambda x: pd.to_datetime(x, format='%Y-%m-%d', errors='ignore'))
TORdf = TORdf[(TORdf['Date'] > pd.to_datetime('20191022', format='%Y%m%d', errors='ignore')) & (TORdf['Date'] <= pd.to_datetime('20200311', format='%Y%m%d', errors='ignore'))]

PHIdf.Date = PHIdf.Date.apply(lambda x: pd.to_datetime(x, format='%Y-%m-%d', errors='ignore'))
PHIdf = PHIdf[(PHIdf['Date'] > pd.to_datetime('20191022', format='%Y%m%d', errors='ignore'))& (PHIdf['Date'] <= pd.to_datetime('20200311', format='%Y%m%d', errors='ignore'))]
```

```python
TORmeanpts = TORdf.TeamPTS.mean()
PHImeanpts = PHIdf.TeamPTS.mean()
TORfgp = TORdf.TeamFGP.mean()
PHIfgp = PHIdf.TeamFGP.mean()
TORsdpts = TORdf.TeamPTS.std()
PHIsdpts = PHIdf.TeamPTS.std()

TORmeaPHIp = TORdf.OpponentPTS.mean()
PHImeaPHIp = PHIdf.OpponentPTS.mean()
TORfgpopp = TORdf.OpponentFGP.mean()
PHIfgpopp = PHIdf.OpponentFGP.mean()
TORsdopp = TORdf.OpponentPTS.std()
PHIsdopp = PHIdf.OpponentPTS.std()

print("TOR Mean ", TORmeanpts)
print("TOR FG% ", TORfgp)
print("TOR Points Standard Deviation ", TORsdpts)
print("PHI Points Mean ", PHImeanpts)
print("PHI FG% ", PHIfgp)
print("PHI Points Standard Deviation ", PHIsdpts)

print("TOR OppPoints Mean ", TORmeaPHIp)
print("TOR OppFG% Mean ", TORfgpopp)
print("TOR OppPoints SD ", TORsdopp)
print("PHI OppPoints Mean ", PHImeaPHIp)
print("PHI OppFG% Mean ", PHIfgpopp)
print("PHI OppPoints SD ", PHIsdopp)
```

```python
def gameSim():
    TORScore = (rnd.gauss(TORmeanpts,TORsdpts)+ rnd.gauss(PHImeaPHIp,PHIsdopp))/2
    PHIScore = (rnd.gauss(PHImeanpts,PHIsdpts)+ rnd.gauss(PHImeaPHIp,PHIsdopp))/2
    if int(round(TORScore)) > int(round(PHIScore)):
        return 1
    elif int(round(TORScore)) < int(round(PHIScore)):
        return -1
    else: return 0
```

```python
def gAmesSim(ns):
    gamesout = []
    team1win = 0
    team2win = 0
    for i in range(ns):
        gm = gameSim()
        gamesout.append(gm)
        if gm == 1:
            team1win +=1 
        elif gm == -1:
            team2win +=1
    print('TOR Win ', team1win/(team1win+team2win),'%')
    print('PHI Win ', team2win/(team1win+team2win),'%')
    return gamesout
        
```

```python
gAmesSim(10000)
```
