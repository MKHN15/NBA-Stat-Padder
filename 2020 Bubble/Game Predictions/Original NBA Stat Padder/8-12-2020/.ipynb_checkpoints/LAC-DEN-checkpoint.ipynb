```python
import pandas as pd 
import random as rnd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
gdf = pd.read_csv('regssdatacsv.csv')
```

```python
gdf.columns
```

```python
LACdf = gdf[gdf.Tm == 'LAC']
DENdf = gdf[gdf.Tm == 'DEN']
```

```python
LACdf.Date = LACdf.Date.apply(lambda x: pd.to_datetime(x, format='%Y-%m-%d', errors='ignore'))
LACdf = LACdf[(LACdf['Date'] > pd.to_datetime('20191022', format='%Y%m%d', errors='ignore')) & (LACdf['Date'] <= pd.to_datetime('20200311', format='%Y%m%d', errors='ignore'))]

DENdf.Date = DENdf.Date.apply(lambda x: pd.to_datetime(x, format='%Y-%m-%d', errors='ignore'))
DENdf = DENdf[(DENdf['Date'] > pd.to_datetime('20191022', format='%Y%m%d', errors='ignore'))& (DENdf['Date'] <= pd.to_datetime('20200311', format='%Y%m%d', errors='ignore'))]
```

```python
LACmeanpts = LACdf.TeamPTS.mean()
DENmeanpts = DENdf.TeamPTS.mean()
LACfgp = LACdf.TeamFGP.mean()
DENfgp = DENdf.TeamFGP.mean()
LACsdpts = LACdf.TeamPTS.std()
DENsdpts = DENdf.TeamPTS.std()

LACmeaDENp = LACdf.OpponentPTS.mean()
DENmeaDENp = DENdf.OpponentPTS.mean()
LACfgpopp = LACdf.OpponentFGP.mean()
DENfgpopp = DENdf.OpponentFGP.mean()
LACsdopp = LACdf.OpponentPTS.std()
DENsdopp = DENdf.OpponentPTS.std()

print("LAC Mean ", LACmeanpts)
print("LAC FG% ", LACfgp)
print("LAC Points Standard Deviation ", LACsdpts)
print("DEN Points Mean ", DENmeanpts)
print("DEN FG% ", DENfgp)
print("DEN Points Standard Deviation ", DENsdpts)

print("LAC OppPoints Mean ", LACmeaDENp)
print("LAC OppFG% Mean ", LACfgpopp)
print("LAC OppPoints SD ", LACsdopp)
print("DEN OppPoints Mean ", DENmeaDENp)
print("DEN OppFG% Mean ", DENfgpopp)
print("DEN OppPoints SD ", DENsdopp)
```

```python
def gameSim():
    LACScore = (rnd.gauss(LACmeanpts,LACsdpts)+ rnd.gauss(DENmeaDENp,DENsdopp))/2
    DENScore = (rnd.gauss(DENmeanpts,DENsdpts)+ rnd.gauss(DENmeaDENp,DENsdopp))/2
    if int(round(LACScore)) > int(round(DENScore)):
        return 1
    elif int(round(LACScore)) < int(round(DENScore)):
        return -1
    else: return 0
```

```python
def gAmesSim(ns):
    gamesout = []
    team1win = 0
    team2win = 0
    for i in range(ns):
        gm = gameSim()
        gamesout.append(gm)
        if gm == 1:
            team1win +=1 
        elif gm == -1:
            team2win +=1
    print('LAC Win ', team1win/(team1win+team2win),'%')
    print('DEN Win ', team2win/(team1win+team2win),'%')
    return gamesout
        
```

```python
gAmesSim(10000)
```
