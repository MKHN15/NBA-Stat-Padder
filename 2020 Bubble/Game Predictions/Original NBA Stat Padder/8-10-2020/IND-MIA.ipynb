```python
import pandas as pd 
import random as rnd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
gdf = pd.read_csv('regssdatacsv.csv')
```

```python
gdf.columns
```

```python
INDdf = gdf[gdf.Tm == 'IND']
MIAdf = gdf[gdf.Tm == 'MIA']
```

```python
INDdf.Date = INDdf.Date.apply(lambda x: pd.to_datetime(x, format='%Y-%m-%d', errors='ignore'))
INDdf = INDdf[(INDdf['Date'] > pd.to_datetime('20191022', format='%Y%m%d', errors='ignore')) & (INDdf['Date'] <= pd.to_datetime('20200311', format='%Y%m%d', errors='ignore'))]

MIAdf.Date = MIAdf.Date.apply(lambda x: pd.to_datetime(x, format='%Y-%m-%d', errors='ignore'))
MIAdf = MIAdf[(MIAdf['Date'] > pd.to_datetime('20191022', format='%Y%m%d', errors='ignore'))& (MIAdf['Date'] <= pd.to_datetime('20200311', format='%Y%m%d', errors='ignore'))]
```

```python
INDmeanpts = INDdf.TeamPTS.mean()
MIAmeanpts = MIAdf.TeamPTS.mean()
INDfgp = INDdf.TeamFGP.mean()
MIAfgp = MIAdf.TeamFGP.mean()
INDsdpts = INDdf.TeamPTS.std()
MIAsdpts = MIAdf.TeamPTS.std()

INDmeaMIAp = INDdf.OpponentPTS.mean()
MIAmeaMIAp = MIAdf.OpponentPTS.mean()
INDfgpopp = INDdf.OpponentFGP.mean()
MIAfgpopp = MIAdf.OpponentFGP.mean()
INDsdopp = INDdf.OpponentPTS.std()
MIAsdopp = MIAdf.OpponentPTS.std()

print("IND Mean ", INDmeanpts)
print("IND FG% ", INDfgp)
print("IND Points Standard Deviation ", INDsdpts)
print("MIA Points Mean ", MIAmeanpts)
print("MIA FG% ", MIAfgp)
print("MIA Points Standard Deviation ", MIAsdpts)

print("IND OppPoints Mean ", INDmeaMIAp)
print("IND OppFG% Mean ", INDfgpopp)
print("IND OppPoints SD ", INDsdopp)
print("MIA OppPoints Mean ", MIAmeaMIAp)
print("MIA OppFG% Mean ", MIAfgpopp)
print("MIA OppPoints SD ", MIAsdopp)
```

```python
def gameSim():
    INDScore = (rnd.gauss(INDmeanpts,INDsdpts)+ rnd.gauss(MIAmeaMIAp,MIAsdopp))/2
    MIAScore = (rnd.gauss(MIAmeanpts,MIAsdpts)+ rnd.gauss(MIAmeaMIAp,MIAsdopp))/2
    if int(round(INDScore)) > int(round(MIAScore)):
        return 1
    elif int(round(INDScore)) < int(round(MIAScore)):
        return -1
    else: return 0
```

```python
def gAmesSim(ns):
    gamesout = []
    team1win = 0
    team2win = 0
    for i in range(ns):
        gm = gameSim()
        gamesout.append(gm)
        if gm == 1:
            team1win +=1 
        elif gm == -1:
            team2win +=1
    print('IND Win ', team1win/(team1win+team2win),'%')
    print('MIA Win ', team2win/(team1win+team2win),'%')
    return gamesout
        
```

```python
gAmesSim(10000)
```
