```python
import pandas as pd 
import random as rnd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
gdf = pd.read_csv('regssdatacsv.csv')
```

```python
gdf.columns
```

```python
OKCdf = gdf[gdf.Tm == 'OKC']
PHOdf = gdf[gdf.Tm == 'PHO']
```

```python
OKCdf.Date = OKCdf.Date.apply(lambda x: pd.to_datetime(x, format='%Y-%m-%d', errors='ignore'))
OKCdf = OKCdf[(OKCdf['Date'] > pd.to_datetime('20191022', format='%Y%m%d', errors='ignore')) & (OKCdf['Date'] <= pd.to_datetime('20200311', format='%Y%m%d', errors='ignore'))]

PHOdf.Date = PHOdf.Date.apply(lambda x: pd.to_datetime(x, format='%Y-%m-%d', errors='ignore'))
PHOdf = PHOdf[(PHOdf['Date'] > pd.to_datetime('20191022', format='%Y%m%d', errors='ignore'))& (PHOdf['Date'] <= pd.to_datetime('20200311', format='%Y%m%d', errors='ignore'))]
```

```python
OKCmeanpts = OKCdf.TeamPTS.mean()
PHOmeanpts = PHOdf.TeamPTS.mean()
OKCfgp = OKCdf.TeamFGP.mean()
PHOfgp = PHOdf.TeamFGP.mean()
OKCsdpts = OKCdf.TeamPTS.std()
PHOsdpts = PHOdf.TeamPTS.std()

OKCmeaPHOp = OKCdf.OpponentPTS.mean()
PHOmeaPHOp = PHOdf.OpponentPTS.mean()
OKCfgpopp = OKCdf.OpponentFGP.mean()
PHOfgpopp = PHOdf.OpponentFGP.mean()
OKCsdopp = OKCdf.OpponentPTS.std()
PHOsdopp = PHOdf.OpponentPTS.std()

print("OKC Mean ", OKCmeanpts)
print("OKC FG% ", OKCfgp)
print("OKC Points Standard Deviation ", OKCsdpts)
print("PHO Points Mean ", PHOmeanpts)
print("PHO FG% ", PHOfgp)
print("PHO Points Standard Deviation ", PHOsdpts)

print("OKC OppPoints Mean ", OKCmeaPHOp)
print("OKC OppFG% Mean ", OKCfgpopp)
print("OKC OppPoints SD ", OKCsdopp)
print("PHO OppPoints Mean ", PHOmeaPHOp)
print("PHO OppFG% Mean ", PHOfgpopp)
print("PHO OppPoints SD ", PHOsdopp)
```

```python
def gameSim():
    OKCScore = (rnd.gauss(OKCmeanpts,OKCsdpts)+ rnd.gauss(PHOmeaPHOp,PHOsdopp))/2
    PHOScore = (rnd.gauss(PHOmeanpts,PHOsdpts)+ rnd.gauss(PHOmeaPHOp,PHOsdopp))/2
    if int(round(OKCScore)) > int(round(PHOScore)):
        return 1
    elif int(round(OKCScore)) < int(round(PHOScore)):
        return -1
    else: return 0
```

```python
def gAmesSim(ns):
    gamesout = []
    team1win = 0
    team2win = 0
    for i in range(ns):
        gm = gameSim()
        gamesout.append(gm)
        if gm == 1:
            team1win +=1 
        elif gm == -1:
            team2win +=1
    print('OKC Win ', team1win/(team1win+team2win),'%')
    print('PHO Win ', team2win/(team1win+team2win),'%')
    return gamesout
        
```

```python
gAmesSim(10000)
```
