```python
import pandas as pd 
import random as rnd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
gdf = pd.read_csv('regssdatacsv.csv')
```

```python
gdf.columns
```

```python
TORdf = gdf[gdf.Tm == 'TOR']
MILdf = gdf[gdf.Tm == 'MIL']
```

```python
TORdf.Date = TORdf.Date.apply(lambda x: pd.to_datetime(x, format='%Y-%m-%d', errors='ignore'))
TORdf = TORdf[(TORdf['Date'] > pd.to_datetime('20191022', format='%Y%m%d', errors='ignore')) & (TORdf['Date'] <= pd.to_datetime('20200311', format='%Y%m%d', errors='ignore'))]

MILdf.Date = MILdf.Date.apply(lambda x: pd.to_datetime(x, format='%Y-%m-%d', errors='ignore'))
MILdf = MILdf[(MILdf['Date'] > pd.to_datetime('20191022', format='%Y%m%d', errors='ignore'))& (MILdf['Date'] <= pd.to_datetime('20200311', format='%Y%m%d', errors='ignore'))]
```

```python
TORmeanpts = TORdf.TeamPTS.mean()
MILmeanpts = MILdf.TeamPTS.mean()
TORfgp = TORdf.TeamFGP.mean()
MILfgp = MILdf.TeamFGP.mean()
TORsdpts = TORdf.TeamPTS.std()
MILsdpts = MILdf.TeamPTS.std()

TORmeaMILp = TORdf.OpponentPTS.mean()
MILmeaMILp = MILdf.OpponentPTS.mean()
TORfgpopp = TORdf.OpponentFGP.mean()
MILfgpopp = MILdf.OpponentFGP.mean()
TORsdopp = TORdf.OpponentPTS.std()
MILsdopp = MILdf.OpponentPTS.std()

print("TOR Mean ", TORmeanpts)
print("TOR FG% ", TORfgp)
print("TOR Points Standard Deviation ", TORsdpts)
print("MIL Points Mean ", MILmeanpts)
print("MIL FG% ", MILfgp)
print("MIL Points Standard Deviation ", MILsdpts)

print("TOR OppPoints Mean ", TORmeaMILp)
print("TOR OppFG% Mean ", TORfgpopp)
print("TOR OppPoints SD ", TORsdopp)
print("MIL OppPoints Mean ", MILmeaMILp)
print("MIL OppFG% Mean ", MILfgpopp)
print("MIL OppPoints SD ", MILsdopp)
```

```python
def gameSim():
    TORScore = (rnd.gauss(TORmeanpts,TORsdpts)+ rnd.gauss(MILmeaMILp,MILsdopp))/2
    MILScore = (rnd.gauss(MILmeanpts,MILsdpts)+ rnd.gauss(MILmeaMILp,MILsdopp))/2
    if int(round(TORScore)) > int(round(MILScore)):
        return 1
    elif int(round(TORScore)) < int(round(MILScore)):
        return -1
    else: return 0
```

```python
def gAmesSim(ns):
    gamesout = []
    team1win = 0
    team2win = 0
    for i in range(ns):
        gm = gameSim()
        gamesout.append(gm)
        if gm == 1:
            team1win +=1 
        elif gm == -1:
            team2win +=1
    print('TOR Win ', team1win/(team1win+team2win),'%')
    print('MIL Win ', team2win/(team1win+team2win),'%')
    return gamesout
        
```

```python
gAmesSim(10000)
```
