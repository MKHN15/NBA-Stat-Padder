```python
import pandas as pd 
import random as rnd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
gdf = pd.read_csv('regssdatacsv.csv')
```

```python
gdf.columns
```

```python
WASdf = gdf[gdf.Tm == 'WAS']
BOSdf = gdf[gdf.Tm == 'BOS']
```

```python
WASdf.Date = WASdf.Date.apply(lambda x: pd.to_datetime(x, format='%Y-%m-%d', errors='ignore'))
WASdf = WASdf[(WASdf['Date'] > pd.to_datetime('20191022', format='%Y%m%d', errors='ignore')) & (WASdf['Date'] <= pd.to_datetime('20200311', format='%Y%m%d', errors='ignore'))]

BOSdf.Date = BOSdf.Date.apply(lambda x: pd.to_datetime(x, format='%Y-%m-%d', errors='ignore'))
BOSdf = BOSdf[(BOSdf['Date'] > pd.to_datetime('20191022', format='%Y%m%d', errors='ignore'))& (BOSdf['Date'] <= pd.to_datetime('20200311', format='%Y%m%d', errors='ignore'))]
```

```python
WASmeanpts = WASdf.TeamPTS.mean()
BOSmeanpts = BOSdf.TeamPTS.mean()
WASfgp = WASdf.TeamFGP.mean()
BOSfgp = BOSdf.TeamFGP.mean()
WASsdpts = WASdf.TeamPTS.std()
BOSsdpts = BOSdf.TeamPTS.std()

WASmeaBOSp = WASdf.OpponentPTS.mean()
BOSmeaBOSp = BOSdf.OpponentPTS.mean()
WASfgpopp = WASdf.OpponentFGP.mean()
BOSfgpopp = BOSdf.OpponentFGP.mean()
WASsdopp = WASdf.OpponentPTS.std()
BOSsdopp = BOSdf.OpponentPTS.std()

print("WAS Mean ", WASmeanpts)
print("WAS FG% ", WASfgp)
print("WAS Points Standard Deviation ", WASsdpts)
print("BOS Points Mean ", BOSmeanpts)
print("BOS FG% ", BOSfgp)
print("BOS Points Standard Deviation ", BOSsdpts)

print("WAS OppPoints Mean ", WASmeaBOSp)
print("WAS OppFG% Mean ", WASfgpopp)
print("WAS OppPoints SD ", WASsdopp)
print("BOS OppPoints Mean ", BOSmeaBOSp)
print("BOS OppFG% Mean ", BOSfgpopp)
print("BOS OppPoints SD ", BOSsdopp)
```

```python
def gameSim():
    WASScore = (rnd.gauss(WASmeanpts,WASsdpts)+ rnd.gauss(BOSmeaBOSp,BOSsdopp))/2
    BOSScore = (rnd.gauss(BOSmeanpts,BOSsdpts)+ rnd.gauss(BOSmeaBOSp,BOSsdopp))/2
    if int(round(WASScore)) > int(round(BOSScore)):
        return 1
    elif int(round(WASScore)) < int(round(BOSScore)):
        return -1
    else: return 0
```

```python
def gAmesSim(ns):
    gamesout = []
    team1win = 0
    team2win = 0
    for i in range(ns):
        gm = gameSim()
        gamesout.append(gm)
        if gm == 1:
            team1win +=1 
        elif gm == -1:
            team2win +=1
    print('WAS Win ', team1win/(team1win+team2win),'%')
    print('BOS Win ', team2win/(team1win+team2win),'%')
    return gamesout
        
```

```python
gAmesSim(10000)
```
