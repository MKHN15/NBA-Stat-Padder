```python
import pandas as pd 
import random as rnd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
gdf = pd.read_csv('regssdatacsv.csv')
```

```python
gdf.columns
```

```python
HOUdf = gdf[gdf.Tm == 'HOU']
DALdf = gdf[gdf.Tm == 'DAL']
```

```python
HOUdf.Date = HOUdf.Date.apply(lambda x: pd.to_datetime(x, format='%Y-%m-%d', errors='ignore'))
HOUdf = HOUdf[(HOUdf['Date'] > pd.to_datetime('20191024', format='%Y%m%d', errors='ignore')) & (HOUdf['Date'] <= pd.to_datetime('20200308', format='%Y%m%d', errors='ignore'))]

DALdf.Date = DALdf.Date.apply(lambda x: pd.to_datetime(x, format='%Y-%m-%d', errors='ignore'))
DALdf = DALdf[(DALdf['Date'] > pd.to_datetime('20191023', format='%Y%m%d', errors='ignore'))& (DALdf['Date'] <= pd.to_datetime('20200311', format='%Y%m%d', errors='ignore'))]
```

```python
HOUmeanpts = HOUdf.TeamPTS.mean()
DALmeanpts = DALdf.TeamPTS.mean()
HOUfgp = HOUdf.TeamFGP.mean()
DALfgp = DALdf.TeamFGP.mean()
HOUsdpts = HOUdf.TeamPTS.std()
DALsdpts = DALdf.TeamPTS.std()

HOUmeaDALp = HOUdf.OpponentPTS.mean()
DALmeaDALp = DALdf.OpponentPTS.mean()
HOUfgpopp = HOUdf.OpponentFGP.mean()
DALfgpopp = DALdf.OpponentFGP.mean()
HOUsdopp = HOUdf.OpponentPTS.std()
DALsdopp = DALdf.OpponentPTS.std()

print("HOU Mean ", HOUmeanpts)
print("HOU FG% ", HOUfgp)
print("HOU Points Standard Deviation ", HOUsdpts)
print("DAL Points Mean ", DALmeanpts)
print("DAL FG% ", DALfgp)
print("DAL Points Standard Deviation ", DALsdpts)

print("HOU OppPoints Mean ", HOUmeaDALp)
print("HOU OppFG% Mean ", HOUfgpopp)
print("HOU OppPoints SD ", HOUsdopp)
print("DAL OppPoints Mean ", DALmeaDALp)
print("DAL OppFG% Mean ", DALfgpopp)
print("DAL OppPoints SD ", DALsdopp)
```

```python
def gameSim():
    HOUScore = (rnd.gauss(HOUmeanpts,HOUsdpts)+ rnd.gauss(DALmeaDALp,DALsdopp))/2
    DALScore = (rnd.gauss(DALmeanpts,DALsdpts)+ rnd.gauss(DALmeaDALp,DALsdopp))/2
    if int(round(HOUScore)) > int(round(DALScore)):
        return 1
    elif int(round(HOUScore)) < int(round(DALScore)):
        return -1
    else: return 0
```

```python
def gAmesSim(ns):
    gamesout = []
    team1win = 0
    team2win = 0
    for i in range(ns):
        gm = gameSim()
        gamesout.append(gm)
        if gm == 1:
            team1win +=1 
        elif gm == -1:
            team2win +=1
    print('HOU Win ', team1win/(team1win+team2win),'%')
    print('DAL Win ', team2win/(team1win+team2win),'%')
    return gamesout
        
```

```python
gAmesSim(10000)
```
