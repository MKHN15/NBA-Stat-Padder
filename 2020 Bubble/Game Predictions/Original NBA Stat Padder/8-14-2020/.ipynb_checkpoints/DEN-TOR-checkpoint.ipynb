```python
import pandas as pd 
import random as rnd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
gdf = pd.read_csv('regssdatacsv.csv')
```

```python
gdf.columns
```

```python
DENdf = gdf[gdf.Tm == 'DEN']
TORdf = gdf[gdf.Tm == 'TOR']
```

```python
DENdf.Date = DENdf.Date.apply(lambda x: pd.to_datetime(x, format='%Y-%m-%d', errors='ignore'))
DENdf = DENdf[(DENdf['Date'] > pd.to_datetime('20191022', format='%Y%m%d', errors='ignore')) & (DENdf['Date'] <= pd.to_datetime('20200311', format='%Y%m%d', errors='ignore'))]

TORdf.Date = TORdf.Date.apply(lambda x: pd.to_datetime(x, format='%Y-%m-%d', errors='ignore'))
TORdf = TORdf[(TORdf['Date'] > pd.to_datetime('20191022', format='%Y%m%d', errors='ignore'))& (TORdf['Date'] <= pd.to_datetime('20200311', format='%Y%m%d', errors='ignore'))]
```

```python
DENmeanpts = DENdf.TeamPTS.mean()
TORmeanpts = TORdf.TeamPTS.mean()
DENfgp = DENdf.TeamFGP.mean()
TORfgp = TORdf.TeamFGP.mean()
DENsdpts = DENdf.TeamPTS.std()
TORsdpts = TORdf.TeamPTS.std()

DENmeaTORp = DENdf.OpponentPTS.mean()
TORmeaTORp = TORdf.OpponentPTS.mean()
DENfgpopp = DENdf.OpponentFGP.mean()
TORfgpopp = TORdf.OpponentFGP.mean()
DENsdopp = DENdf.OpponentPTS.std()
TORsdopp = TORdf.OpponentPTS.std()

print("DEN Mean ", DENmeanpts)
print("DEN FG% ", DENfgp)
print("DEN Points Standard Deviation ", DENsdpts)
print("TOR Points Mean ", TORmeanpts)
print("TOR FG% ", TORfgp)
print("TOR Points Standard Deviation ", TORsdpts)

print("DEN OppPoints Mean ", DENmeaTORp)
print("DEN OppFG% Mean ", DENfgpopp)
print("DEN OppPoints SD ", DENsdopp)
print("TOR OppPoints Mean ", TORmeaTORp)
print("TOR OppFG% Mean ", TORfgpopp)
print("TOR OppPoints SD ", TORsdopp)
```

```python
def gameSim():
    DENScore = (rnd.gauss(DENmeanpts,DENsdpts)+ rnd.gauss(TORmeaTORp,TORsdopp))/2
    TORScore = (rnd.gauss(TORmeanpts,TORsdpts)+ rnd.gauss(TORmeaTORp,TORsdopp))/2
    if int(round(DENScore)) > int(round(TORScore)):
        return 1
    elif int(round(DENScore)) < int(round(TORScore)):
        return -1
    else: return 0
```

```python
def gAmesSim(ns):
    gamesout = []
    team1win = 0
    team2win = 0
    for i in range(ns):
        gm = gameSim()
        gamesout.append(gm)
        if gm == 1:
            team1win +=1 
        elif gm == -1:
            team2win +=1
    print('DEN Win ', team1win/(team1win+team2win),'%')
    print('TOR Win ', team2win/(team1win+team2win),'%')
    return gamesout
        
```

```python
gAmesSim(10000)
```
