```python
import pandas as pd 
import random as rnd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
gdf = pd.read_csv('regssdatacsv.csv')
```

```python
gdf.columns
```

```python
LALdf = gdf[gdf.Tm == 'LAL']
HOUdf = gdf[gdf.Tm == 'HOU']
```

```python
LALdf.Date = LALdf.Date.apply(lambda x: pd.to_datetime(x, format='%Y-%m-%d', errors='ignore'))
LALdf = LALdf[(LALdf['Date'] > pd.to_datetime('20191022', format='%Y%m%d', errors='ignore')) & (LALdf['Date'] <= pd.to_datetime('20200311', format='%Y%m%d', errors='ignore'))]

HOUdf.Date = HOUdf.Date.apply(lambda x: pd.to_datetime(x, format='%Y-%m-%d', errors='ignore'))
HOUdf = HOUdf[(HOUdf['Date'] > pd.to_datetime('20191022', format='%Y%m%d', errors='ignore'))& (HOUdf['Date'] <= pd.to_datetime('20200311', format='%Y%m%d', errors='ignore'))]
```

```python
LALmeanpts = LALdf.TeamPTS.mean()
HOUmeanpts = HOUdf.TeamPTS.mean()
LALfgp = LALdf.TeamFGP.mean()
HOUfgp = HOUdf.TeamFGP.mean()
LALsdpts = LALdf.TeamPTS.std()
HOUsdpts = HOUdf.TeamPTS.std()

LALmeaHOUp = LALdf.OpponentPTS.mean()
HOUmeaHOUp = HOUdf.OpponentPTS.mean()
LALfgpopp = LALdf.OpponentFGP.mean()
HOUfgpopp = HOUdf.OpponentFGP.mean()
LALsdopp = LALdf.OpponentPTS.std()
HOUsdopp = HOUdf.OpponentPTS.std()

print("LAL Mean ", LALmeanpts)
print("LAL FG% ", LALfgp)
print("LAL Points Standard Deviation ", LALsdpts)
print("HOU Points Mean ", HOUmeanpts)
print("HOU FG% ", HOUfgp)
print("HOU Points Standard Deviation ", HOUsdpts)

print("LAL OppPoints Mean ", LALmeaHOUp)
print("LAL OppFG% Mean ", LALfgpopp)
print("LAL OppPoints SD ", LALsdopp)
print("HOU OppPoints Mean ", HOUmeaHOUp)
print("HOU OppFG% Mean ", HOUfgpopp)
print("HOU OppPoints SD ", HOUsdopp)
```

```python
def gameSim():
    LALScore = (rnd.gauss(LALmeanpts,LALsdpts)+ rnd.gauss(HOUmeaHOUp,HOUsdopp))/2
    HOUScore = (rnd.gauss(HOUmeanpts,HOUsdpts)+ rnd.gauss(HOUmeaHOUp,HOUsdopp))/2
    if int(round(LALScore)) > int(round(HOUScore)):
        return 1
    elif int(round(LALScore)) < int(round(HOUScore)):
        return -1
    else: return 0
```

```python
def gAmesSim(ns):
    gamesout = []
    team1win = 0
    team2win = 0
    for i in range(ns):
        gm = gameSim()
        gamesout.append(gm)
        if gm == 1:
            team1win +=1 
        elif gm == -1:
            team2win +=1
    print('LAL Win ', team1win/(team1win+team2win),'%')
    print('HOU Win ', team2win/(team1win+team2win),'%')
    return gamesout
        
```

```python
gAmesSim(10000)
```
