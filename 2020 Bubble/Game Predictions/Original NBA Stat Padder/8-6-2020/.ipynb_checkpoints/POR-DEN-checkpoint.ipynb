```python
import pandas as pd 
import random as rnd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
gdf = pd.read_csv('regssdatacsv.csv')
```

```python
gdf.columns
```

```python
PORdf = gdf[gdf.Tm == 'POR']
DENdf = gdf[gdf.Tm == 'DEN']
```

```python
PORdf.Date = PORdf.Date.apply(lambda x: pd.to_datetime(x, format='%Y-%m-%d', errors='ignore'))
PORdf = PORdf[(PORdf['Date'] > pd.to_datetime('20191022', format='%Y%m%d', errors='ignore')) & (PORdf['Date'] <= pd.to_datetime('20200311', format='%Y%m%d', errors='ignore'))]

DENdf.Date = DENdf.Date.apply(lambda x: pd.to_datetime(x, format='%Y-%m-%d', errors='ignore'))
DENdf = DENdf[(DENdf['Date'] > pd.to_datetime('20191022', format='%Y%m%d', errors='ignore'))& (DENdf['Date'] <= pd.to_datetime('20200311', format='%Y%m%d', errors='ignore'))]
```

```python
PORmeanpts = PORdf.TeamPTS.mean()
DENmeanpts = DENdf.TeamPTS.mean()
PORfgp = PORdf.TeamFGP.mean()
DENfgp = DENdf.TeamFGP.mean()
PORsdpts = PORdf.TeamPTS.std()
DENsdpts = DENdf.TeamPTS.std()

PORmeaDENp = PORdf.OpponentPTS.mean()
DENmeaDENp = DENdf.OpponentPTS.mean()
PORfgpopp = PORdf.OpponentFGP.mean()
DENfgpopp = DENdf.OpponentFGP.mean()
PORsdopp = PORdf.OpponentPTS.std()
DENsdopp = DENdf.OpponentPTS.std()

print("POR Mean ", PORmeanpts)
print("POR FG% ", PORfgp)
print("POR Points Standard Deviation ", PORsdpts)
print("DEN Points Mean ", DENmeanpts)
print("DEN FG% ", DENfgp)
print("DEN Points Standard Deviation ", DENsdpts)

print("POR OppPoints Mean ", PORmeaDENp)
print("POR OppFG% Mean ", PORfgpopp)
print("POR OppPoints SD ", PORsdopp)
print("DEN OppPoints Mean ", DENmeaDENp)
print("DEN OppFG% Mean ", DENfgpopp)
print("DEN OppPoints SD ", DENsdopp)
```

```python
def gameSim():
    PORScore = (rnd.gauss(PORmeanpts,PORsdpts)+ rnd.gauss(DENmeaDENp,DENsdopp))/2
    DENScore = (rnd.gauss(DENmeanpts,DENsdpts)+ rnd.gauss(DENmeaDENp,DENsdopp))/2
    if int(round(PORScore)) > int(round(DENScore)):
        return 1
    elif int(round(PORScore)) < int(round(DENScore)):
        return -1
    else: return 0
```

```python
def gAmesSim(ns):
    gamesout = []
    team1win = 0
    team2win = 0
    for i in range(ns):
        gm = gameSim()
        gamesout.append(gm)
        if gm == 1:
            team1win +=1 
        elif gm == -1:
            team2win +=1
    print('POR Win ', team1win/(team1win+team2win),'%')
    print('DEN Win ', team2win/(team1win+team2win),'%')
    return gamesout
        
```

```python
gAmesSim(10000)
```
