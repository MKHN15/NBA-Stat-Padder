```python
import pandas as pd 
import random as rnd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
gdf = pd.read_csv('regssdatacsv.csv')
```

```python
gdf.columns
```

```python
NOPdf = gdf[gdf.Tm == 'NOP']
SACdf = gdf[gdf.Tm == 'SAC']
```

```python
NOPdf.Date = NOPdf.Date.apply(lambda x: pd.to_datetime(x, format='%Y-%m-%d', errors='ignore'))
NOPdf = NOPdf[(NOPdf['Date'] > pd.to_datetime('20191022', format='%Y%m%d', errors='ignore')) & (NOPdf['Date'] <= pd.to_datetime('20200311', format='%Y%m%d', errors='ignore'))]

SACdf.Date = SACdf.Date.apply(lambda x: pd.to_datetime(x, format='%Y-%m-%d', errors='ignore'))
SACdf = SACdf[(SACdf['Date'] > pd.to_datetime('20191022', format='%Y%m%d', errors='ignore'))& (SACdf['Date'] <= pd.to_datetime('20200311', format='%Y%m%d', errors='ignore'))]
```

```python
NOPmeanpts = NOPdf.TeamPTS.mean()
SACmeanpts = SACdf.TeamPTS.mean()
NOPfgp = NOPdf.TeamFGP.mean()
SACfgp = SACdf.TeamFGP.mean()
NOPsdpts = NOPdf.TeamPTS.std()
SACsdpts = SACdf.TeamPTS.std()

NOPmeaSACp = NOPdf.OpponentPTS.mean()
SACmeaSACp = SACdf.OpponentPTS.mean()
NOPfgpopp = NOPdf.OpponentFGP.mean()
SACfgpopp = SACdf.OpponentFGP.mean()
NOPsdopp = NOPdf.OpponentPTS.std()
SACsdopp = SACdf.OpponentPTS.std()

print("NOP Mean ", NOPmeanpts)
print("NOP FG% ", NOPfgp)
print("NOP Points Standard Deviation ", NOPsdpts)
print("SAC Points Mean ", SACmeanpts)
print("SAC FG% ", SACfgp)
print("SAC Points Standard Deviation ", SACsdpts)

print("NOP OppPoints Mean ", NOPmeaSACp)
print("NOP OppFG% Mean ", NOPfgpopp)
print("NOP OppPoints SD ", NOPsdopp)
print("SAC OppPoints Mean ", SACmeaSACp)
print("SAC OppFG% Mean ", SACfgpopp)
print("SAC OppPoints SD ", SACsdopp)
```

```python
def gameSim():
    NOPScore = (rnd.gauss(NOPmeanpts,NOPsdpts)+ rnd.gauss(SACmeaSACp,SACsdopp))/2
    SACScore = (rnd.gauss(SACmeanpts,SACsdpts)+ rnd.gauss(SACmeaSACp,SACsdopp))/2
    if int(round(NOPScore)) > int(round(SACScore)):
        return 1
    elif int(round(NOPScore)) < int(round(SACScore)):
        return -1
    else: return 0
```

```python
def gAmesSim(ns):
    gamesout = []
    team1win = 0
    team2win = 0
    for i in range(ns):
        gm = gameSim()
        gamesout.append(gm)
        if gm == 1:
            team1win +=1 
        elif gm == -1:
            team2win +=1
    print('NOP Win ', team1win/(team1win+team2win),'%')
    print('SAC Win ', team2win/(team1win+team2win),'%')
    return gamesout
        
```

```python
gAmesSim(10000)
```
