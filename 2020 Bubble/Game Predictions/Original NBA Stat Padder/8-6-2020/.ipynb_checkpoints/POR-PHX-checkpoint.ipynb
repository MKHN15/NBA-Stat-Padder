```python
import pandas as pd 
import random as rnd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
gdf = pd.read_csv('regssdatacsv.csv')
```

```python
gdf.columns
```

```python
PORdf = gdf[gdf.Tm == 'POR']
PHOdf = gdf[gdf.Tm == 'PHO']
```

```python
PORdf.Date = PORdf.Date.apply(lambda x: pd.to_datetime(x, format='%Y-%m-%d', errors='ignore'))
PORdf = PORdf[(PORdf['Date'] > pd.to_datetime('20191022', format='%Y%m%d', errors='ignore')) & (PORdf['Date'] <= pd.to_datetime('20200311', format='%Y%m%d', errors='ignore'))]

PHOdf.Date = PHOdf.Date.apply(lambda x: pd.to_datetime(x, format='%Y-%m-%d', errors='ignore'))
PHOdf = PHOdf[(PHOdf['Date'] > pd.to_datetime('20191022', format='%Y%m%d', errors='ignore'))& (PHOdf['Date'] <= pd.to_datetime('20200311', format='%Y%m%d', errors='ignore'))]
```

```python
PORmeanpts = PORdf.TeamPTS.mean()
PHOmeanpts = PHOdf.TeamPTS.mean()
PORfgp = PORdf.TeamFGP.mean()
PHOfgp = PHOdf.TeamFGP.mean()
PORsdpts = PORdf.TeamPTS.std()
PHOsdpts = PHOdf.TeamPTS.std()

PORmeaPHOp = PORdf.OpponentPTS.mean()
PHOmeaPHOp = PHOdf.OpponentPTS.mean()
PORfgpopp = PORdf.OpponentFGP.mean()
PHOfgpopp = PHOdf.OpponentFGP.mean()
PORsdopp = PORdf.OpponentPTS.std()
PHOsdopp = PHOdf.OpponentPTS.std()

print("POR Mean ", PORmeanpts)
print("POR FG% ", PORfgp)
print("POR Points Standard Deviation ", PORsdpts)
print("PHO Points Mean ", PHOmeanpts)
print("PHO FG% ", PHOfgp)
print("PHO Points Standard Deviation ", PHOsdpts)

print("POR OppPoints Mean ", PORmeaPHOp)
print("POR OppFG% Mean ", PORfgpopp)
print("POR OppPoints SD ", PORsdopp)
print("PHO OppPoints Mean ", PHOmeaPHOp)
print("PHO OppFG% Mean ", PHOfgpopp)
print("PHO OppPoints SD ", PHOsdopp)
```

```python
def gameSim():
    PORScore = (rnd.gauss(PORmeanpts,PORsdpts)+ rnd.gauss(PHOmeaPHOp,PHOsdopp))/2
    PHOScore = (rnd.gauss(PHOmeanpts,PHOsdpts)+ rnd.gauss(PHOmeaPHOp,PHOsdopp))/2
    if int(round(PORScore)) > int(round(PHOScore)):
        return 1
    elif int(round(PORScore)) < int(round(PHOScore)):
        return -1
    else: return 0
```

```python
def gAmesSim(ns):
    gamesout = []
    team1win = 0
    team2win = 0
    for i in range(ns):
        gm = gameSim()
        gamesout.append(gm)
        if gm == 1:
            team1win +=1 
        elif gm == -1:
            team2win +=1
    print('POR Win ', team1win/(team1win+team2win),'%')
    print('PHO Win ', team2win/(team1win+team2win),'%')
    return gamesout
        
```

```python
gAmesSim(10000)
```
