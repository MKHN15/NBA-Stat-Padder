```python
import pandas as pd 
import random as rnd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
gdf = pd.read_csv('regssdatacsv.csv')
```

```python
gdf.columns
```

```python
MILdf = gdf[gdf.Tm == 'MIL']
DALdf = gdf[gdf.Tm == 'DAL']
```

```python
MILdf.Date = MILdf.Date.apply(lambda x: pd.to_datetime(x, format='%Y-%m-%d', errors='ignore'))
MILdf = MILdf[(MILdf['Date'] > pd.to_datetime('20191022', format='%Y%m%d', errors='ignore')) & (MILdf['Date'] <= pd.to_datetime('20200311', format='%Y%m%d', errors='ignore'))]

DALdf.Date = DALdf.Date.apply(lambda x: pd.to_datetime(x, format='%Y-%m-%d', errors='ignore'))
DALdf = DALdf[(DALdf['Date'] > pd.to_datetime('20191022', format='%Y%m%d', errors='ignore'))& (DALdf['Date'] <= pd.to_datetime('20200311', format='%Y%m%d', errors='ignore'))]
```

```python
MILmeanpts = MILdf.TeamPTS.mean()
DALmeanpts = DALdf.TeamPTS.mean()
MILfgp = MILdf.TeamFGP.mean()
DALfgp = DALdf.TeamFGP.mean()
MILsdpts = MILdf.TeamPTS.std()
DALsdpts = DALdf.TeamPTS.std()

MILmeaDALp = MILdf.OpponentPTS.mean()
DALmeaDALp = DALdf.OpponentPTS.mean()
MILfgpopp = MILdf.OpponentFGP.mean()
DALfgpopp = DALdf.OpponentFGP.mean()
MILsdopp = MILdf.OpponentPTS.std()
DALsdopp = DALdf.OpponentPTS.std()

print("MIL Mean ", MILmeanpts)
print("MIL FG% ", MILfgp)
print("MIL Points Standard Deviation ", MILsdpts)
print("DAL Points Mean ", DALmeanpts)
print("DAL FG% ", DALfgp)
print("DAL Points Standard Deviation ", DALsdpts)

print("MIL OppPoints Mean ", MILmeaDALp)
print("MIL OppFG% Mean ", MILfgpopp)
print("MIL OppPoints SD ", MILsdopp)
print("DAL OppPoints Mean ", DALmeaDALp)
print("DAL OppFG% Mean ", DALfgpopp)
print("DAL OppPoints SD ", DALsdopp)
```

```python
def gameSim():
    MILScore = (rnd.gauss(MILmeanpts,MILsdpts)+ rnd.gauss(DALmeaDALp,DALsdopp))/2
    DALScore = (rnd.gauss(DALmeanpts,DALsdpts)+ rnd.gauss(DALmeaDALp,DALsdopp))/2
    if int(round(MILScore)) > int(round(DALScore)):
        return 1
    elif int(round(MILScore)) < int(round(DALScore)):
        return -1
    else: return 0
```

```python
def gAmesSim(ns):
    gamesout = []
    team1win = 0
    team2win = 0
    for i in range(ns):
        gm = gameSim()
        gamesout.append(gm)
        if gm == 1:
            team1win +=1 
        elif gm == -1:
            team2win +=1
    print('MIL Win ', team1win/(team1win+team2win),'%')
    print('DAL Win ', team2win/(team1win+team2win),'%')
    return gamesout
        
```

```python
gAmesSim(10000)
```
