```python
import pandas as pd 
import random as rnd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
gdf = pd.read_csv('regssdatacsv.csv')
```

```python
gdf.columns
```

```python
HOUdf = gdf[gdf.Tm == 'HOU']
SACdf = gdf[gdf.Tm == 'SAC']
```

```python
HOUdf.Date = HOUdf.Date.apply(lambda x: pd.to_datetime(x, format='%Y-%m-%d', errors='ignore'))
HOUdf = HOUdf[(HOUdf['Date'] > pd.to_datetime('20191022', format='%Y%m%d', errors='ignore')) & (HOUdf['Date'] <= pd.to_datetime('20200311', format='%Y%m%d', errors='ignore'))]

SACdf.Date = SACdf.Date.apply(lambda x: pd.to_datetime(x, format='%Y-%m-%d', errors='ignore'))
SACdf = SACdf[(SACdf['Date'] > pd.to_datetime('20191022', format='%Y%m%d', errors='ignore'))& (SACdf['Date'] <= pd.to_datetime('20200311', format='%Y%m%d', errors='ignore'))]
```

```python
HOUmeanpts = HOUdf.TeamPTS.mean()
SACmeanpts = SACdf.TeamPTS.mean()
HOUfgp = HOUdf.TeamFGP.mean()
SACfgp = SACdf.TeamFGP.mean()
HOUsdpts = HOUdf.TeamPTS.std()
SACsdpts = SACdf.TeamPTS.std()

HOUmeaSACp = HOUdf.OpponentPTS.mean()
SACmeaSACp = SACdf.OpponentPTS.mean()
HOUfgpopp = HOUdf.OpponentFGP.mean()
SACfgpopp = SACdf.OpponentFGP.mean()
HOUsdopp = HOUdf.OpponentPTS.std()
SACsdopp = SACdf.OpponentPTS.std()

print("HOU Mean ", HOUmeanpts)
print("HOU FG% ", HOUfgp)
print("HOU Points Standard Deviation ", HOUsdpts)
print("SAC Points Mean ", SACmeanpts)
print("SAC FG% ", SACfgp)
print("SAC Points Standard Deviation ", SACsdpts)

print("HOU OppPoints Mean ", HOUmeaSACp)
print("HOU OppFG% Mean ", HOUfgpopp)
print("HOU OppPoints SD ", HOUsdopp)
print("SAC OppPoints Mean ", SACmeaSACp)
print("SAC OppFG% Mean ", SACfgpopp)
print("SAC OppPoints SD ", SACsdopp)
```

```python
def gameSim():
    HOUScore = (rnd.gauss(HOUmeanpts,HOUsdpts)+ rnd.gauss(SACmeaSACp,SACsdopp))/2
    SACScore = (rnd.gauss(SACmeanpts,SACsdpts)+ rnd.gauss(SACmeaSACp,SACsdopp))/2
    if int(round(HOUScore)) > int(round(SACScore)):
        return 1
    elif int(round(HOUScore)) < int(round(SACScore)):
        return -1
    else: return 0
```

```python
def gAmesSim(ns):
    gamesout = []
    team1win = 0
    team2win = 0
    for i in range(ns):
        gm = gameSim()
        gamesout.append(gm)
        if gm == 1:
            team1win +=1 
        elif gm == -1:
            team2win +=1
    print('HOU Win ', team1win/(team1win+team2win),'%')
    print('SAC Win ', team2win/(team1win+team2win),'%')
    return gamesout
        
```

```python
gAmesSim(10000)
```
