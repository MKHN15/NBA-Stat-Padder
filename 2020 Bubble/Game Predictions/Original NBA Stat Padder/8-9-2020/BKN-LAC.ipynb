```python
import pandas as pd 
import random as rnd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
gdf = pd.read_csv('regssdatacsv.csv')
```

```python
gdf.columns
```

```python
BRKdf = gdf[gdf.Tm == 'BRK']
LACdf = gdf[gdf.Tm == 'LAC']
```

```python
BRKdf.Date = BRKdf.Date.apply(lambda x: pd.to_datetime(x, format='%Y-%m-%d', errors='ignore'))
BRKdf = BRKdf[(BRKdf['Date'] > pd.to_datetime('20191022', format='%Y%m%d', errors='ignore')) & (BRKdf['Date'] <= pd.to_datetime('20200311', format='%Y%m%d', errors='ignore'))]

LACdf.Date = LACdf.Date.apply(lambda x: pd.to_datetime(x, format='%Y-%m-%d', errors='ignore'))
LACdf = LACdf[(LACdf['Date'] > pd.to_datetime('20191022', format='%Y%m%d', errors='ignore'))& (LACdf['Date'] <= pd.to_datetime('20200311', format='%Y%m%d', errors='ignore'))]
```

```python
BRKmeanpts = BRKdf.TeamPTS.mean()
LACmeanpts = LACdf.TeamPTS.mean()
BRKfgp = BRKdf.TeamFGP.mean()
LACfgp = LACdf.TeamFGP.mean()
BRKsdpts = BRKdf.TeamPTS.std()
LACsdpts = LACdf.TeamPTS.std()

BRKmeaLACp = BRKdf.OpponentPTS.mean()
LACmeaLACp = LACdf.OpponentPTS.mean()
BRKfgpopp = BRKdf.OpponentFGP.mean()
LACfgpopp = LACdf.OpponentFGP.mean()
BRKsdopp = BRKdf.OpponentPTS.std()
LACsdopp = LACdf.OpponentPTS.std()

print("BRK Mean ", BRKmeanpts)
print("BRK FG% ", BRKfgp)
print("BRK Points Standard Deviation ", BRKsdpts)
print("LAC Points Mean ", LACmeanpts)
print("LAC FG% ", LACfgp)
print("LAC Points Standard Deviation ", LACsdpts)

print("BRK OppPoints Mean ", BRKmeaLACp)
print("BRK OppFG% Mean ", BRKfgpopp)
print("BRK OppPoints SD ", BRKsdopp)
print("LAC OppPoints Mean ", LACmeaLACp)
print("LAC OppFG% Mean ", LACfgpopp)
print("LAC OppPoints SD ", LACsdopp)
```

```python
def gameSim():
    BRKScore = (rnd.gauss(BRKmeanpts,BRKsdpts)+ rnd.gauss(LACmeaLACp,LACsdopp))/2
    LACScore = (rnd.gauss(LACmeanpts,LACsdpts)+ rnd.gauss(LACmeaLACp,LACsdopp))/2
    if int(round(BRKScore)) > int(round(LACScore)):
        return 1
    elif int(round(BRKScore)) < int(round(LACScore)):
        return -1
    else: return 0
```

```python
def gAmesSim(ns):
    gamesout = []
    team1win = 0
    team2win = 0
    for i in range(ns):
        gm = gameSim()
        gamesout.append(gm)
        if gm == 1:
            team1win +=1 
        elif gm == -1:
            team2win +=1
    print('BRK Win ', team1win/(team1win+team2win),'%')
    print('LAC Win ', team2win/(team1win+team2win),'%')
    return gamesout
        
```

```python
gAmesSim(10000)
```
