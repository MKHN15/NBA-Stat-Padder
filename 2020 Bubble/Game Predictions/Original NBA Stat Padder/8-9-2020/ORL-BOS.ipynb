```python
import pandas as pd 
import random as rnd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
gdf = pd.read_csv('regssdatacsv.csv')
```

```python
gdf.columns
```

```python
ORLdf = gdf[gdf.Tm == 'ORL']
BOSdf = gdf[gdf.Tm == 'BOS']
```

```python
ORLdf.Date = ORLdf.Date.apply(lambda x: pd.to_datetime(x, format='%Y-%m-%d', errors='ignore'))
ORLdf = ORLdf[(ORLdf['Date'] > pd.to_datetime('20191022', format='%Y%m%d', errors='ignore')) & (ORLdf['Date'] <= pd.to_datetime('20200311', format='%Y%m%d', errors='ignore'))]

BOSdf.Date = BOSdf.Date.apply(lambda x: pd.to_datetime(x, format='%Y-%m-%d', errors='ignore'))
BOSdf = BOSdf[(BOSdf['Date'] > pd.to_datetime('20191022', format='%Y%m%d', errors='ignore'))& (BOSdf['Date'] <= pd.to_datetime('20200311', format='%Y%m%d', errors='ignore'))]
```

```python
ORLmeanpts = ORLdf.TeamPTS.mean()
BOSmeanpts = BOSdf.TeamPTS.mean()
ORLfgp = ORLdf.TeamFGP.mean()
BOSfgp = BOSdf.TeamFGP.mean()
ORLsdpts = ORLdf.TeamPTS.std()
BOSsdpts = BOSdf.TeamPTS.std()

ORLmeaBOSp = ORLdf.OpponentPTS.mean()
BOSmeaBOSp = BOSdf.OpponentPTS.mean()
ORLfgpopp = ORLdf.OpponentFGP.mean()
BOSfgpopp = BOSdf.OpponentFGP.mean()
ORLsdopp = ORLdf.OpponentPTS.std()
BOSsdopp = BOSdf.OpponentPTS.std()

print("ORL Mean ", ORLmeanpts)
print("ORL FG% ", ORLfgp)
print("ORL Points Standard Deviation ", ORLsdpts)
print("BOS Points Mean ", BOSmeanpts)
print("BOS FG% ", BOSfgp)
print("BOS Points Standard Deviation ", BOSsdpts)

print("ORL OppPoints Mean ", ORLmeaBOSp)
print("ORL OppFG% Mean ", ORLfgpopp)
print("ORL OppPoints SD ", ORLsdopp)
print("BOS OppPoints Mean ", BOSmeaBOSp)
print("BOS OppFG% Mean ", BOSfgpopp)
print("BOS OppPoints SD ", BOSsdopp)
```

```python
def gameSim():
    ORLScore = (rnd.gauss(ORLmeanpts,ORLsdpts)+ rnd.gauss(BOSmeaBOSp,BOSsdopp))/2
    BOSScore = (rnd.gauss(BOSmeanpts,BOSsdpts)+ rnd.gauss(BOSmeaBOSp,BOSsdopp))/2
    if int(round(ORLScore)) > int(round(BOSScore)):
        return 1
    elif int(round(ORLScore)) < int(round(BOSScore)):
        return -1
    else: return 0
```

```python
def gAmesSim(ns):
    gamesout = []
    team1win = 0
    team2win = 0
    for i in range(ns):
        gm = gameSim()
        gamesout.append(gm)
        if gm == 1:
            team1win +=1 
        elif gm == -1:
            team2win +=1
    print('ORL Win ', team1win/(team1win+team2win),'%')
    print('BOS Win ', team2win/(team1win+team2win),'%')
    return gamesout
        
```

```python
gAmesSim(10000)
```
