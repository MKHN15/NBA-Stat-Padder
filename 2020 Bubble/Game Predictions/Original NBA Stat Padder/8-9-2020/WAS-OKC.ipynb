```python
import pandas as pd 
import random as rnd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
gdf = pd.read_csv('regssdatacsv.csv')
```

```python
gdf.columns
```

```python
WASdf = gdf[gdf.Tm == 'WAS']
OKCdf = gdf[gdf.Tm == 'OKC']
```

```python
WASdf.Date = WASdf.Date.apply(lambda x: pd.to_datetime(x, format='%Y-%m-%d', errors='ignore'))
WASdf = WASdf[(WASdf['Date'] > pd.to_datetime('20191022', format='%Y%m%d', errors='ignore')) & (WASdf['Date'] <= pd.to_datetime('20200311', format='%Y%m%d', errors='ignore'))]

OKCdf.Date = OKCdf.Date.apply(lambda x: pd.to_datetime(x, format='%Y-%m-%d', errors='ignore'))
OKCdf = OKCdf[(OKCdf['Date'] > pd.to_datetime('20191022', format='%Y%m%d', errors='ignore'))& (OKCdf['Date'] <= pd.to_datetime('20200311', format='%Y%m%d', errors='ignore'))]
```

```python
WASmeanpts = WASdf.TeamPTS.mean()
OKCmeanpts = OKCdf.TeamPTS.mean()
WASfgp = WASdf.TeamFGP.mean()
OKCfgp = OKCdf.TeamFGP.mean()
WASsdpts = WASdf.TeamPTS.std()
OKCsdpts = OKCdf.TeamPTS.std()

WASmeaOKCp = WASdf.OpponentPTS.mean()
OKCmeaOKCp = OKCdf.OpponentPTS.mean()
WASfgpopp = WASdf.OpponentFGP.mean()
OKCfgpopp = OKCdf.OpponentFGP.mean()
WASsdopp = WASdf.OpponentPTS.std()
OKCsdopp = OKCdf.OpponentPTS.std()

print("WAS Mean ", WASmeanpts)
print("WAS FG% ", WASfgp)
print("WAS Points Standard Deviation ", WASsdpts)
print("OKC Points Mean ", OKCmeanpts)
print("OKC FG% ", OKCfgp)
print("OKC Points Standard Deviation ", OKCsdpts)

print("WAS OppPoints Mean ", WASmeaOKCp)
print("WAS OppFG% Mean ", WASfgpopp)
print("WAS OppPoints SD ", WASsdopp)
print("OKC OppPoints Mean ", OKCmeaOKCp)
print("OKC OppFG% Mean ", OKCfgpopp)
print("OKC OppPoints SD ", OKCsdopp)
```

```python
def gameSim():
    WASScore = (rnd.gauss(WASmeanpts,WASsdpts)+ rnd.gauss(OKCmeaOKCp,OKCsdopp))/2
    OKCScore = (rnd.gauss(OKCmeanpts,OKCsdpts)+ rnd.gauss(OKCmeaOKCp,OKCsdopp))/2
    if int(round(WASScore)) > int(round(OKCScore)):
        return 1
    elif int(round(WASScore)) < int(round(OKCScore)):
        return -1
    else: return 0
```

```python
def gAmesSim(ns):
    gamesout = []
    team1win = 0
    team2win = 0
    for i in range(ns):
        gm = gameSim()
        gamesout.append(gm)
        if gm == 1:
            team1win +=1 
        elif gm == -1:
            team2win +=1
    print('WAS Win ', team1win/(team1win+team2win),'%')
    print('OKC Win ', team2win/(team1win+team2win),'%')
    return gamesout
        
```

```python
gAmesSim(10000)
```
