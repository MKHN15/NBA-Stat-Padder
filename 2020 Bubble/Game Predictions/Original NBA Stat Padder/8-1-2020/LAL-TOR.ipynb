```python
import pandas as pd 
import random as rnd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
gdf = pd.read_csv('regssdatacsv.csv')
```

```python
gdf.columns
```

```python
LALdf = gdf[gdf.Tm == 'LAL']
TORdf = gdf[gdf.Tm == 'TOR']
```

```python
LALdf.Date = LALdf.Date.apply(lambda x: pd.to_datetime(x, format='%Y-%m-%d', errors='ignore'))
LALdf = LALdf[(LALdf['Date'] > pd.to_datetime('20191022', format='%Y%m%d', errors='ignore')) & (LALdf['Date'] <= pd.to_datetime('20200310', format='%Y%m%d', errors='ignore'))]

TORdf.Date = TORdf.Date.apply(lambda x: pd.to_datetime(x, format='%Y-%m-%d', errors='ignore'))
TORdf = TORdf[(TORdf['Date'] > pd.to_datetime('20191022', format='%Y%m%d', errors='ignore'))& (TORdf['Date'] <= pd.to_datetime('20200309', format='%Y%m%d', errors='ignore'))]
```

```python
LALmeanpts = LALdf.TeamPTS.mean()
TORmeanpts = TORdf.TeamPTS.mean()
LALfgp = LALdf.TeamFGP.mean()
TORfgp = TORdf.TeamFGP.mean()
LALsdpts = LALdf.TeamPTS.std()
TORsdpts = TORdf.TeamPTS.std()

LALmeaTORp = LALdf.OpponentPTS.mean()
TORmeaTORp = TORdf.OpponentPTS.mean()
LALfgpopp = LALdf.OpponentFGP.mean()
TORfgpopp = TORdf.OpponentFGP.mean()
LALsdopp = LALdf.OpponentPTS.std()
TORsdopp = TORdf.OpponentPTS.std()

print("LAL Mean ", LALmeanpts)
print("LAL FG% ", LALfgp)
print("LAL Points Standard Deviation ", LALsdpts)
print("TOR Points Mean ", TORmeanpts)
print("TOR FG% ", TORfgp)
print("TOR Points Standard Deviation ", TORsdpts)

print("LAL OppPoints Mean ", LALmeaTORp)
print("LAL OppFG% Mean ", LALfgpopp)
print("LAL OppPoints SD ", LALsdopp)
print("TOR OppPoints Mean ", TORmeaTORp)
print("TOR OppFG% Mean ", TORfgpopp)
print("TOR OppPoints SD ", TORsdopp)
```

```python
def gameSim():
    LALScore = (rnd.gauss(LALmeanpts,LALsdpts)+ rnd.gauss(TORmeaTORp,TORsdopp))/2
    TORScore = (rnd.gauss(TORmeanpts,TORsdpts)+ rnd.gauss(TORmeaTORp,TORsdopp))/2
    if int(round(LALScore)) > int(round(TORScore)):
        return 1
    elif int(round(LALScore)) < int(round(TORScore)):
        return -1
    else: return 0
```

```python
def gAmesSim(ns):
    gamesout = []
    team1win = 0
    team2win = 0
    for i in range(ns):
        gm = gameSim()
        gamesout.append(gm)
        if gm == 1:
            team1win +=1 
        elif gm == -1:
            team2win +=1
    print('LAL Win ', team1win/(team1win+team2win),'%')
    print('TOR Win ', team2win/(team1win+team2win),'%')
    return gamesout
        
```

```python
gAmesSim(10000)
```
