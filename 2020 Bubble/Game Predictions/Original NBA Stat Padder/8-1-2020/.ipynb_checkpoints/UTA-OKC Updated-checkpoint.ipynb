```python
import pandas as pd 
import random as rnd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
gdf = pd.read_csv('regssdatacsv.csv')
```

```python
gdf.columns
```

```python
UTAdf = gdf[gdf.Tm == 'UTA']
OKCdf = gdf[gdf.Tm == 'OKC']
```

```python
UTAdf.Date = UTAdf.Date.apply(lambda x: pd.to_datetime(x, format='%Y-%m-%d', errors='ignore'))
UTAdf = UTAdf[(UTAdf['Date'] > pd.to_datetime('20191023', format='%Y%m%d', errors='ignore')) & (UTAdf['Date'] <= pd.to_datetime('20200309', format='%Y%m%d', errors='ignore'))]

OKCdf.Date = OKCdf.Date.apply(lambda x: pd.to_datetime(x, format='%Y-%m-%d', errors='ignore'))
OKCdf = OKCdf[(OKCdf['Date'] > pd.to_datetime('20191023', format='%Y%m%d', errors='ignore'))& (OKCdf['Date'] <= pd.to_datetime('20200308', format='%Y%m%d', errors='ignore'))]
```

```python
UTAmeanpts = UTAdf.TeamPTS.mean()
OKCmeanpts = OKCdf.TeamPTS.mean()
UTAfgp = UTAdf.TeamFGP.mean()
OKCfgp = OKCdf.TeamFGP.mean()
UTAmeanftp = UTAdf.TeamFTP.mean()
OKCmeanftp = OKCdf.TeamFTP.mean()
UTAsdpts = UTAdf.TeamPTS.std()
OKCsdpts = OKCdf.TeamPTS.std()

UTAmeaOKCp = UTAdf.OpponentPTS.mean()
OKCmeaOKCp = OKCdf.OpponentPTS.mean()
UTAfgpopp = UTAdf.OpponentFGP.mean()
OKCfgpopp = OKCdf.OpponentFGP.mean()
UTAmeanftp = UTAdf.TeamFTP.mean()
OKCmeanftp = OKCdf.TeamFTP.mean()
UTAsdopp = UTAdf.OpponentPTS.std()
OKCsdopp = OKCdf.OpponentPTS.std()

print("UTA Mean ", UTAmeanpts)
print("UTA FG% ", UTAfgp)
print("UTA Points Standard Deviation ", UTAsdpts)
print("OKC Points Mean ", OKCmeanpts)
print("OKC FG% ", OKCfgp)
print("OKC Points Standard Deviation ", OKCsdpts)

print("UTA OppPoints Mean ", UTAmeaOKCp)
print("UTA OppFG% Mean ", UTAfgpopp)
print("UTA OppPoints SD ", UTAsdopp)
print("OKC OppPoints Mean ", OKCmeaOKCp)
print("OKC OppFG% Mean ", OKCfgpopp)
print("OKC OppPoints SD ", OKCsdopp)
```

```python
def gameSim():
    UTAScore = (rnd.gauss(UTAmeanpts,UTAsdpts)+ rnd.gauss(OKCmeaOKCp,OKCsdopp))/2
    OKCScore = (rnd.gauss(OKCmeanpts,OKCsdpts)+ rnd.gauss(OKCmeaOKCp,OKCsdopp))/2
    if int(round(UTAScore)) > int(round(OKCScore)):
        return 1
    elif int(round(UTAScore)) < int(round(OKCScore)):
        return -1
    else: return 0
```

```python
def gAmesSim(ns):
    gamesout = []
    team1win = 0
    team2win = 0
    for i in range(ns):
        gm = gameSim()
        gamesout.append(gm)
        if gm == 1:
            team1win +=1 
        elif gm == -1:
            team2win +=1
    print('UTA Win ', team1win/(team1win+team2win),'%')
    print('OKC Win ', team2win/(team1win+team2win),'%')
    return gamesout
        
```

```python
gAmesSim(10000)
```
