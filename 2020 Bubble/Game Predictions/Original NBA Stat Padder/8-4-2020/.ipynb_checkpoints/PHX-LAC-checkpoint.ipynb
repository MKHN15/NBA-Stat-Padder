```python
import pandas as pd 
import random as rnd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
gdf = pd.read_csv('regssdatacsv.csv')
```

```python
gdf.columns
```

```python
PHOdf = gdf[gdf.Tm == 'PHO']
LACdf = gdf[gdf.Tm == 'LAC']
```

```python
PHOdf.Date = PHOdf.Date.apply(lambda x: pd.to_datetime(x, format='%Y-%m-%d', errors='ignore'))
PHOdf = PHOdf[(PHOdf['Date'] > pd.to_datetime('20191022', format='%Y%m%d', errors='ignore')) & (PHOdf['Date'] <= pd.to_datetime('20200311', format='%Y%m%d', errors='ignore'))]

LACdf.Date = LACdf.Date.apply(lambda x: pd.to_datetime(x, format='%Y-%m-%d', errors='ignore'))
LACdf = LACdf[(LACdf['Date'] > pd.to_datetime('20191022', format='%Y%m%d', errors='ignore'))& (LACdf['Date'] <= pd.to_datetime('20200311', format='%Y%m%d', errors='ignore'))]
```

```python
PHOmeanpts = PHOdf.TeamPTS.mean()
LACmeanpts = LACdf.TeamPTS.mean()
PHOfgp = PHOdf.TeamFGP.mean()
LACfgp = LACdf.TeamFGP.mean()
PHOsdpts = PHOdf.TeamPTS.std()
LACsdpts = LACdf.TeamPTS.std()

PHOmeaLACp = PHOdf.OpponentPTS.mean()
LACmeaLACp = LACdf.OpponentPTS.mean()
PHOfgpopp = PHOdf.OpponentFGP.mean()
LACfgpopp = LACdf.OpponentFGP.mean()
PHOsdopp = PHOdf.OpponentPTS.std()
LACsdopp = LACdf.OpponentPTS.std()

print("PHO Mean ", PHOmeanpts)
print("PHO FG% ", PHOfgp)
print("PHO Points Standard Deviation ", PHOsdpts)
print("LAC Points Mean ", LACmeanpts)
print("LAC FG% ", LACfgp)
print("LAC Points Standard Deviation ", LACsdpts)

print("PHO OppPoints Mean ", PHOmeaLACp)
print("PHO OppFG% Mean ", PHOfgpopp)
print("PHO OppPoints SD ", PHOsdopp)
print("LAC OppPoints Mean ", LACmeaLACp)
print("LAC OppFG% Mean ", LACfgpopp)
print("LAC OppPoints SD ", LACsdopp)
```

```python
def gameSim():
    PHOScore = (rnd.gauss(PHOmeanpts,PHOsdpts)+ rnd.gauss(LACmeaLACp,LACsdopp))/2
    LACScore = (rnd.gauss(LACmeanpts,LACsdpts)+ rnd.gauss(LACmeaLACp,LACsdopp))/2
    if int(round(PHOScore)) > int(round(LACScore)):
        return 1
    elif int(round(PHOScore)) < int(round(LACScore)):
        return -1
    else: return 0
```

```python
def gAmesSim(ns):
    gamesout = []
    team1win = 0
    team2win = 0
    for i in range(ns):
        gm = gameSim()
        gamesout.append(gm)
        if gm == 1:
            team1win +=1 
        elif gm == -1:
            team2win +=1
    print('PHO Win ', team1win/(team1win+team2win),'%')
    print('LAC Win ', team2win/(team1win+team2win),'%')
    return gamesout
        
```

```python
gAmesSim(10000)
```
