```python
import pandas as pd 
import random as rnd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
gdf = pd.read_csv('regssdatacsv.csv')
```

```python
gdf.columns
```

```python
ORLdf = gdf[gdf.Tm == 'ORL']
INDdf = gdf[gdf.Tm == 'IND']
```

```python
ORLdf.Date = ORLdf.Date.apply(lambda x: pd.to_datetime(x, format='%Y-%m-%d', errors='ignore'))
ORLdf = ORLdf[(ORLdf['Date'] > pd.to_datetime('20191022', format='%Y%m%d', errors='ignore')) & (ORLdf['Date'] <= pd.to_datetime('20200311', format='%Y%m%d', errors='ignore'))]

INDdf.Date = INDdf.Date.apply(lambda x: pd.to_datetime(x, format='%Y-%m-%d', errors='ignore'))
INDdf = INDdf[(INDdf['Date'] > pd.to_datetime('20191022', format='%Y%m%d', errors='ignore'))& (INDdf['Date'] <= pd.to_datetime('20200311', format='%Y%m%d', errors='ignore'))]
```

```python
ORLmeanpts = ORLdf.TeamPTS.mean()
INDmeanpts = INDdf.TeamPTS.mean()
ORLfgp = ORLdf.TeamFGP.mean()
INDfgp = INDdf.TeamFGP.mean()
ORLsdpts = ORLdf.TeamPTS.std()
INDsdpts = INDdf.TeamPTS.std()

ORLmeaINDp = ORLdf.OpponentPTS.mean()
INDmeaINDp = INDdf.OpponentPTS.mean()
ORLfgpopp = ORLdf.OpponentFGP.mean()
INDfgpopp = INDdf.OpponentFGP.mean()
ORLsdopp = ORLdf.OpponentPTS.std()
INDsdopp = INDdf.OpponentPTS.std()

print("ORL Mean ", ORLmeanpts)
print("ORL FG% ", ORLfgp)
print("ORL Points Standard Deviation ", ORLsdpts)
print("IND Points Mean ", INDmeanpts)
print("IND FG% ", INDfgp)
print("IND Points Standard Deviation ", INDsdpts)

print("ORL OppPoints Mean ", ORLmeaINDp)
print("ORL OppFG% Mean ", ORLfgpopp)
print("ORL OppPoints SD ", ORLsdopp)
print("IND OppPoints Mean ", INDmeaINDp)
print("IND OppFG% Mean ", INDfgpopp)
print("IND OppPoints SD ", INDsdopp)
```

```python
def gameSim():
    ORLScore = (rnd.gauss(ORLmeanpts,ORLsdpts)+ rnd.gauss(INDmeaINDp,INDsdopp))/2
    INDScore = (rnd.gauss(INDmeanpts,INDsdpts)+ rnd.gauss(INDmeaINDp,INDsdopp))/2
    if int(round(ORLScore)) > int(round(INDScore)):
        return 1
    elif int(round(ORLScore)) < int(round(INDScore)):
        return -1
    else: return 0
```

```python
def gAmesSim(ns):
    gamesout = []
    team1win = 0
    team2win = 0
    for i in range(ns):
        gm = gameSim()
        gamesout.append(gm)
        if gm == 1:
            team1win +=1 
        elif gm == -1:
            team2win +=1
    print('ORL Win ', team1win/(team1win+team2win),'%')
    print('IND Win ', team2win/(team1win+team2win),'%')
    return gamesout
        
```

```python
gAmesSim(10000)
```
