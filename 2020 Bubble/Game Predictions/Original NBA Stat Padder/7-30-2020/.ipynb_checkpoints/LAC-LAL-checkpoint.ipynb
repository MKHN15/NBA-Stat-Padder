```python
import pandas as pd 
import random as rnd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
gdf = pd.read_csv('regssdatacsv.csv')
```

```python
gdf.columns
```

```python
LACdf = gdf[gdf.Tm == 'LAC']
LALdf = gdf[gdf.Tm == 'LAL']
```

```python
LACdf.Date = LACdf.Date.apply(lambda x: pd.to_datetime(x, format='%Y-%m-%d', errors='ignore'))
LACdf = LACdf[(LACdf['Date'] > pd.to_datetime('20191022', format='%Y%m%d', errors='ignore')) & (LACdf['Date'] <= pd.to_datetime('20200310', format='%Y%m%d', errors='ignore'))]

LALdf.Date = LALdf.Date.apply(lambda x: pd.to_datetime(x, format='%Y-%m-%d', errors='ignore'))
LALdf = LALdf[(LALdf['Date'] > pd.to_datetime('20191022', format='%Y%m%d', errors='ignore'))& (LALdf['Date'] <= pd.to_datetime('20200310', format='%Y%m%d', errors='ignore'))]
```

```python
LACmeanpts = LACdf.TeamPTS.mean()
LALmeanpts = LALdf.TeamPTS.mean()
LACfgp = LACdf.TeamFGP.mean()
LALfgp = LALdf.TeamFGP.mean()
LACsdpts = LACdf.TeamPTS.std()
LALsdpts = LALdf.TeamPTS.std()

LACmeaLALp = LACdf.OpponentPTS.mean()
LALmeaLALp = LALdf.OpponentPTS.mean()
LACfgpopp = LACdf.OpponentFGP.mean()
LALfgpopp = LALdf.OpponentFGP.mean()
LACsdopp = LACdf.OpponentPTS.std()
LALsdopp = LALdf.OpponentPTS.std()

print("LAC Mean ", LACmeanpts)
print("LAC FG% ", LACfgp)
print("LAC Points Standard Deviation ", LACsdpts)
print("LAL Points Mean ", LALmeanpts)
print("LAL FG% ", LALfgp)
print("LAL Points Standard Deviation ", LALsdpts)

print("LAC OppPoints Mean ", LACmeaLALp)
print("LAC OppFG% Mean ", LACfgpopp)
print("LAC OppPoints SD ", LACsdopp)
print("LAL OppPoints Mean ", LALmeaLALp)
print("LAL OppFG% Mean ", LALfgpopp)
print("LAL OppPoints SD ", LALsdopp)
```

```python
def gameSim():
    LACScore = (rnd.gauss(LACmeanpts,LACsdpts)+ rnd.gauss(LALmeaLALp,LALsdopp))/2
    LALScore = (rnd.gauss(LALmeanpts,LALsdpts)+ rnd.gauss(LALmeaLALp,LALsdopp))/2
    if int(round(LACScore)) > int(round(LALScore)):
        return 1
    elif int(round(LACScore)) < int(round(LALScore)):
        return -1
    else: return 0
```

```python
def gAmesSim(ns):
    gamesout = []
    team1win = 0
    team2win = 0
    for i in range(ns):
        gm = gameSim()
        gamesout.append(gm)
        if gm == 1:
            team1win +=1 
        elif gm == -1:
            team2win +=1
    print('LAC Win ', team1win/(team1win+team2win),'%')
    print('LAL Win ', team2win/(team1win+team2win),'%')
    return gamesout
        
```

```python
gAmesSim(10000)
```
