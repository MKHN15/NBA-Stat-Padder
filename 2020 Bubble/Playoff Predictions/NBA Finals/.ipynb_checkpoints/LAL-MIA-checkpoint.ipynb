```python
LAL_scoring = LAL_meanpts*(LAL_fgp*LAL_tpp*LAL_ftp+LAL_offrtg)
LAL_misc = LAL_scoring+LAL_dreb+LAL_ast+LAL_stl+LAL_blk+LAL_pfd+LAL_defrtg
LAL_top = LAL_misc*LAL_winpct*LAL_clutchwinpct
LAL_score = (LAL_top/(LAL_tov*10000))/(LAL_ornk+LAL_drnk)
LAL_score

print(LAL_score)
```

```python
MIA_scoring = MIA_meanpts*(MIA_fgp*MIA_tpp*MIA_ftp+MIA_offrtg)
MIA_misc = MIA_scoring+MIA_dreb+MIA_ast+MIA_stl+MIA_blk+MIA_pfd+MIA_defrtg
MIA_top = MIA_misc*MIA_winpct*MIA_clutchwinpct
MIA_score = (MIA_top/(MIA_tov*10000))/(MIA_ornk+MIA_drnk)
MIA_score

print(MIA_score)
```

```python
LAL_meanpoints = LAL1df.TeamPTS.mean()
LAL_meanfgp = LAL1df.TeamFGP.mean()
LAL_sdpts = LAL1df.TeamPTS.std()
LAL_sdfgp = LAL1df.TeamFGP.std()
LAL_oppmeanpoints = LAL1df.OpponentPTS.mean()
LAL_oppfgp = LAL1df.OpponentFGP.mean()
LAL_oppstdpts = LAL1df.OpponentPTS.std()
LAL_oppstdfgp = LAL1df.OpponentFGP.std()

print("LAL Avg. Pts ", LAL_meanpoints)
print("LAL Avg. FG% ", LAL_meanfgp)
print("LAL Pts Deviation ", LAL_sdpts)
print("LAL FG% Deviation ", LAL_sdfgp)
print("LAL Opp. Avg. Pts ", LAL_oppmeanpoints)
print("LAL Opp. Avg. FG% ", LAL_oppfgp)
print("LAL Opp. Pts Deviation ", LAL_oppstdpts)
print("LAL Opp. FG% Deviation ", LAL_oppstdfgp)
```

```python
MIA_meanpoints = MIA1df.TeamPTS.mean()
MIA_meanfgp = MIA1df.TeamFGP.mean()
MIA_sdpts = MIA1df.TeamPTS.std()
MIA_sdfgp = MIA1df.TeamFGP.std()
MIA_oppmeanpoints = MIA1df.OpponentPTS.mean()
MIA_oppfgp = MIA1df.OpponentFGP.mean()
MIA_oppstdpts = MIA1df.OpponentPTS.std()
MIA_oppstdfgp = MIA1df.OpponentFGP.std()

print("MIA Avg. Pts ", MIA_meanpoints)
print("MIA Avg. FG% ", MIA_meanfgp)
print("MIA Pts Deviation ", MIA_sdpts)
print("MIA FG% Deviation ", MIA_sdfgp)
print("MIA Opp. Avg. Pts ", MIA_oppmeanpoints)
print("MIA Opp. Avg. FG% ", MIA_oppfgp)
print("MIA Opp. Pts Deviation ", MIA_oppstdpts)
print("MIA Opp. FG% Deviation ", MIA_oppstdfgp)
```

```python
def gameSim():
    LALScore = (rnd.gauss(LAL_score,LAL_sdpts)+ rnd.gauss(LAL_oppmeanpoints,LAL_oppstdpts))/2
    MIAScore = (rnd.gauss(MIA_score,MIA_sdpts)+ rnd.gauss(MIA_oppmeanpoints,MIA_oppstdpts))/2
    if int(round(LALScore)) > int(round(MIAScore)):
        return 1
    elif int(round(LALScore)) < int(round(MIAScore)):
        return -1
    else: return 0
```

```python
def gamesSim(ns):
    gamesout = []
    team1win = 0
    team2win = 0
    for i in range(ns):
        gm = gameSim()
        gamesout.append(gm)
        if gm == 1:
            team1win +=1 
        elif gm == -1:
            team2win +=1
    print('LAL Win ', round((team1win/(team1win+team2win))*100,2),'%')
    print('MIA Win ', round((team2win/(team1win+team2win))*100,2),'%')
    return gamesout
```

```python
gamesSim(10000)
```
