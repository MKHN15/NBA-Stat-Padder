```python
MIL_scoring = MIL_meanpts*(MIL_fgp*MIL_tpp*MIL_ftp+MIL_offrtg)
MIL_misc = MIL_scoring+MIL_dreb+MIL_ast+MIL_stl+MIL_blk+MIL_pfd+MIL_defrtg
MIL_top = MIL_misc*MIL_winpct*MIL_clutchwinpct
MIL_score = (MIL_top/(MIL_tov*10000))/(MIL_ornk+MIL_drnk)
MIL_score

print(MIL_score)
```

```python
ORL_scoring = ORL_meanpts*(ORL_fgp*ORL_tpp*ORL_ftp+ORL_offrtg)
ORL_misc = ORL_scoring+ORL_dreb+ORL_ast+ORL_stl+ORL_blk+ORL_pfd+ORL_defrtg
ORL_top = ORL_misc*ORL_winpct*ORL_clutchwinpct
ORL_score = (ORL_top/(ORL_tov*10000))/(ORL_ornk+ORL_drnk)
ORL_score

print(ORL_score)
```

```python
MIL_meanpoints = MIL1df.TeamPTS.mean()
MIL_meanfgp = MIL1df.TeamFGP.mean()
MIL_sdpts = MIL1df.TeamPTS.std()
MIL_sdfgp = MIL1df.TeamFGP.std()
MIL_oppmeanpoints = MIL1df.OpponentPTS.mean()
MIL_oppfgp = MIL1df.OpponentFGP.mean()
MIL_oppstdpts = MIL1df.OpponentPTS.std()
MIL_oppstdfgp = MIL1df.OpponentFGP.std()

print("MIL Avg. Pts ", MIL_meanpoints)
print("MIL Avg. FG% ", MIL_meanfgp)
print("MIL Pts Deviation ", MIL_sdpts)
print("MIL FG% Deviation ", MIL_sdfgp)
print("MIL Opp. Avg. Pts ", MIL_oppmeanpoints)
print("MIL Opp. Avg. FG% ", MIL_oppfgp)
print("MIL Opp. Pts Deviation ", MIL_oppstdpts)
print("MIL Opp. FG% Deviation ", MIL_oppstdfgp)
```

```python
ORL_meanpoints = ORL1df.TeamPTS.mean()
ORL_meanfgp = ORL1df.TeamFGP.mean()
ORL_sdpts = ORL1df.TeamPTS.std()
ORL_sdfgp = ORL1df.TeamFGP.std()
ORL_oppmeanpoints = ORL1df.OpponentPTS.mean()
ORL_oppfgp = ORL1df.OpponentFGP.mean()
ORL_oppstdpts = ORL1df.OpponentPTS.std()
ORL_oppstdfgp = ORL1df.OpponentFGP.std()

print("ORL Avg. Pts ", ORL_meanpoints)
print("ORL Avg. FG% ", ORL_meanfgp)
print("ORL Pts Deviation ", ORL_sdpts)
print("ORL FG% Deviation ", ORL_sdfgp)
print("ORL Opp. Avg. Pts ", ORL_oppmeanpoints)
print("ORL Opp. Avg. FG% ", ORL_oppfgp)
print("ORL Opp. Pts Deviation ", ORL_oppstdpts)
print("ORL Opp. FG% Deviation ", ORL_oppstdfgp)
```

```python
def gameSim():
    MILScore = (rnd.gauss(MIL_score,MIL_sdpts)+ rnd.gauss(MIL_oppmeanpoints,MIL_oppstdpts))/2
    ORLScore = (rnd.gauss(ORL_score,ORL_sdpts)+ rnd.gauss(ORL_oppmeanpoints,ORL_oppstdpts))/2
    if int(round(MILScore)) > int(round(ORLScore)):
        return 1
    elif int(round(MILScore)) < int(round(ORLScore)):
        return -1
    else: return 0
```

```python
def gamesSim(ns):
    gamesout = []
    team1win = 0
    team2win = 0
    for i in range(ns):
        gm = gameSim()
        gamesout.append(gm)
        if gm == 1:
            team1win +=1 
        elif gm == -1:
            team2win +=1
    print('MIL Win ', round((team1win/(team1win+team2win))*100,2),'%')
    print('ORL Win ', round((team2win/(team1win+team2win))*100,2),'%')
    return gamesout
```

```python
gamesSim(10000)
```
