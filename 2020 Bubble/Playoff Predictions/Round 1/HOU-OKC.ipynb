```python
HOU_meanpoints = HOU1df.TeamPTS.mean()
HOU_meanfgp = HOU1df.TeamFGP.mean()
HOU_sdpts = HOU1df.TeamPTS.std()
HOU_sdfgp = HOU1df.TeamFGP.std()
HOU_oppmeanpoints = HOU1df.OpponentPTS.mean()
HOU_oppfgp = HOU1df.OpponentFGP.mean()
HOU_oppstdpts = HOU1df.OpponentPTS.std()
HOU_oppstdfgp = HOU1df.OpponentFGP.std()

print("HOU Avg. Pts ", HOU_meanpoints)
print("HOU Avg. FG% ", HOU_meanfgp)
print("HOU Pts Deviation ", HOU_sdpts)
print("HOU FG% Deviation ", HOU_sdfgp)
print("HOU Opp. Avg. Pts ", HOU_oppmeanpoints)
print("HOU Opp. Avg. FG% ", HOU_oppfgp)
print("HOU Opp. Pts Deviation ", HOU_oppstdpts)
print("HOU Opp. FG% Deviation ", HOU_oppstdfgp)
```

```python
OKC_meanpoints = OKC1df.TeamPTS.mean()
OKC_meanfgp = OKC1df.TeamFGP.mean()
OKC_sdpts = OKC1df.TeamPTS.std()
OKC_sdfgp = OKC1df.TeamFGP.std()
OKC_oppmeanpoints = OKC1df.OpponentPTS.mean()
OKC_oppfgp = OKC1df.OpponentFGP.mean()
OKC_oppstdpts = OKC1df.OpponentPTS.std()
OKC_oppstdfgp = OKC1df.OpponentFGP.std()

print("OKC Avg. Pts ", OKC_meanpoints)
print("OKC Avg. FG% ", OKC_meanfgp)
print("OKC Pts Deviation ", OKC_sdpts)
print("OKC FG% Deviation ", OKC_sdfgp)
print("OKC Opp. Avg. Pts ", OKC_oppmeanpoints)
print("OKC Opp. Avg. FG% ", OKC_oppfgp)
print("OKC Opp. Pts Deviation ", OKC_oppstdpts)
print("OKC Opp. FG% Deviation ", OKC_oppstdfgp)
```

```python
def gameSim():
    HOUScore = (rnd.gauss(HOU_score,HOU_sdpts)+ rnd.gauss(HOU_oppmeanpoints,HOU_oppstdpts))/2
    OKCScore = (rnd.gauss(OKC_score,OKC_sdpts)+ rnd.gauss(OKC_oppmeanpoints,OKC_oppstdpts))/2
    if int(round(HOUScore)) > int(round(OKCScore)):
        return 1
    elif int(round(HOUScore)) < int(round(OKCScore)):
        return -1
    else: return 0
```

```python
def gamesSim(ns):
    gamesout = []
    team1win = 0
    team2win = 0
    for i in range(ns):
        gm = gameSim()
        gamesout.append(gm)
        if gm == 1:
            team1win +=1 
        elif gm == -1:
            team2win +=1
    print('HOU Win ', round((team1win/(team1win+team2win))*100,2),'%')
    print('OKC Win ', round((team2win/(team1win+team2win))*100,2),'%')
    return gamesout
```
