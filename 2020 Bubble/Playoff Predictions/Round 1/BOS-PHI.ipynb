```python
BOS_meanpoints = BOS1df.TeamPTS.mean()
BOS_meanfgp = BOS1df.TeamFGP.mean()
BOS_sdpts = BOS1df.TeamPTS.std()
BOS_sdfgp = BOS1df.TeamFGP.std()
BOS_oppmeanpoints = BOS1df.OpponentPTS.mean()
BOS_oppfgp = BOS1df.OpponentFGP.mean()
BOS_oppstdpts = BOS1df.OpponentPTS.std()
BOS_oppstdfgp = BOS1df.OpponentFGP.std()

print("BOS Avg. Pts ", BOS_meanpoints)
print("BOS Avg. FG% ", BOS_meanfgp)
print("BOS Pts Deviation ", BOS_sdpts)
print("BOS FG% Deviation ", BOS_sdfgp)
print("BOS Opp. Avg. Pts ", BOS_oppmeanpoints)
print("BOS Opp. Avg. FG% ", BOS_oppfgp)
print("BOS Opp. Pts Deviation ", BOS_oppstdpts)
print("BOS Opp. FG% Deviation ", BOS_oppstdfgp)
```

```python
PHI_meanpoints = PHI1df.TeamPTS.mean()
PHI_meanfgp = PHI1df.TeamFGP.mean()
PHI_sdpts = PHI1df.TeamPTS.std()
PHI_sdfgp = PHI1df.TeamFGP.std()
PHI_oppmeanpoints = PHI1df.OpponentPTS.mean()
PHI_oppfgp = PHI1df.OpponentFGP.mean()
PHI_oppstdpts = PHI1df.OpponentPTS.std()
PHI_oppstdfgp = PHI1df.OpponentFGP.std()

print("PHI Avg. Pts ", PHI_meanpoints)
print("PHI Avg. FG% ", PHI_meanfgp)
print("PHI Pts Deviation ", PHI_sdpts)
print("PHI FG% Deviation ", PHI_sdfgp)
print("PHI Opp. Avg. Pts ", PHI_oppmeanpoints)
print("PHI Opp. Avg. FG% ", PHI_oppfgp)
print("PHI Opp. Pts Deviation ", PHI_oppstdpts)
print("PHI Opp. FG% Deviation ", PHI_oppstdfgp)
```

```python
def gameSim():
    BOSScore = (rnd.gauss(BOS_score,BOS_sdpts)+ rnd.gauss(BOS_oppmeanpoints,BOS_oppstdpts))/2
    PHIScore = (rnd.gauss(PHI_score,PHI_sdpts)+ rnd.gauss(PHI_oppmeanpoints,PHI_oppstdpts))/2
    if int(round(BOSScore)) > int(round(PHIScore)):
        return 1
    elif int(round(BOSScore)) < int(round(PHIScore)):
        return -1
    else: return 0
```

```python
def gamesSim(ns):
    gamesout = []
    team1win = 0
    team2win = 0
    for i in range(ns):
        gm = gameSim()
        gamesout.append(gm)
        if gm == 1:
            team1win +=1 
        elif gm == -1:
            team2win +=1
    print('BOS Win ', round((team1win/(team1win+team2win))*100,2),'%')
    print('PHI Win ', round((team2win/(team1win+team2win))*100,2),'%')
    return gamesout
```
