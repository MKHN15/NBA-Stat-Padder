```python
DENdf = gdf[gdf.TEAM == 'DEN']
UTAdf = gdf[gdf.TEAM == 'UTA']
```

```python
DEN_winpct = DENdf.WINP.mean()
DEN_meanpts = DENdf.PTS.mean()
DEN_fgp = DENdf.FGP.mean()
DEN_tpp = DENdf.TPP.mean()
DEN_ftp = DENdf.FTP.mean()
DEN_dreb = DENdf.DREB.mean()
DEN_ast = DENdf.AST.mean()
DEN_tov = DENdf.TOV.mean()
DEN_stl = DENdf.STL.mean()
DEN_blk = DENdf.BLK.mean()
DEN_pfd = DENdf.PFD.mean()
DEN_offrtg = DENdf.ORTG.mean()
DEN_defrtg = DENdf.DRTG.mean()
DEN_clutchwinpct = DENdf.CWINP.mean()
DEN_ornk = DENdf.ORNK.mean()
DEN_drnk = DENdf.DRNK.mean()


print("DEN Win% ", DEN_winpct)
print("DEN Avg. Points ", DEN_meanpts)
print("DEN FG% ", DEN_fgp)
print("DEN 3P% ", DEN_tpp)
print("DEN FT% ", DEN_ftp)
print("DEN Def. Reb. ", DEN_dreb)
print("DEN Avg. Assists ", DEN_ast)
print("DEN Avg. Turnovers ", DEN_tov)
print("DEN Avg. Steals ", DEN_stl)
print("DEN Avg. Blocks ", DEN_blk)
print("DEN Avg. Personal Fouls Drawn ", DEN_pfd)
print("DEN Off. Rtg ", DEN_offrtg)
print("DEN Def. Rtg ", DEN_defrtg)
print("DEN Clutch Win% ", DEN_clutchwinpct)
print("DEN Off. Rank ", DEN_ornk)
print("DEN Def. Rank", DEN_drnk)
```

```python
UTA_winpct = UTAdf.WINP.mean()
UTA_meanpts = UTAdf.PTS.mean()
UTA_fgp = UTAdf.FGP.mean()
UTA_tpp = UTAdf.TPP.mean()
UTA_ftp = UTAdf.FTP.mean()
UTA_dreb = UTAdf.DREB.mean()
UTA_ast = UTAdf.AST.mean()
UTA_tov = UTAdf.TOV.mean()
UTA_stl = UTAdf.STL.mean()
UTA_blk = UTAdf.BLK.mean()
UTA_pfd = UTAdf.PFD.mean()
UTA_offrtg = UTAdf.ORTG.mean()
UTA_defrtg = UTAdf.DRTG.mean()
UTA_clutchwinpct = UTAdf.CWINP.mean()
UTA_ornk = UTAdf.ORNK.mean()
UTA_drnk = UTAdf.DRNK.mean()

print("UTA Win% ", UTA_winpct)
print("UTA Avg. Points ", UTA_meanpts)
print("UTA FG% ", UTA_fgp)
print("UTA 3P% ", UTA_tpp)
print("UTA FT% ", UTA_ftp)
print("UTA Def. Reb. ", UTA_dreb)
print("UTA Avg. Assists ", UTA_ast)
print("UTA Avg. Turnovers ", UTA_tov)
print("UTA Avg. Steals ", UTA_stl)
print("UTA Avg. Blocks ", UTA_blk)
print("UTA Avg. Personal Fouls Drawn ", UTA_pfd)
print("UTA Off. Rtg ", UTA_offrtg)
print("UTA Def. Rtg ", UTA_defrtg)
print("UTA Clutch Win% ", UTA_clutchwinpct)
print("UTA Off. Rank ", UTA_ornk)
print("UTA Def. Rank", UTA_drnk)
```

```python
DEN_scoring = DEN_meanpts*(DEN_fgp*DEN_tpp*DEN_ftp+DEN_offrtg)
DEN_misc = DEN_scoring+DEN_dreb+DEN_ast+DEN_stl+DEN_blk+DEN_pfd+DEN_defrtg
DEN_top = DEN_misc*DEN_winpct*DEN_clutchwinpct
DEN_score = (DEN_top/(DEN_tov*10000))/(DEN_ornk+DEN_drnk)
DEN_score

print(DEN_score)
```

```python
UTA_scoring = UTA_meanpts*(UTA_fgp*UTA_tpp*UTA_ftp+UTA_offrtg)
UTA_misc = UTA_scoring+UTA_dreb+UTA_ast+UTA_stl+UTA_blk+UTA_pfd+UTA_defrtg
UTA_top = UTA_misc*UTA_winpct*UTA_clutchwinpct
UTA_score = (UTA_top/(UTA_tov*10000))/(UTA_ornk+UTA_drnk)
UTA_score

print(UTA_score)
```

```python
DEN_meanpoints = DEN1df.TeamPTS.mean()
DEN_meanfgp = DEN1df.TeamFGP.mean()
DEN_sdpts = DEN1df.TeamPTS.std()
DEN_sdfgp = DEN1df.TeamFGP.std()
DEN_oppmeanpoints = DEN1df.OpponentPTS.mean()
DEN_oppfgp = DEN1df.OpponentFGP.mean()
DEN_oppstdpts = DEN1df.OpponentPTS.std()
DEN_oppstdfgp = DEN1df.OpponentFGP.std()

print("DEN Avg. Pts ", DEN_meanpoints)
print("DEN Avg. FG% ", DEN_meanfgp)
print("DEN Pts Deviation ", DEN_sdpts)
print("DEN FG% Deviation ", DEN_sdfgp)
print("DEN Opp. Avg. Pts ", DEN_oppmeanpoints)
print("DEN Opp. Avg. FG% ", DEN_oppfgp)
print("DEN Opp. Pts Deviation ", DEN_oppstdpts)
print("DEN Opp. FG% Deviation ", DEN_oppstdfgp)
```

```python
UTA_meanpoints = UTA1df.TeamPTS.mean()
UTA_meanfgp = UTA1df.TeamFGP.mean()
UTA_sdpts = UTA1df.TeamPTS.std()
UTA_sdfgp = UTA1df.TeamFGP.std()
UTA_oppmeanpoints = UTA1df.OpponentPTS.mean()
UTA_oppfgp = UTA1df.OpponentFGP.mean()
UTA_oppstdpts = UTA1df.OpponentPTS.std()
UTA_oppstdfgp = UTA1df.OpponentFGP.std()

print("UTA Avg. Pts ", UTA_meanpoints)
print("UTA Avg. FG% ", UTA_meanfgp)
print("UTA Pts Deviation ", UTA_sdpts)
print("UTA FG% Deviation ", UTA_sdfgp)
print("UTA Opp. Avg. Pts ", UTA_oppmeanpoints)
print("UTA Opp. Avg. FG% ", UTA_oppfgp)
print("UTA Opp. Pts Deviation ", UTA_oppstdpts)
print("UTA Opp. FG% Deviation ", UTA_oppstdfgp)
```

```python
def gameSim():
    DENScore = (rnd.gauss(DEN_score,DEN_sdpts)+ rnd.gauss(DEN_oppmeanpoints,DEN_oppstdpts))/2
    UTAScore = (rnd.gauss(UTA_score,UTA_sdpts)+ rnd.gauss(UTA_oppmeanpoints,UTA_oppstdpts))/2
    if int(round(DENScore)) > int(round(UTAScore)):
        return 1
    elif int(round(DENScore)) < int(round(UTAScore)):
        return -1
    else: return 0
```

```python
def gamesSim(ns):
    gamesout = []
    team1win = 0
    team2win = 0
    for i in range(ns):
        gm = gameSim()
        gamesout.append(gm)
        if gm == 1:
            team1win +=1 
        elif gm == -1:
            team2win +=1
    print('DEN Win ', round((team1win/(team1win+team2win))*100,2),'%')
    print('UTA Win ', round((team2win/(team1win+team2win))*100,2),'%')
    return gamesout
```

```python
gamesSim(10000)
```
