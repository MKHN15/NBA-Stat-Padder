```python
LAC_scoring = LAC_meanpts*(LAC_fgp*LAC_tpp*LAC_ftp+LAC_offrtg)
LAC_misc = LAC_scoring+LAC_dreb+LAC_ast+LAC_stl+LAC_blk+LAC_pfd+LAC_defrtg
LAC_top = LAC_misc*LAC_winpct*LAC_clutchwinpct
LAC_score = (LAC_top/(LAC_tov*10000))/(LAC_ornk+LAC_drnk)
LAC_score

print(LAC_score)
```

```python
DAL_scoring = DAL_meanpts*(DAL_fgp*DAL_tpp*DAL_ftp+DAL_offrtg)
DAL_misc = DAL_scoring+DAL_dreb+DAL_ast+DAL_stl+DAL_blk+DAL_pfd+DAL_defrtg
DAL_top = DAL_misc*DAL_winpct*DAL_clutchwinpct
DAL_score = (DAL_top/(DAL_tov*10000))/(DAL_ornk+DAL_drnk)
DAL_score

print(DAL_score)
```

```python
LAC_meanpoints = LAC1df.TeamPTS.mean()
LAC_meanfgp = LAC1df.TeamFGP.mean()
LAC_sdpts = LAC1df.TeamPTS.std()
LAC_sdfgp = LAC1df.TeamFGP.std()
LAC_oppmeanpoints = LAC1df.OpponentPTS.mean()
LAC_oppfgp = LAC1df.OpponentFGP.mean()
LAC_oppstdpts = LAC1df.OpponentPTS.std()
LAC_oppstdfgp = LAC1df.OpponentFGP.std()

print("LAC Avg. Pts ", LAC_meanpoints)
print("LAC Avg. FG% ", LAC_meanfgp)
print("LAC Pts Deviation ", LAC_sdpts)
print("LAC FG% Deviation ", LAC_sdfgp)
print("LAC Opp. Avg. Pts ", LAC_oppmeanpoints)
print("LAC Opp. Avg. FG% ", LAC_oppfgp)
print("LAC Opp. Pts Deviation ", LAC_oppstdpts)
print("LAC Opp. FG% Deviation ", LAC_oppstdfgp)
```

```python
DAL_meanpoints = DAL1df.TeamPTS.mean()
DAL_meanfgp = DAL1df.TeamFGP.mean()
DAL_sdpts = DAL1df.TeamPTS.std()
DAL_sdfgp = DAL1df.TeamFGP.std()
DAL_oppmeanpoints = DAL1df.OpponentPTS.mean()
DAL_oppfgp = DAL1df.OpponentFGP.mean()
DAL_oppstdpts = DAL1df.OpponentPTS.std()
DAL_oppstdfgp = DAL1df.OpponentFGP.std()

print("DAL Avg. Pts ", DAL_meanpoints)
print("DAL Avg. FG% ", DAL_meanfgp)
print("DAL Pts Deviation ", DAL_sdpts)
print("DAL FG% Deviation ", DAL_sdfgp)
print("DAL Opp. Avg. Pts ", DAL_oppmeanpoints)
print("DAL Opp. Avg. FG% ", DAL_oppfgp)
print("DAL Opp. Pts Deviation ", DAL_oppstdpts)
print("DAL Opp. FG% Deviation ", DAL_oppstdfgp)
```

```python
def gameSim():
    LACScore = (rnd.gauss(LAC_score,LAC_sdpts)+ rnd.gauss(LAC_oppmeanpoints,LAC_oppstdpts))/2
    DALScore = (rnd.gauss(DAL_score,DAL_sdpts)+ rnd.gauss(DAL_oppmeanpoints,DAL_oppstdpts))/2
    if int(round(LACScore)) > int(round(DALScore)):
        return 1
    elif int(round(LACScore)) < int(round(DALScore)):
        return -1
    else: return 0
```

```python
def gamesSim(ns):
    gamesout = []
    team1win = 0
    team2win = 0
    for i in range(ns):
        gm = gameSim()
        gamesout.append(gm)
        if gm == 1:
            team1win +=1 
        elif gm == -1:
            team2win +=1
    print('LAC Win ', round((team1win/(team1win+team2win))*100,2),'%')
    print('DAL Win ', round((team2win/(team1win+team2win))*100,2),'%')
    return gamesout
```

```python
gamesSim(10000)
```
