```python
PHIdf = gdf[gdf.TEAM == 'Philadelphia 76ers']
HOUdf = gdf[gdf.TEAM == 'Houston Rockets']
```

```python
PHI_winpct = PHIdf.WINP.mean()
PHI_meanpts = PHIdf.PTS.mean()
PHI_fgp = PHIdf.FGP.mean()
PHI_tpp = PHIdf.TPP.mean()
PHI_ftp = PHIdf.FTP.mean()
PHI_dreb = PHIdf.DREB.mean()
PHI_ast = PHIdf.AST.mean()
PHI_tov = PHIdf.TOV.mean()
PHI_stl = PHIdf.STL.mean()
PHI_blk = PHIdf.BLK.mean()
PHI_pfd = PHIdf.PFD.mean()
PHI_offrtg = PHIdf.ORTG.mean()
PHI_defrtg = PHIdf.DRTG.mean()
PHI_clutchwinpct = PHIdf.CWINP.mean()

print("PHI Win% ", PHI_winpct)
print("PHI Avg. Points ", PHI_meanpts)
print("PHI FG% ", PHI_fgp)
print("PHI 3P% ", PHI_tpp)
print("PHI FT% ", PHI_ftp)
print("PHI Def. Reb. ", PHI_dreb)
print("PHI Avg. Assists ", PHI_ast)
print("PHI Avg. Turnovers ", PHI_tov)
print("PHI Avg. Steals ", PHI_stl)
print("PHI Avg. Blocks ", PHI_blk)
print("PHI Avg. Personal Fouls Drawn ", PHI_pfd)
print("PHI Off. Rtg ", PHI_offrtg)
print("PHI Def. Rtg ", PHI_defrtg)
print("PHI Clutch Win% ", PHI_clutchwinpct)
```

```python
HOU_winpct = HOUdf.WINP.mean()
HOU_meanpts = HOUdf.PTS.mean()
HOU_fgp = HOUdf.FGP.mean()
HOU_tpp = HOUdf.TPP.mean()
HOU_ftp = HOUdf.FTP.mean()
HOU_dreb = HOUdf.DREB.mean()
HOU_ast = HOUdf.AST.mean()
HOU_tov = HOUdf.TOV.mean()
HOU_stl = HOUdf.STL.mean()
HOU_blk = HOUdf.BLK.mean()
HOU_pfd = HOUdf.PFD.mean()
HOU_offrtg = PHIdf.ORTG.mean()
HOU_defrtg = PHIdf.DRTG.mean()
HOU_clutchwinpct = PHIdf.CWINP.mean()

print("HOU Win% ", HOU_winpct)
print("HOU Avg. Points ", HOU_meanpts)
print("HOU FG% ", HOU_fgp)
print("HOU 3P% ", HOU_tpp)
print("HOU FT% ", HOU_ftp)
print("HOU Def. Reb. ", HOU_dreb)
print("HOU Avg. Assists ", HOU_ast)
print("HOU Avg. Turnovers ", HOU_tov)
print("HOU Avg. Steals ", HOU_stl)
print("HOU Avg. Blocks ", HOU_blk)
print("HOU Avg. Personal Fouls Drawn ", HOU_pfd)
print("HOU Off. Rtg ", HOU_offrtg)
print("HOU Def. Rtg ", HOU_defrtg)
print("HOU Clutch Win% ", HOU_clutchwinpct)
```

```python
PHI_scoring = PHI_meanpts*(PHI_fgp*PHI_tpp*PHI_ftp+PHI_offrtg)
PHI_misc = PHI_scoring+PHI_dreb+PHI_ast+PHI_stl+PHI_blk+PHI_pfd+PHI_defrtg
PHI_top = PHI_misc*PHI_winpct*PHI_clutchwinpct
PHI_score = PHI_top/(PHI_tov*10000)
PHI_score

print(PHI_score)
```

```python
HOU_scoring = HOU_meanpts*(HOU_fgp*HOU_tpp*HOU_ftp+HOU_offrtg)
HOU_misc = HOU_scoring+HOU_dreb+HOU_ast+HOU_stl+HOU_blk+HOU_pfd+HOU_defrtg
HOU_top = HOU_misc*HOU_winpct*HOU_clutchwinpct
HOU_score = HOU_top/(HOU_tov*10000)
HOU_score

print(HOU_score)
```

```python
PHI_meanpoints = PHI1df.TeamPTS.mean()
PHI_meanfgp = PHI1df.TeamFGP.mean()
PHI_sdpts = PHI1df.TeamPTS.std()
PHI_sdfgp = PHI1df.TeamFGP.std()
PHI_oppmeanpoints = PHI1df.OpponentPTS.mean()
PHI_oppfgp = PHI1df.OpponentFGP.mean()
PHI_oppstdpts = PHI1df.OpponentPTS.std()
PHI_oppstdfgp = PHI1df.OpponentFGP.std()

print("PHI Avg. Pts ", PHI_meanpoints)
print("PHI Avg. FG% ", PHI_meanfgp)
print("PHI Pts Deviation ", PHI_sdpts)
print("PHI FG% Deviation ", PHI_sdfgp)
print("PHI Opp. Avg. Pts ", PHI_oppmeanpoints)
print("PHI Opp. Avg. FG% ", PHI_oppfgp)
print("PHI Opp. Pts Deviation ", PHI_oppstdpts)
print("PHI Opp. FG% Deviation ", PHI_oppstdfgp)
```

```python
HOU_meanpoints = HOU1df.TeamPTS.mean()
HOU_meanfgp = HOU1df.TeamFGP.mean()
HOU_sdpts = HOU1df.TeamPTS.std()
HOU_sdfgp = HOU1df.TeamFGP.std()
HOU_oppmeanpoints = HOU1df.OpponentPTS.mean()
HOU_oppfgp = HOU1df.OpponentFGP.mean()
HOU_oppstdpts = HOU1df.OpponentPTS.std()
HOU_oppstdfgp = HOU1df.OpponentFGP.std()

print("HOU Avg. Pts ", HOU_meanpoints)
print("HOU Avg. FG% ", HOU_meanfgp)
print("HOU Pts Deviation ", HOU_sdpts)
print("HOU FG% Deviation ", HOU_sdfgp)
print("HOU Opp. Avg. Pts ", HOU_oppmeanpoints)
print("HOU Opp. Avg. FG% ", HOU_oppfgp)
print("HOU Opp. Pts Deviation ", HOU_oppstdpts)
print("HOU Opp. FG% Deviation ", HOU_oppstdfgp)
```

```python
def gameSim():
    PHIScore = (rnd.gauss(PHI_score,PHI_sdpts)+ rnd.gauss(PHI_oppmeanpoints,PHI_oppstdpts))/2
    HOUScore = (rnd.gauss(HOU_score,HOU_sdpts)+ rnd.gauss(HOU_oppmeanpoints,HOU_oppstdpts))/2
    if int(round(PHIScore)) > int(round(HOUScore)):
        return 1
    elif int(round(PHIScore)) < int(round(HOUScore)):
        return -1
    else: return 0
```

```python
def gamesSim(ns):
    gamesout = []
    team1win = 0
    team2win = 0
    for i in range(ns):
        gm = gameSim()
        gamesout.append(gm)
        if gm == 1:
            team1win +=1 
        elif gm == -1:
            team2win +=1
    print('PHI Win ', round((team1win/(team1win+team2win))*100,2),'%')
    print('HOU Win ', round((team2win/(team1win+team2win))*100,2),'%')
    return gamesout
```

```python
gamesSim(10000)
```
