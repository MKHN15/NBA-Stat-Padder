```python
LAL_meanpoints = LAL1df.TeamPTS.mean()
LAL_meanfgp = LAL1df.TeamFGP.mean()
LAL_sdpts = LAL1df.TeamPTS.std()
LAL_sdfgp = LAL1df.TeamFGP.std()
LAL_oppmeanpoints = LAL1df.OpponentPTS.mean()
LAL_oppfgp = LAL1df.OpponentFGP.mean()
LAL_oppstdpts = LAL1df.OpponentPTS.std()
LAL_oppstdfgp = LAL1df.OpponentFGP.std()

print("LAL Avg. Pts ", LAL_meanpoints)
print("LAL Avg. FG% ", LAL_meanfgp)
print("LAL Pts Deviation ", LAL_sdpts)
print("LAL FG% Deviation ", LAL_sdfgp)
print("LAL Opp. Avg. Pts ", LAL_oppmeanpoints)
print("LAL Opp. Avg. FG% ", LAL_oppfgp)
print("LAL Opp. Pts Deviation ", LAL_oppstdpts)
print("LAL Opp. FG% Deviation ", LAL_oppstdfgp)
```

```python
POR_meanpoints = POR1df.TeamPTS.mean()
POR_meanfgp = POR1df.TeamFGP.mean()
POR_sdpts = POR1df.TeamPTS.std()
POR_sdfgp = POR1df.TeamFGP.std()
POR_oppmeanpoints = POR1df.OpponentPTS.mean()
POR_oppfgp = POR1df.OpponentFGP.mean()
POR_oppstdpts = POR1df.OpponentPTS.std()
POR_oppstdfgp = POR1df.OpponentFGP.std()

print("POR Avg. Pts ", POR_meanpoints)
print("POR Avg. FG% ", POR_meanfgp)
print("POR Pts Deviation ", POR_sdpts)
print("POR FG% Deviation ", POR_sdfgp)
print("POR Opp. Avg. Pts ", POR_oppmeanpoints)
print("POR Opp. Avg. FG% ", POR_oppfgp)
print("POR Opp. Pts Deviation ", POR_oppstdpts)
print("POR Opp. FG% Deviation ", POR_oppstdfgp)
```

```python
def gameSim():
    LALScore = (rnd.gauss(LAL_score,LAL_sdpts)+ rnd.gauss(LAL_oppmeanpoints,LAL_oppstdpts))/2
    PORScore = (rnd.gauss(POR_score,POR_sdpts)+ rnd.gauss(POR_oppmeanpoints,POR_oppstdpts))/2
    if int(round(LALScore)) > int(round(PORScore)):
        return 1
    elif int(round(LALScore)) < int(round(PORScore)):
        return -1
    else: return 0
```

```python
def gamesSim(ns):
    gamesout = []
    team1win = 0
    team2win = 0
    for i in range(ns):
        gm = gameSim()
        gamesout.append(gm)
        if gm == 1:
            team1win +=1 
        elif gm == -1:
            team2win +=1
    print('LAL Win ', round((team1win/(team1win+team2win))*100,2),'%')
    print('POR Win ', round((team2win/(team1win+team2win))*100,2),'%')
    return gamesout
```

```python
gamesSim(10000)
```
