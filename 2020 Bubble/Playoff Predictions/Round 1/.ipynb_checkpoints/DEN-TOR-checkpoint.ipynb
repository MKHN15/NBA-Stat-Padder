```python
import pandas as pd 
import random as rnd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
DENdf = gdf[gdf.TEAM == 'Denver Nuggets']
TORdf = gdf[gdf.TEAM == 'Toronto Raptors']
```

```python
DEN_winpct = DENdf.WINP.mean()
DEN_meanpts = DENdf.PTS.mean()
DEN_fgp = DENdf.FGP.mean()
DEN_tpp = DENdf.TPP.mean()
DEN_ftp = DENdf.FTP.mean()
DEN_dreb = DENdf.DREB.mean()
DEN_ast = DENdf.AST.mean()
DEN_tov = DENdf.TOV.mean()
DEN_stl = DENdf.STL.mean()
DEN_blk = DENdf.BLK.mean()
DEN_pfd = DENdf.PFD.mean()
DEN_offrtg = DENdf.ORTG.mean()
DEN_defrtg = DENdf.DRTG.mean()
DEN_clutchwinpct = DENdf.CWINP.mean()

print("DEN Win% ", DEN_winpct)
print("DEN Avg. Points ", DEN_meanpts)
print("DEN FG% ", DEN_fgp)
print("DEN 3P% ", DEN_tpp)
print("DEN FT% ", DEN_ftp)
print("DEN Def. Reb. ", DEN_dreb)
print("DEN Avg. Assists ", DEN_ast)
print("DEN Avg. Turnovers ", DEN_tov)
print("DEN Avg. Steals ", DEN_stl)
print("DEN Avg. Blocks ", DEN_blk)
print("DEN Avg. Personal Fouls Drawn ", DEN_pfd)
print("DEN Off. Rtg ", DEN_offrtg)
print("DEN Def. Rtg ", DEN_defrtg)
print("DEN Clutch Win% ", DEN_clutchwinpct)
```

```python
TOR_winpct = TORdf.WINP.mean()
TOR_meanpts = TORdf.PTS.mean()
TOR_fgp = TORdf.FGP.mean()
TOR_tpp = TORdf.TPP.mean()
TOR_ftp = TORdf.FTP.mean()
TOR_dreb = TORdf.DREB.mean()
TOR_ast = TORdf.AST.mean()
TOR_tov = TORdf.TOV.mean()
TOR_stl = TORdf.STL.mean()
TOR_blk = TORdf.BLK.mean()
TOR_pfd = TORdf.PFD.mean()
TOR_offrtg = DENdf.ORTG.mean()
TOR_defrtg = DENdf.DRTG.mean()
TOR_clutchwinpct = DENdf.CWINP.mean()

print("TOR Win% ", TOR_winpct)
print("TOR Avg. Points ", TOR_meanpts)
print("TOR FG% ", TOR_fgp)
print("TOR 3P% ", TOR_tpp)
print("TOR FT% ", TOR_ftp)
print("TOR Def. Reb. ", TOR_dreb)
print("TOR Avg. Assists ", TOR_ast)
print("TOR Avg. Turnovers ", TOR_tov)
print("TOR Avg. Steals ", TOR_stl)
print("TOR Avg. Blocks ", TOR_blk)
print("TOR Avg. Personal Fouls Drawn ", TOR_pfd)
print("TOR Off. Rtg ", TOR_offrtg)
print("TOR Def. Rtg ", TOR_defrtg)
print("TOR Clutch Win% ", TOR_clutchwinpct)
```

```python
DEN_scoring = DEN_meanpts*(DEN_fgp*DEN_tpp*DEN_ftp+DEN_offrtg)
DEN_misc = DEN_scoring+DEN_dreb+DEN_ast+DEN_stl+DEN_blk+DEN_pfd+DEN_defrtg
DEN_top = DEN_misc*DEN_winpct*DEN_clutchwinpct
DEN_score = DEN_top/(DEN_tov*10000)
DEN_score

print(DEN_score)
```

```python
TOR_scoring = TOR_meanpts*(TOR_fgp*TOR_tpp*TOR_ftp+TOR_offrtg)
TOR_misc = TOR_scoring+TOR_dreb+TOR_ast+TOR_stl+TOR_blk+TOR_pfd+TOR_defrtg
TOR_top = TOR_misc*TOR_winpct*TOR_clutchwinpct
TOR_score = TOR_top/(TOR_tov*10000)
TOR_score

print(TOR_score)
```

```python
DEN_meanpoints = DEN1df.TeamPTS.mean()
DEN_meanfgp = DEN1df.TeamFGP.mean()
DEN_sdpts = DEN1df.TeamPTS.std()
DEN_sdfgp = DEN1df.TeamFGP.std()
DEN_oppmeanpoints = DEN1df.OpponentPTS.mean()
DEN_oppfgp = DEN1df.OpponentFGP.mean()
DEN_oppstdpts = DEN1df.OpponentPTS.std()
DEN_oppstdfgp = DEN1df.OpponentFGP.std()

print("DEN Avg. Pts ", DEN_meanpoints)
print("DEN Avg. FG% ", DEN_meanfgp)
print("DEN Pts Deviation ", DEN_sdpts)
print("DEN FG% Deviation ", DEN_sdfgp)
print("DEN Opp. Avg. Pts ", DEN_oppmeanpoints)
print("DEN Opp. Avg. FG% ", DEN_oppfgp)
print("DEN Opp. Pts Deviation ", DEN_oppstdpts)
print("DEN Opp. FG% Deviation ", DEN_oppstdfgp)
```

```python
TOR_meanpoints = TOR1df.TeamPTS.mean()
TOR_meanfgp = TOR1df.TeamFGP.mean()
TOR_sdpts = TOR1df.TeamPTS.std()
TOR_sdfgp = TOR1df.TeamFGP.std()
TOR_oppmeanpoints = TOR1df.OpponentPTS.mean()
TOR_oppfgp = TOR1df.OpponentFGP.mean()
TOR_oppstdpts = TOR1df.OpponentPTS.std()
TOR_oppstdfgp = TOR1df.OpponentFGP.std()

print("TOR Avg. Pts ", TOR_meanpoints)
print("TOR Avg. FG% ", TOR_meanfgp)
print("TOR Pts Deviation ", TOR_sdpts)
print("TOR FG% Deviation ", TOR_sdfgp)
print("TOR Opp. Avg. Pts ", TOR_oppmeanpoints)
print("TOR Opp. Avg. FG% ", TOR_oppfgp)
print("TOR Opp. Pts Deviation ", TOR_oppstdpts)
print("TOR Opp. FG% Deviation ", TOR_oppstdfgp)
```

```python
def gameSim():
    DENScore = (rnd.gauss(DEN_score,DEN_sdpts)+ rnd.gauss(DEN_oppmeanpoints,DEN_oppstdpts))/2
    TORScore = (rnd.gauss(TOR_score,TOR_sdpts)+ rnd.gauss(TOR_oppmeanpoints,TOR_oppstdpts))/2
    if int(round(DENScore)) > int(round(TORScore)):
        return 1
    elif int(round(DENScore)) < int(round(TORScore)):
        return -1
    else: return 0
```

```python
def gamesSim(ns):
    gamesout = []
    team1win = 0
    team2win = 0
    for i in range(ns):
        gm = gameSim()
        gamesout.append(gm)
        if gm == 1:
            team1win +=1 
        elif gm == -1:
            team2win +=1
    print('DEN Win ', round((team1win/(team1win+team2win))*100,2),'%')
    print('TOR Win ', round((team2win/(team1win+team2win))*100,2),'%')
    return gamesout
```

```python
gamesSim(10000)
```
