```python
LAL_meanpoints = LAL1df.TeamPTS.mean()
LAL_meanfgp = LAL1df.TeamFGP.mean()
LAL_sdpts = LAL1df.TeamPTS.std()
LAL_sdfgp = LAL1df.TeamFGP.std()
LAL_oppmeanpoints = LAL1df.OpponentPTS.mean()
LAL_oppfgp = LAL1df.OpponentFGP.mean()
LAL_oppstdpts = LAL1df.OpponentPTS.std()
LAL_oppstdfgp = LAL1df.OpponentFGP.std()

print("LAL Avg. Pts ", LAL_meanpoints)
print("LAL Avg. FG% ", LAL_meanfgp)
print("LAL Pts Deviation ", LAL_sdpts)
print("LAL FG% Deviation ", LAL_sdfgp)
print("LAL Opp. Avg. Pts ", LAL_oppmeanpoints)
print("LAL Opp. Avg. FG% ", LAL_oppfgp)
print("LAL Opp. Pts Deviation ", LAL_oppstdpts)
print("LAL Opp. FG% Deviation ", LAL_oppstdfgp)
```

```python
DEN_meanpoints = DEN1df.TeamPTS.mean()
DEN_meanfgp = DEN1df.TeamFGP.mean()
DEN_sdpts = DEN1df.TeamPTS.std()
DEN_sdfgp = DEN1df.TeamFGP.std()
DEN_oppmeanpoints = DEN1df.OpponentPTS.mean()
DEN_oppfgp = DEN1df.OpponentFGP.mean()
DEN_oppstdpts = DEN1df.OpponentPTS.std()
DEN_oppstdfgp = DEN1df.OpponentFGP.std()

print("DEN Avg. Pts ", DEN_meanpoints)
print("DEN Avg. FG% ", DEN_meanfgp)
print("DEN Pts Deviation ", DEN_sdpts)
print("DEN FG% Deviation ", DEN_sdfgp)
print("DEN Opp. Avg. Pts ", DEN_oppmeanpoints)
print("DEN Opp. Avg. FG% ", DEN_oppfgp)
print("DEN Opp. Pts Deviation ", DEN_oppstdpts)
print("DEN Opp. FG% Deviation ", DEN_oppstdfgp)
```

```python
def gameSim():
    LALScore = (rnd.gauss(LAL_score,LAL_sdpts)+ rnd.gauss(LAL_oppmeanpoints,LAL_oppstdpts))/2
    DENScore = (rnd.gauss(DEN_score,DEN_sdpts)+ rnd.gauss(DEN_oppmeanpoints,DEN_oppstdpts))/2
    if int(round(LALScore)) > int(round(DENScore)):
        return 1
    elif int(round(LALScore)) < int(round(DENScore)):
        return -1
    else: return 0
```

```python
def gamesSim(ns):
    gamesout = []
    team1win = 0
    team2win = 0
    for i in range(ns):
        gm = gameSim()
        gamesout.append(gm)
        if gm == 1:
            team1win +=1 
        elif gm == -1:
            team2win +=1
    print('LAL Win ', round((team1win/(team1win+team2win))*100,2),'%')
    print('DEN Win ', round((team2win/(team1win+team2win))*100,2),'%')
    return gamesout
```
