```python
BOS_meanpoints = BOS1df.TeamPTS.mean()
BOS_meanfgp = BOS1df.TeamFGP.mean()
BOS_sdpts = BOS1df.TeamPTS.std()
BOS_sdfgp = BOS1df.TeamFGP.std()
BOS_oppmeanpoints = BOS1df.OpponentPTS.mean()
BOS_oppfgp = BOS1df.OpponentFGP.mean()
BOS_oppstdpts = BOS1df.OpponentPTS.std()
BOS_oppstdfgp = BOS1df.OpponentFGP.std()

print("BOS Avg. Pts ", BOS_meanpoints)
print("BOS Avg. FG% ", BOS_meanfgp)
print("BOS Pts Deviation ", BOS_sdpts)
print("BOS FG% Deviation ", BOS_sdfgp)
print("BOS Opp. Avg. Pts ", BOS_oppmeanpoints)
print("BOS Opp. Avg. FG% ", BOS_oppfgp)
print("BOS Opp. Pts Deviation ", BOS_oppstdpts)
print("BOS Opp. FG% Deviation ", BOS_oppstdfgp)
```

```python
MIA_meanpoints = MIA1df.TeamPTS.mean()
MIA_meanfgp = MIA1df.TeamFGP.mean()
MIA_sdpts = MIA1df.TeamPTS.std()
MIA_sdfgp = MIA1df.TeamFGP.std()
MIA_oppmeanpoints = MIA1df.OpponentPTS.mean()
MIA_oppfgp = MIA1df.OpponentFGP.mean()
MIA_oppstdpts = MIA1df.OpponentPTS.std()
MIA_oppstdfgp = MIA1df.OpponentFGP.std()

print("MIA Avg. Pts ", MIA_meanpoints)
print("MIA Avg. FG% ", MIA_meanfgp)
print("MIA Pts Deviation ", MIA_sdpts)
print("MIA FG% Deviation ", MIA_sdfgp)
print("MIA Opp. Avg. Pts ", MIA_oppmeanpoints)
print("MIA Opp. Avg. FG% ", MIA_oppfgp)
print("MIA Opp. Pts Deviation ", MIA_oppstdpts)
print("MIA Opp. FG% Deviation ", MIA_oppstdfgp)
```

```python
def gameSim():
    BOSScore = (rnd.gauss(BOS_score,BOS_sdpts)+ rnd.gauss(BOS_oppmeanpoints,BOS_oppstdpts))/2
    MIAScore = (rnd.gauss(MIA_score,MIA_sdpts)+ rnd.gauss(MIA_oppmeanpoints,MIA_oppstdpts))/2
    if int(round(BOSScore)) > int(round(MIAScore)):
        return 1
    elif int(round(BOSScore)) < int(round(MIAScore)):
        return -1
    else: return 0
```

```python
def gamesSim(ns):
    gamesout = []
    team1win = 0
    team2win = 0
    for i in range(ns):
        gm = gameSim()
        gamesout.append(gm)
        if gm == 1:
            team1win +=1 
        elif gm == -1:
            team2win +=1
    print('BOS Win ', round((team1win/(team1win+team2win))*100,2),'%')
    print('MIA Win ', round((team2win/(team1win+team2win))*100,2),'%')
    return gamesout
```
